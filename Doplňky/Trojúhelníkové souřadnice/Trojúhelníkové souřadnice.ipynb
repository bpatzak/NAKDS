# Trojúhelníkové souřadnice

Pro zjednodušení práce s trojúhelníkovými prvky používáme tzv. trojúhelníkové (plošné/přirozené) souřadnice. V tomto parametrickém souřadném systému je každému bodu $P$ přiřazena trojice bezrozměrných souřadnic $\xi_i$, které jsou definovány jako poměr 

$\displaystyle{\xi_i = \frac{A_i}{A} ,}$

kde $A_i$ je plocha trojúhelníka vymezeného dvěma vrcholy prvku $V_j$, $j\not=i$, a daným bodem $P$ a $A$ je plocha celého trojúhelníkového prvku.

<img src="figures/area_coords.svg" width="650"> 

Tyto souřadnice $\xi_1, \xi_2, \xi_3$ jsou vzájemně závislé, platí

$\displaystyle{\label{ksi}
\xi_1 + \xi_2 + \xi_3 = 1.
}$

Z uvedených vztahů je patrné, jak jednoduché je vyjádření stran nebo vrcholů prvku. Pokud bude bod $P$ ležet na jedné ze stran trojúhelníka, bude jedna z ploch $A_i$, a tedy i odpovídající souřadnice $\xi_i$ nulová. Pokud budeme uvažovat bod P ve vrcholech, pokryje jeden z dílčích trojúhelníků celý prvek, zbylé dva trojúhelníky budou mít nulovou plochu. Jednotlivé vrcholy mají tedy souřadnice $[1, 0, 0], [0, 1, 0], [0, 0, 1].$

Ačkoliv se plošné souřadnice zdají být velmi výhodným nástrojem pro práci s trojúhelníkovým prvkem, některé veličiny (např. posunutí, deformace, napětí) jsou spjaty s kartézskými souřadnicemi. Je proto vhodné znát vzájemný vztah souřadnic. Tuto transformaci je možné vyjádřit maticovým zápisem, vyjadřujícím lineární vztah mezi kartézskými a plošnými souřadnicemi doplněný rovnicí vyjadřující, že součet souřadnic je roven jedné 

$\displaystyle{
\left\{
\begin{array}{c}
1\\
x\\
y
\end{array}\right\}
=\left[
\begin{array}{ccc}
1&1&1\\
x^e_1&x^e_2&x^e_3\\
y^e_1&y^e_2&y^e_3
\end{array}\right]\left\{
\begin{array}{c}
\xi_1\\
\xi_2\\
\xi_3
\end{array}\right\}.
}$

Pro převod v opačném směru lze použít inverzní vztah

$\displaystyle{
\left\{
\begin{array}{c}
\xi_1\\
\xi_2\\
\xi_3
\end{array}\right\}
=\frac{1}{2A}\left[
\begin{array}{ccc}
x^e_2 y^e_3 - x^e_3 y^e_2 & y^e_2 - y^e_3 & x^e_3 - x^e_2\\
x^e_3 y^e_1 - x^e_1 y^e_3 & y^e_3 - y^e_1 & x^e_1 - x^e_3\\
x^e_1 y^e_2 - x^e_2 y^e_1 & y^e_1 - y^e_2 & x^e_2 - x^e_1
\end{array}
\right]\left\{
\begin{array}{c}
1\\
x\\
y
\end{array}
\right\}.
}$

Derivace funkce $f(\xi_1, \xi_2, \xi_3)$ podle $x,y$ je definována na základě věty o derivaci složené funkce. Platí

$\displaystyle{
\frac{\partial f}{\partial x}=\frac{\partial f}{\partial \xi_1}\frac{\partial \xi_1}{\partial x}+\frac{\partial f}{\partial \xi_2}\frac{\partial \xi_2}{\partial x}+\frac{\partial f}{\partial \xi_3}\frac{\partial \xi_3}{\partial x}=\frac{1}{2A}\left( \frac{\partial f}{\partial \xi_1}y_{23}+\frac{\partial f}{\partial \xi_2}y_{31}+\frac{\partial f}{\partial \xi_3}y_{12}\right),
}$

$\displaystyle{
\frac{\partial f}{\partial y}=\frac{\partial f}{\partial \xi_1}\frac{\partial \xi_1}{\partial y}+\frac{\partial f}{\partial \xi_2}\frac{\partial \xi_2}{\partial y}+\frac{\partial f}{\partial \xi_3}\frac{\partial \xi_3}{\partial y}=\frac{1}{2A}\left( \frac{\partial f}{\partial \xi_1}x_{32}+\frac{\partial f}{\partial \xi_2}x_{13}+\frac{\partial f}{\partial \xi_3}x_{21}\right).
}$

V maticovém zápisu


$\displaystyle{
\left\{
\begin{array}{c}
\frac{\partial f}{\partial x}\\
\frac{\partial f}{\partial y}
\end{array}
\right\} 
= \frac{1}{2A}
\left[
\begin{array}{ccc}
y_{23}&y_{31}&y_{12}\\
x_{32}&x_{13}&x_{21}
\end{array}
\right]
\left\{
\begin{array}{c}
\frac{\partial f}{\partial \xi_1}\\
\frac{\partial f}{\partial \xi_2}\\
\frac{\partial f}{\partial \xi_3}
\end{array}
\right\}.
}$

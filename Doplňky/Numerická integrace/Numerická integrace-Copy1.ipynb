## Numerická integrace

Slabé řešení vyžaduje výpočet integrálů, jen výjimečně lze provést integraci analyticky, proto se používá integrace numerická. Rozlišujeme různé metody numerické integrace, my se zaměříme na Gaussovu numerickou integraci, která je zvláště výhodná pro polynomiální funkce.

### Gaussova numerická integrace
 
#### Princip 

Uvažujme následující integrál: 

$\displaystyle{I=\int_{-1}^{1}f(\xi)d\xi}$

Hodnotu integrálu budeme aproximovat jako 

$\displaystyle{\hat{I}=\sum_{i=1}^{n}w_if(\xi_i)}$


Idea Gaussovy integrace spočívá v tom, že se snažíme stanovit hodnoty vah $w_i$ a souřadnic integračních bodů $\xi_i$ tak, abychom integrovali přesně polynom co nejvyššího řádu. 
Máme tedy celkem $2n$ parametrů, které můžeme zvolit.
Důsledkem toho je, že máme-li $n$ integračních bodů, pak můžeme přesně integrovat polygon řádu $p\le2n-1$.
Nutný počet integračních bodů pro přesnou integraci polynomu řádu p je tedy $n \ge \del{p+1}{2}$.



Uvažujme případ $n=1$:

$\displaystyle{w_{1}f(\xi_1) \approx \int_{-1}^{1}f(\xi)\ d\xi}$

  Hledáme takovou váhu $w_1$ a polohu $\xi_1$ aby integrace byla přesná pro polynomy co největšího (prvního) stupně.
  Proto uvažujeme $f(\xi)=a\xi+b$, kde $a,b\in R$.
  Tedy máme

$\displaystyle{\int_{-1}^{1}f(\xi)\ d\xi = w_1f(\xi_1) \Rightarrow \int_{-1}^{1}(a\xi+b)\ d\xi = w_1(a\xi_1+b)}$

$\displaystyle{\Rightarrow [a\del{\xi^2}{2}+b\xi]_{-1}^{1} = w_1(a\xi_1+b)\Rightarrow 0\cdot a+2\cdot b = a\xi_1 w_1 + bw_1}$

Aby bylo splněno pro libovolný polynom 1 řádu ($a,b\ne 0$):

$\displaystyle{
\begin{eqnarray}
    0 &=& \xi_1w_1\\
    2 &=& w_1
  \end{eqnarray}
}$

Hledané řešení je tedy

$\displaystyle{\xi_1 = 0,w_1 = 2,\ \int_{-1}^{1} f(\xi)\ d\xi \approx 2f(0)}$


  Uvažujme případ $n=2$: 
  
  $\displaystyle{w_{1}f(\xi_1)+w_2 f(\xi_2) \approx \int_{-1}^{1}f(\xi)\ dx}$
  
  Hledáme takové váhy $w_1, w_2$ a polohy $\xi_1, \xi_2$ aby integrace byla přesná pro polynomy co největšího stupně.
  Proto uvažujeme $f(\xi)=a\xi^3+b\xi^2+c\xi+d$
  Dosazením postupně dostáváme:
  
  $\displaystyle{\int_{-1}^{1}f(\xi)\ d\xi = w_{1}f(\xi_1)+w_2 f(\xi_2)\Rightarrow}$
  
  $\displaystyle{\int_{-1}^{1}[a\xi^3+b\xi^2+c\xi+d]d\xi=w_1[a\xi_1^3+b\xi_1^2+c\xi_1+d]+w_2[a\xi_2^3+b\xi_2^2+c\xi_2+d]\Rightarrow}$
  
  $\displaystyle{[\del{a\xi^4}{4}+\del{b\xi^3}{3}+\del{c\xi^2}{2}+d\xi]_{-1}^{1} = w_1[a\xi_1^3+b\xi_1^2+c\xi_1+d]+w_2[a\xi_2^3+b\xi_2^2+c\xi_2+d]\Rightarrow}$
  
  $\displaystyle{a[w_1\xi_1^3+w_1\xi_2^3]+b[w_1\xi_1^2+w_2\xi_2^2-2/3]+c[w_1\xi_1+w_2\xi_2]+d[w_1+w_2-2] = 0}$
  
  
  Aby bylo splněno pro libovolný polynom 3 řádu (tedy pro $a,b,c,d\ne 0$) musí platit:
  
  $\displaystyle{
  \begin{eqnarray}
    w_1\xi_1^3+w_2\xi_2^3 &=& 0\\
    w_1\xi_1^2+w_2\xi_2^2-2/3 &=& 0\\
    w_1\xi_1+w_2\xi_2 &=& 0\\
    w_1+w_2-2 &=& 0
  \end{eqnarray}
  }$
  
  Hledané řešení je
  
  $\displaystyle{w_1=w_2=1,\ \ \xi_1=\del{-1}{\sqrt{3}}, \xi_2=\del{1}{\sqrt{3}}}$


  Ve skutečnosti se tento algoritmus pro stanovení vah a souřadnic bodů nepoužívá.
  Lze ukázat (teorie ortogonálních polynomů), že pro dosažení optimální přesnosti (tj přesné integrace polynomu do řádu $2n-1$ včetně na intervalu $[-1, 1]$ je třeba souřadnice integračních bodů $x_1,\cdots,x_n$ stanovit jako kořeny Legendreova polynomu $P_n(x)$ řádu $n$.
  Legendreovy polynomy jsou definovány rekurzivní formulí

$\displaystyle{P_{k+1}(x)=\del{2k+1}{k+1}xP_k(x)-\del{k}{k+1}P_{k-1}(x),\ \rm{kde}\ P_0(x)=1,\ P_1(x)=x}$

  Integrační váhy jsou pak dány následujícím vztahem

$\displaystyle{w_i=\int_{-1}^{1}\prod_{k=1, k\ne i}^n \del{x-x_i}{x_k-x_i}\ dx }$

#### Numerická integrace na obdélníkové oblasti
 
Numerická integrace v 1D:
 
 $\displaystyle{\int^1_{-1}F(\xi)\ d\xi\approx\sum_{i=1}^Mw_i\ F(\xi_i)}$
 
 $\displaystyle{\int^a_bF(x)\ dx=\int^1_{-1}F(\xi)J\ d\xi\approx\sum_{i=1}^Mw_i\ F(\xi_i) J(\xi_i),}$
 
 kde např. pro lineární aproximaci geometrie 
    
 $\displaystyle{J=dx/d\xi=\del{d}{dx}(0.5(1-\xi)a+0.5(1+\xi)b)=0.5(b-a)=0.5l.}$
  
 Integraci na obdélníku lze získat jako produkt jednorozměrných integrací v příslušných směrech
    
 $\displaystyle{
    \begin{array}{rcl}
      I&=&\int^1_{-1}\int^1_{-1}F(\xi,\eta)\ d\xi d\eta\\
      &\approx&\int^1_{-1}\left(\sum_{j=1}^Mw_jF(\xi,\eta_j)\right)d\xi\\
      &\approx&\sum_{i=1}^{M}\sum_{j=1}^Mw_iw_jF(\xi_i,\eta_j)\\
      &=&\sum_{i=1}^{M}\sum_{j=1}^Mw_{ij}F(\xi_i,\eta_j)
    \end{array}
    }$
 

 Numerická integrace ve 2D:

$\displaystyle{\mbf{K}^e=\int_{\Omega}F(x,y) d\Omega = \int_{-1}^1\int_{-1}^1F(\xi,\eta) J\ d\xi d\eta}$

Plocha diferenciálního elementu:

$\displaystyle{
\begin{array}{rcl}
\Omega&=&\vert\vec{OA}\times\vec{OC}\vert\\
 &=& \del{\partial x}{\partial\xi}d\xi\del{\partial y}{\partial\eta}d\eta-\del{\partial x}{\partial\eta}d\eta\del{\partial y}{\partial\xi}d\xi \\
&=& \left|\begin{array}{cc}\pard{x}{\xi}& \pard{x}{\eta}\\ \pard{y}{\xi} & \pard{y}{\eta}\end{array}\right|\ d\xi d\eta=\vert J\vert d\xi d\eta
\end{array}
}$

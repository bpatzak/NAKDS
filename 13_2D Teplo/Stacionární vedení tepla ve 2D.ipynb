# Stacionární vedení tepla ve 2D



Řešení stacionárního vedení tepla v rovině je rozšířením jednorozměrné úlohy do dvou dimenzí $(x, y)$.

   * Gradient teploty v bodě $(x, y)$

$\qquad \displaystyle{
   \left[ 
   \begin{array}{c}
   {  \frac{\partial T}{\partial x} }(x,y)\\
   \frac{\partial T}{\partial y}(x,y)
   \end{array}   
   \right] =
   \left[ 
   \begin{array}{c}
   \frac{\partial}{\partial x}\\
   \frac{\partial}{\partial y}
   \end{array}   
   \right]  {T}(x,y) = 
   {\boldsymbol \nabla} T(\boldsymbol x)
   }$
   
## Diferenciální rovnice problému

Diferenciální rovnice úlohy vedení tepla plyne z <b>bilance energie</b>. Uvnitř tělesa (pro $x \in \Omega$)

<img src="Figures/bilance2.png" width=300>

$\qquad \displaystyle{
\begin{array}{ccc}
-q_{x}( x +\Delta x,y ) \Delta y + q_{x}(x,y) \Delta y
&  & 
( \rightarrow x ) \\
-q_{y}( x, y + \Delta y ) \Delta x + q_{y}(x,y) \Delta x 
&  & 
( \uparrow y ) 
\\
+ 
\overline{Q}\left(x + \frac{\Delta x}{2}, y + \frac{\Delta y}{2} \right) \Delta x \Delta y & = & 0. 
\end{array}
}$

Podělením $\Delta x \Delta y$ a limitním přechodem $\Delta x \rightarrow 0, \Delta y \rightarrow 0$:

$\qquad \displaystyle{ \begin{array}{rcl} - \frac{\partial q_x}{\partial x}(x,y) - \frac{\partial q_y}{\partial y}(x,y) + \overline{Q}( x,y ) &=& 0 \end{array}}$

což můžeme maticově zapsat jako

$\qquad \displaystyle{ \begin{array}{rcl} -\left[ {\frac{\partial}{\partial x} \qquad \frac{\partial}{\partial y}}\right]
\left[ \begin{array}{c} q_x(x,y) \\q_y(x,y)\end{array}\right] +
\overline{Q}(x,y) & = & 0 \\ - \boldsymbol{\nabla}^T \boldsymbol{q}(\boldsymbol{x}) + \overline{Q}(\boldsymbol{x}) & = & 0
\;\rm{   pro } \boldsymbol{x} \in \Omega \end{array}
}$



Konstitutivní rovnice (Fourierův zákon) (pro $\boldsymbol{x} \in \Omega$)
  
$\qquad \displaystyle{ \begin{array}{rcl} \left[ \begin{array}{c} q_x(x,y) \\ q_y(x,y) \end{array} \right]
& = & - \left[ \begin{array}{cc}
\lambda_{xx}(x,y) & \lambda_{xy}(x,y) \\
\lambda_{yx}(x,y) & \lambda_{yy}(x,y)
\end{array} \right]
\left[
\begin{array}{c}
\frac{\partial T}{\partial x}(x,y) \\
\frac{\partial T}{\partial y}(x,y)
\end{array}
\right]
\\
\boldsymbol{q}(\boldsymbol{x}) & = & - \boldsymbol{\lambda}  \boldsymbol{\nabla}T(\boldsymbol{x}) \end{array}
}$

kde $\boldsymbol{\lambda}$ je symetrická ($\lambda_{xy} = \lambda_{yx}$) a
  pozitivně definitní matice tepelných vodivostí. Pro izotropní materiál
 
$\qquad \displaystyle{
\boldsymbol{\lambda}(\boldsymbol{x})=
\left[
\begin{array}{cc}
\lambda(\boldsymbol{x}) & 0 \\ 0 & \lambda(\boldsymbol{x})
\end{array}
\right]
= \lambda(\boldsymbol{x}) \boldsymbol{I}
}$

### Okrajové podmínky

  * Dirichletovské okrajové podmínky: 
  
$\qquad\displaystyle{
\begin{array}{rcl}
T(\boldsymbol{x}) = \overline{T}(\boldsymbol{x}) &
\mbox{pro} & \boldsymbol{x} \in \Gamma_T
\end{array}}
$

  * Energetická bilance na hranici
  
$\qquad\displaystyle{
\begin{array}{rcl}
q_x(x,y) n_x(x,y) + q_y(x,y) n_y(x,y) 
& = & \overline{q}_{\boldsymbol{n}}(x,y) \mbox{ pro } (x,y) \in \Gamma_{qp}
\\
\left[\begin{array}{cc}n_x(\boldsymbol{x}) & n_y(\boldsymbol{x})\end{array}\right]
\left[\begin{array}{c}q_x(\boldsymbol{x}) \\ q_y(\boldsymbol{x})\end{array}\right]
& = & 
\overline{q}_{\boldsymbol{n}}(\boldsymbol{x})
\\
\boldsymbol{n}^T(\boldsymbol{x}) \boldsymbol{q}(\boldsymbol{x}) 
& = &
\overline{q}_{\boldsymbol{n}}(\boldsymbol{x})
\end{array}}
$

  * Neumannova okrajová podínka: je dáno $\overline{q}_{\boldsymbol{n}}(\boldsymbol{x})\;\mbox{pro}\;\boldsymbol{x}\in\Gamma_{qc}$ 

  * Smíšená okrajová podmínka (Mixed/Robin/Newton):
  
$\qquad\displaystyle{
\begin{array}{rcl}
\overline{q}_{\boldsymbol{n}}( \boldsymbol{x} ) = \alpha( \boldsymbol{x} ) \left( T( \boldsymbol{x} ) -
T_0( \boldsymbol{x} ) \right),
& \mbox{pro} & \boldsymbol{x} \in \Gamma_{qc}
\end{array}}
$

$\qquad$ kde $\alpha$ [Jm$^{-2}$K$^{-1}$s$^{-1}$] je koeficient přestupu tepla a
$T_0$ je teplota okolního prostředí

  * Radiace (Newton)
  
$\qquad\displaystyle{
\begin{array}
\overline{q}_{\boldsymbol n}( \boldsymbol{x} ) = \varepsilon( \boldsymbol{x} ) \sigma( \boldsymbol{x} )
\left( T^4( \boldsymbol{x} ) - T^4_{\infty} \right)
& \mbox{pro} & \boldsymbol{x} \in \Gamma_{qr}
\end{array}}
$

Význam koeficientů obdobný jako pro [jednorozměrnou úlohu](../12_1D%20Teplo/Stacionární%20vedení%20tepla%20v%201D.ipynb).

### Slabé řešení 

Hledáme $T(\boldsymbol{x})$ (dostatečně hladké), takové aby splňovalo okrajové podmínky (nelineární okrajovou podmínku (radiaci) neuvažujeme). Pro libovolnou váhovou funkci $\delta T(\boldsymbol{x})$, která splňuje $\delta
  T(\boldsymbol{x})=0$ pro $\boldsymbol{x} \in \Gamma_T$:
  
$\qquad \displaystyle{
\int_\Omega 
\delta T( \boldsymbol{x} ) 
\left( 
\boldsymbol{\nabla}^T
\bigl(
\boldsymbol{\lambda}(\boldsymbol{x}) \boldsymbol{\nabla}T(\boldsymbol{x}) 
\bigr)
+
\overline{Q}(\boldsymbol{x})
\right)
\ d\boldsymbol{x}
= 0}
$

Integrace per-partes ve dvou dimenzích (Divergenční teorém):

$\qquad \displaystyle{ \begin{array}{rcl} \int_\Omega g( \boldsymbol{x}) \frac{\partial f_x}{\partial x}(\boldsymbol{x})  \ d\boldsymbol{x} & = &
\int_\Gamma g(\boldsymbol{x}) n_x(\boldsymbol{x}) f_x(\boldsymbol{x})  \ ds - 
\int_\Omega \frac{\partial g}{\partial x}(\boldsymbol{x})  f_x(\boldsymbol{x}) \ d\boldsymbol{x} \\
\int_\Omega g( \boldsymbol{x}) \frac{\partial f_y}{\partial y}(\boldsymbol{x})  \ d\boldsymbol{x} & = &
\int_\Gamma g(\boldsymbol{x}) n_y(\boldsymbol{x}) f_y(\boldsymbol{x})  \ ds -
\int_\Omega \frac{\partial g}{\partial y}(\boldsymbol{x})  f_y(\boldsymbol{x}) \ d\boldsymbol{x} \\
\hline
\int_\Omega g(\boldsymbol{x}) \boldsymbol{\nabla}^T \boldsymbol{f}(\boldsymbol{x}) \ d\boldsymbol{x} & = &
\int_\Gamma g(\boldsymbol{x}) \boldsymbol{n}^T(\boldsymbol{x}) \boldsymbol{f}(\boldsymbol{x}) \ ds -
\int_\Omega \left( \boldsymbol{\nabla} g(\boldsymbol{x}) \right)^T \boldsymbol{f}(\boldsymbol{x}) \ d\boldsymbol{x}
\end{array}}
$

Aplikací divergenčního teorému na úlohu vedení tepla ve 2D ($g \equiv \delta T, \boldsymbol{f} \equiv \boldsymbol{\lambda}
  \boldsymbol{\nabla} T$) 

$\qquad \displaystyle{ \begin{array}{rcl} 0 & = &
\int_\Gamma  \delta T( \boldsymbol{x} )    \boldsymbol{\lambda}(\boldsymbol{x}) \boldsymbol{\nabla} T(\boldsymbol{x})  \boldsymbol{n}(\boldsymbol{x}) \ ds -
\int_\Omega  \left( \boldsymbol{\nabla} \delta T( \boldsymbol{x}) \right)^T \boldsymbol{\lambda}(\boldsymbol{x})  \boldsymbol{\nabla} T( \boldsymbol{x}) \ d\boldsymbol{x} \\
& + & \int_\Omega   \delta T( \boldsymbol{x} )    \overline{Q}(\boldsymbol{x}) \ d\boldsymbol{x}
\end{array}}
$

Hraniční integrál můžeme dále rozdělit na část hranice $\Gamma_T$ a $\Gamma_q$

$\qquad \displaystyle{
\begin{array}{rcl}
\int_\Gamma
 \delta T( \boldsymbol{x} )  
 \boldsymbol{\lambda}(\boldsymbol{x}) \boldsymbol{\nabla} T(\boldsymbol{x})
 \boldsymbol{n}(\boldsymbol{x})
\ ds
& = &
\int_{\Gamma_T}
 \overbrace{\delta T( \boldsymbol{x} )}^{=0}
 \boldsymbol{\lambda}(\boldsymbol{x}) \boldsymbol{\nabla} T(\boldsymbol{x})
 \boldsymbol{n}(\boldsymbol{x})
\ ds \\
& + &
\int_{\Gamma_q}
 \delta T( \boldsymbol{x} )  
\underbrace{\boldsymbol{\lambda}(\boldsymbol{x}) \boldsymbol{\nabla} T(\boldsymbol{x})
 \boldsymbol{n}(\boldsymbol{x})}_{=-q_{\boldsymbol{n}}}
\ ds
\end{array}
}$




Hledáme tedy $T(x)$ (dostatečně integrovatelné), takové aby:

$
\qquad\displaystyle{
\begin{array}{rcl} \int_\Omega \left(\boldsymbol{\nabla} \delta T(\boldsymbol{x})\right)^T \lambda(x)  \boldsymbol{\nabla} T(x) \ d\boldsymbol{x} +
\int_{\Gamma_{qc}}  \delta T( \boldsymbol{x} ) \alpha( \boldsymbol{x} ) T( \boldsymbol{x} ) \ ds &=& 
\int_{\Gamma_{qc}}  \delta T( \boldsymbol{x} )    \alpha( \boldsymbol{x} ) T_0( \boldsymbol{x} ) \ ds -
\int_{\Gamma_{qp}}  \delta T( \boldsymbol{x} )    \overline{q}_{\boldsymbol{n}}(\boldsymbol{x}) \ ds +
\int_{\Omega}  \delta T(\boldsymbol{x})  \overline{Q}(\boldsymbol{x}) \ d\boldsymbol{x}
\end{array}
}$

pro všechna $\delta T(\boldsymbol{x})$ taková, aby $\delta T=0$ na $\Gamma_T$.


## Diskretizovaná úloha
$\displaystyle{ \left(\boldsymbol{K}^e_{\Omega} + \boldsymbol{K}^e_\Gamma\right)r^e =
f^e_{\Gamma_c} + f^e_{\Gamma_p} + f^e_{\Omega}    }$

* Levá strana
    * Matice vodivosti $\displaystyle{K^e_{\Omega}}$

     $\displaystyle{\boldsymbol{K}^e_{\Omega} = \int_{\Omega} \boldsymbol{B}^{eT}\lambda\  \boldsymbol{B}^{e} \ d\Omega }$

    * Příspěvek do levé strany od přestupu tepla $\displaystyle{K^e_{\Gamma}}$

     $\displaystyle{\boldsymbol{K}^e_{\Gamma} =\int_{\Gamma_c} \boldsymbol{N}^{eT}\alpha^e \boldsymbol{N}^e\ d\Omega}$

* Pravá strana
    * Přestup tepla  $\displaystyle{f^e_{\Gamma_c}}$
     
     $\displaystyle{ f^e_{\Gamma_c} = \int_{\Gamma_c}\boldsymbol{N}^{eT}\alpha^e T_0^e(x)  \ d\Omega}$

    * Předepsaný tok  $\displaystyle{f^e_{\Gamma_p}}$

     $\displaystyle{f^e_{\Gamma_p} = - \int_{\Gamma_p}\boldsymbol{N}^{eT}  \bar{q}^e(x)  \ d\Omega}$

    * Vnitřní zdroj  $\displaystyle{ f^e_{\Omega} }$
    
     $\displaystyle{ f^e_{\Omega} = \int_{\Omega} \boldsymbol{N}^{eT}  Q^e(x) \ d\Omega    }$

### MKP program

```octave
% funkce pocita plochu trojuhelnikoveho prvku
% In: 
%   xe - vektor x-ovych souradnic uzlu prvku (x1,x2,x3)
%   ye - vektor y-ovych souradnic uzlu prvku (y1,y2,y3)
%
% Out:
%   Ae - plocha prvku
%
function Ae = area_triangle (xe,ye) 

Ae=(1/2)*((xe(2)*ye(3)-xe(3)*ye(2))-(xe(1)*ye(3)-xe(3)*ye(1))+(xe(1)*ye(2)-xe(2)*ye(1))) ;
end

```

```octave
% funkce pocita delky stran trojuhelnikoveho prvku
% In: 
%   xe - vektor x-ovych souradnic uzlu prvku (x1,x2,x3)
%   ye - vektor y-ovych souradnic uzlu prvku (y1,y2,y3)
%
% Out:
%   le - delky hran prvku (3,1)
%
function le = length_triangle (xe,ye) 

le = zeros(3,1) ;
le(1) = sqrt((xe(2)-xe(1))*(xe(2)-xe(1))+(ye(2)-ye(1))*(ye(2)-ye(1)));
le(2) = sqrt((xe(3)-xe(2))*(xe(3)-xe(2))+(ye(3)-ye(2))*(ye(3)-ye(2)));
le(3) = sqrt((xe(1)-xe(3))*(xe(1)-xe(3))+(ye(1)-ye(3))*(ye(1)-ye(3)));
end

```

```octave
% funkce pro lokalizaci matice tuhosti prvku 
% do matice tuhosti konstrukce
% vyzaduje existenci globalni matice kodovych cisel - lm
% 
%
% Vstup:
%   k - matice tuhosti konstrukce
%   ke- matice tuhosti prvku
%   lm- matice kodovych cisel
%   e - cislo prvku 
%
% Vystup:
%   k - matice tuhosti konstrukce
%
function k = assembly (k,ke,lm,e)

ndof = size(ke,1);
for i=1:ndof
        ia = lm(e,i);
        if ia ~=0 
            for j=1:ndof
                ja=lm(e,j);
                if ja ~=0
                  k(ia,ja)=k(ia,ja)+ke(i,j);
                end   
            end
        end
    end
end
```

```octave
% funkce pro lokalizaci matice prave strany 
% vyzaduje existenci globalni matice kodovych cisel - lm
% 
%
% Vstup:
%   f - matice prave strany
%   fe- matice prave strany prvku
%   lm- matice kodovych cisel
%   e - cislo prvku 
%
% Vystup:
%   f - matice prave strany
%
function f = assemblyf (f,fe,lm,e)

ndof = size(fe,1);
for i=1:ndof
        ia = lm(e,i);
        if ia ~=0 
           f(ia,1)=f(ia,1)+fe(i,1);
        end
    end
end
```

```octave
% funkce pocita matici vodivosti (conductivity matrix) 2d prvku s linearni aproximaci
% In: 
%   xe - vektor x-ovych souradnic uzlu prvku (x1,x2,x3)
%   ye - vektor y-ovych souradnic uzlu prvku (y1,y2,y3)
%   c- koeficient teplotni vodivosti
%
% Out:
%   ke - matice vodovosti prvku (3,3)
%
function ke = heat2_cond (xe,ye,c) 

Ae = area_triangle(xe,ye) ;

Be = (1/(2*Ae))*[(ye(2)-ye(3)), (ye(3)-ye(1)), (ye(1)-ye(2));
                 (xe(3)-xe(2)), (xe(1)-xe(3)), (xe(2)-xe(1))];

ke = c*Ae*Be'*Be;

end
```

```octave
% funkce pocita matici prave strany od predepsaneho konstatniho toku (heat flux) 2d prvku s linearni aproximaci
% In: 
%   xe - vektor x-ovych souradnic uzlu prvku (x1,x2,x3)
%   ye - vektor y-ovych souradnic uzlu prvku (y1,y2,y3)
%   q- predepsany tok
%   b1,b2,b3 - hranice, kde je predepsany tok, voli se hodnoty 0 nebo 1
%
% Out:
%   ff - matice prave strany (3,1)
%
function ff = heat2_flux (xe,ye,q,b1,b2,b3) 

le = length_triangle(xe,ye);

ff = -q*(b1*[ le(1)/2 ; le(1)/2 ; 0] + b2*[ 0 ; le(2)/2 ; le(2)/2] + b3*[ le(3)/2 ; 0 ; le(3)/2 ]);
end

```

```octave
% funkce pocita matici prave strany od vlivu tepla (heat source) 2d prvku s linearni aproximaci
% In: 
%   xe - vektor x-ovych souradnic uzlu prvku (x1,x2,x3)
%   ye - vektor y-ovych souradnic uzlu prvku (y1,y2,y3)
%   Q- teplo
%
% Out:
%   fe - matice prave strany (3,1)
%
function fe = heat2_source (xe,ye,Q) 

Ae = area_triangle(xe,ye) ;

fe = (Q*Ae/3)*[1; 1; 1]; 
end
```

```octave
% reseni 

%pocet prvku
nelem=2;
lambda =5.0;
Q = 6.0;
q2 = 20;

% pole kodovych cisel
lm = [1 2 3; 2 4 3];
% pocet neznamych
nu = 4;

% pole souradnic
x =[0.0, 2.0, 0.0, 2.0];
y =[0.0, 0.5, 1.0, 1.0];

ke = zeros(3,3);
k = zeros(nu,nu);
f = zeros(nu,1);

for i = 1:nelem
    xe = [x(lm(i,:))] ;
    ye = [y(lm(i,:))] ;
    ke = heat2_cond(xe,ye,lambda);
    k=assembly(k,ke,lm,i);
    fe = heat2_source(xe,ye,Q);
    f=assemblyf(f,fe,lm,i);
end

ff2 = heat2_flux (x(lm(2,:)),y(lm(2,:)),q2,0,1,0);
f=assemblyf(f,ff2,lm,2);

%reseni teploty
kuu=k(4:4,4:4);
fuu=f(4:4);
u=kuu\fuu;
u;

t = zeros(4,1);
t(4)=u

% vykresleni 

%figure; 
% in MATLAB/Octave 'patch' is the convenient function for plotting results:
%for i = 1:nelem
%  xe = [x(lm(i,:))] ;
%  ye = [y(lm(i,:))] ;
%  te = [t(lm(i,:))] ;
%  XX = [xe(1) xe(2) xe(3) xe(1)];
%  YY = [ye(1) ye(2) ye(3) ye(1)];
%  tt = [te(1) te(2) te(3) te(1)]
%  patch(XX,YY,tt);hold on;  
%end
%title('Temperature distribution'); xlabel('X'); ylabel('Y'); colorbar;


% in Jupyter environment it is not possible to interpolate colors, 
% therefore 'sccater' is used to illustrate results 
figure;
for i = 1:nelem
  xe = [x(lm(i,:))]; 
  ye = [y(lm(i,:))]; 
  te = [t(lm(i,:))]; 
  n = 75;
  xi = eta = 0:1/n:1;
  X = []; Y = []; T = [];
  help = 1;
  for i = 1:n+1
    for j = 1:n+2-i
       N = [xi(i), eta(j), 1-xi(i)-eta(j)];
       X(help) = dot(N,xe);
       Y(help) = dot(N,ye);
       T(help) = dot(N,te);
       help++;
    end
  end
  scatter(X,Y,15000/n,T(:),'filled');
  hold on;  
end
title('Temperature distribution'); xlabel('X'); ylabel('Y'); colorbar;axis equal;
```

# Modelování pružného podloží

Pro modelování interakce konstrukce se základy pořebujeme formulovat model podloží. V praxi se můžeme setkat s následujícími modely popisujícímy chování podloží jako pružné vrstvy:

 * Boussinesqův model (pružný poloprostor): homogenní izotropní
podloží, charakterizováno dvěma materiálovými parametry E a ν. Plně trojrozměrný model

 $\displaystyle{
u(x, y, z) \neq 0,\quad v(x, y, z) \neq 0,\quad w(x, y, z) \neq 0
  }$


 * Westergaardův model (pružný poloprostor): homogenní anisotropní podloží, charakterizováno parametry E a ν. Kinematické předpoklady
 
 $\displaystyle{
u(x, y, z) = 0,\quad
v(x, y, z) = 0,\quad
w(x, y, z) \neq 0
}$


 * Model pružné vrstvy: založen na představě deformační zóny

## Winkler-Pasternakův model podloží

<img src="Figures/podlozi.png" width = 500>

Představíme si dvouparametrický ($E,\nu$) model, který je vhodný pro modelování podkladu tvořeného ohraničenou pružnou vrstvou tloušťky $h$. 

### Kinematické předpoklady
 * Posuny u a v jsou zanedbatelné vůči posunu w

 $\displaystyle{
u(x, y, z) = 0,\\
v(x, y, z) = 0.
}$


 * Posun w lze vyjádřit v závislosti na posunu povrchu

 $\displaystyle{
w(x, y, z) = w(x, y, 0)\psi(z) = w(x, y)\psi(z),
}$

   kde funkce $\psi(z)$ závisí na materiálových vlastnostech podloží a vlastnostech základové konstrukce. Vzhledem ke značné neurčitosti vstupních dat v geotechnických problémech postačuje u tenkých vrstev uvažovat lineární průběh. V každém případě $\psi$ splňuje podmínky

   $\displaystyle{
\psi(0) = 1,\\
\psi(h) = 0.
}$

### Geometrické rovnice
 *  Nenulové složky tenzoru deformace

  $\displaystyle{
\begin{array}{rcl}
\varepsilon_z(x,y,z) &=&\displaystyle{ \frac{\partial w}{\partial z} = \frac{\partial }{\partial z}(w(x,y)\psi(z)) = w(x,y)\frac{d\psi(z)}{d z}}\\
\gamma_{zx}(x,y,z) &=& \displaystyle{\frac{\partial w}{\partial x}+\frac{\partial u}{\partial z} = \frac{\partial w(x,y)}{\partial x}\psi(z) }\\
\gamma_{zy}(x,y,z) &=&  \displaystyle{\frac{\partial w}{\partial y}+\frac{\partial v}{\partial z} = \frac{\partial w(x,y)}{\partial y}\psi(z)}
\end{array}
}$


 * Ostatní složky

  $\displaystyle{
  \begin{array}{rcl}
\varepsilon_x(x,y,z) &=& 0,\\
\varepsilon_y(x,y,z) &=& 0,\\
\gamma_{xy}(x,y,z) &=& 0
\end{array}
}$


 * Kompaktní zápis
 
  $\displaystyle{
  \begin{array}{rcl}
\varepsilon_z(x,y,z) &=&\displaystyle{w(x,y)\frac{d\psi(z)}{dz}}\\
\left\{
\begin{array}{r}
\gamma_{zx}(x,y,z)\\
\gamma_{zy}(x,y,z)
\end{array}
\right\}
&=&
\left\{
\begin{array}{r}
\displaystyle{\frac{\partial}{\partial x}}\\
\displaystyle{\frac{\partial}{\partial y}}
\end{array}
\right\}
w(x,y)\psi(z)\\
\boldsymbol{\gamma}(x,y,z) &=& \boldsymbol{\nabla} w(x,y)\psi(z)
\end{array}
}$

### Konstitutivní rovnice

Pro jednoduchost nebudeme v předložené formulaci uvažovat vliv počátečních deformací.

 * Normálová napětí 
 
  $\displaystyle{
\begin{array}{rcl}
\sigma_z(x,y,z) &=& \lambda(x,y,z)(1-\nu(x,y,z))\varepsilon_z(x,y,z)\\
&=& E_{oed}(x,y,z)\varepsilon_z(x,y,z),
\end{array}
}$

  kde $E_{oed}$ je oedometrický modul (experimentálně měřen v oedometrické nádobě, která brání boční deformaci $\Rightarrow$ jednoosá napjatost). 


 * Smyková napětí
 
  $\displaystyle{
\begin{array}{rcl}
\tau_{zx}(x,y,z) &=& G_x(x,y,z)\gamma_{zx}(x,y,z)\\
\tau_{yz}(x,y,z) &=& G_y(x,y,z)\gamma_{yz}(x,y,z)
\end{array}
}$


 * Kompaktní zápis
 
  $\displaystyle{
  \begin{array}{rcl}
  \left\{
\begin{array}{r}
\tau_{zx}(x,y,z)\\
\tau_{yz}(x,y,z)
\end{array}
\right\}
&=&
\left[
\begin{array}{cc}
G_x(x,y,z) & 0\\
0 & G_y(x,y,z)
\end{array}
\right]
\left\{
\begin{array}{r}
\gamma_{zx}(x,y,z)\\
\gamma_{yz}(x,y,z)
\end{array}
\right\}\\
\boldsymbol{\tau}(x,y,z) &=& \boldsymbol{G} (x,y,z)\boldsymbol{\gamma}(x,y,z)
\end{array}
}$


### Statické rovnice
 * Všechny nenulové složky napětí působí ve směru osy z ⇒ jediná podmínka rovnováhy

  $\displaystyle{\frac{\partial\tau_{zx}(x,y,z)}{\partial x}+\frac{\partial\tau_{yz}(x,y,z)}{\partial y}+\frac{\partial\sigma_{z}(x,y,z)}{\partial z}+\bar{Z}(x,y,z) = 0}$


 * Kompaktní zápis
  
  $\displaystyle{\begin{array}{rcl}  \left\{ \begin{array}{cc} \displaystyle{\frac{\partial}{\partial x}}\\
  \displaystyle{\frac{\partial}{\partial y}} \end{array} \right\}
  \left\{\begin{array}{c} \tau_{zx}(x,y,z)\\
  \tau_{yz}(x,y,z)\end{array}\right\} +
  \displaystyle{\frac{\partial\sigma_z(x,y,z)}{\partial z}  +\bar{Z}(x,y,z) }&=& 0\\
  \boldsymbol{\nabla^T\tau}(x,y,z)  + \displaystyle{\frac{\partial\sigma_z(x,y,z)}{\partial z}
  +\bar{Z}(x,y,z) }&=& 0  \end{array}
  }$

### Okrajové podmínky
* Kinematické okrajové podmínky - předepsaný 

* Statické okrajové podmínky
  * Na povrchu pružné vrstvy ($z = 0$ m)
  
   $\displaystyle{
   \begin{array}{rcl}
\sigma_z(x,y,0)n_z(x,y)-\bar{p_z}(x,y) &=& 0\\
-\sigma_z(x,y,0)-\bar{p_z}(x,y) &=&0
\end{array}
}$

<img src="Figures/bc.png" width=350>

  * Na „svislých hranách ($x \in Γ \times 0; h $)

   $\displaystyle{
  \begin{array}{rcl}
\tau_{zx}(x,y,z)n_x(x,y)+\tau_{yz}(x,y,z)n_y(x,y)-\bar{\tau}(x,y,z) &=&0\\
\left\{ \begin{array}{cc} n_x(x,y)&n_y(x,y) \end{array} \right\}
\left\{ \begin{array}{r} \tau_{zx}(x,y,z)\\ \tau_{zy}(x,y,z) \end{array} \right\} - 
\bar{\tau}(x,y,z) &=&0\\ \boldsymbol{n}^T(x,y)\boldsymbol{\tau}(x,y,z) -\bar{\tau}(x,y,z) &=&0
\end{array}
}$

### Řídicí rovnice
#### Dimenzionální redukce problému
 *  Integrací podmínky rovnováhy (statických rovnic) podle $z$ s vahou $\psi$ dostáváme
   
   $\displaystyle{
   \int_0^h \left(
   \boldsymbol{\nabla}^T\boldsymbol{\tau}(x,y,z)+\underline{\frac{\partial\sigma_z(x,y,z)}{\partial z}} + \bar{Z}(x,y,z)   
   \right)
   \psi(z)\ dz =0
   }$


 *  Podtržený člen upravíme pomocí integrace per partes
  
   $\displaystyle{
   \begin{array}{rcl}
   \displaystyle{\int_0^h\frac{\partial\sigma_z(x,y,z)}{\partial z}\psi(z)\ dz} 
   &=&\displaystyle{\left[ \sigma_z(x,y,z)\psi(z)\right]_0^h - \int_0^h\sigma_z(x,y,z)\frac{d\psi(z)}{dz}\ dz}\\
   &=&\displaystyle{-\sigma_z(x,y,0) - \int_0^h\sigma_z(x,y,z)\frac{d\psi(z)}{dz}\ dz}\\
   &=&\displaystyle{\bar{p_z}(x,y) - \int_0^h\sigma_z(x,y,z)\frac{d\psi(z)}{dz}\ dz}
   \end{array}
   }$


 *  Po dosazení dostáváme
  
   $\displaystyle{
   0 = \boldsymbol{\nabla}^T\int_0^h\boldsymbol{\tau}(x,y,z)\psi(z)\ dz - \int_0^h \sigma_z(x,y,z)\frac{d\psi(z)}{dz}\ dz + \bar{p_z}(x,y)+ \int_0^h \bar{Z}(x,y,z)\psi(z)\ dz
   }$

 *  Tato úprava nám umožňuje přejít z třírozměrné úlohy na dvojrozměrnou
 * Poloha je nyní charakterizována pomocí dvou prostorových souřadnic
  
   $\displaystyle{
   \boldsymbol{x} = \left\{x,y \right\}^T
   }$


 * Podmínku rovnováhy ve svislém směru vyjádříme pomocí
    * (zobecněných měrných) posouvajících sil $\boldsymbol{q}(\boldsymbol{x})$
  
      $\displaystyle{\boldsymbol{q}(\boldsymbol{x}) = \int_0^h \boldsymbol{\tau}(\boldsymbol{x},z)\psi(z)\ dz}$ 

   * (zobecněné měrné) normálové síly $n_z$
  
      $\displaystyle{n_z(\boldsymbol{x}) = \int_0^h \sigma_z(\boldsymbol{x},z)\frac{d\psi(z)}{dz}\ dz }$

   *  (zobecněného) plošného zatížení $\bar{p}$
  
      $\displaystyle{\bar{p}(\boldsymbol{x}) = \bar{p_z}(\boldsymbol{x}) + \int_0^h \bar{Z}(\boldsymbol{x},z)\psi(z)\ dz}$



  * Tedy
  
   $\displaystyle{\boldsymbol{\nabla}^T\boldsymbol{q}(\boldsymbol{x}) - n_x(\boldsymbol{x}) + \bar{p}(\boldsymbol{x}) = 0 }$


 * Obdobným způsobem modifikujeme statické okrajové podmínky
  
   $\displaystyle{
   \begin{array}{rcl}
   \displaystyle{\boldsymbol{n}^T(\boldsymbol{x}) \int_0^h \boldsymbol{\tau}(\boldsymbol{x},z)\psi(z)\ dz - \int_0^h \boldsymbol{\tau}(\boldsymbol{x},z)\psi(z)\ dz }&=&0\\
   \boldsymbol{n}^T(\boldsymbol{x})\boldsymbol{q}(\boldsymbol{x})-\bar{q}(\boldsymbol{x})&=&0
   \end{array}
   }$

#### Dosazení z konstitutivních rovnic
 *  Normálové síly $n_x$
  
   $\displaystyle{
   \begin{array}{rcl}
   n_z(\boldsymbol{x}) &=&\displaystyle{ \int_0^h \sigma_z(\boldsymbol{x},z)\frac{d\psi(z)}{dz}\ dz = \int_0^h E_{oed}(\boldsymbol{x},z)\varepsilon_z(\boldsymbol{x},z)\frac{d\psi(z)}{dz}\ dz}\\
   &=& \displaystyle{\left( \int_0^h E_{oed}(\boldsymbol{x},z)\frac{d\psi(z)}{dz}\frac{d\psi(z)}{dz}\ dz \right) w(x,y) = C_1(\boldsymbol{x})w(x,y)}
   \end{array}
   }$


 *  Výsledkem je tedy vztah
  
   $\displaystyle{
   n_z(\boldsymbol{x}) = C_1(\boldsymbol{x})w(x,y)
   }$

   kde konstanta $C_1$ [Nm$^{−3}$]
  
   $\displaystyle{
   C_1(\boldsymbol{x}) = \int_0^h E_{oed}(\boldsymbol{x},z)\left(\frac{d\psi(z)}{dz}\right)^2 \ dz
   }$

   
  je též někdy nazývána součinitel ložnosti.

 * Posouvající síly $\boldsymbol{q}$
  
   $\displaystyle{
   \begin{array}{rcl}
   \left\{
   \begin{array}{c}
   q_x(\boldsymbol{x})\\
   q_y(\boldsymbol{x})
   \end{array}
   \right\} &=& \displaystyle{\int_0^h 
   \left\{
   \begin{array}{c}
   \tau_{zx}(\boldsymbol{x},z)\\
   \tau_{yz}(\boldsymbol{x},z)
   \end{array}
   \right\} 
   \psi(z)\ dz}\\
   &=&\displaystyle{ \int_0^h  
   \left[
   \begin{array}{cc}
   G_{x}(\boldsymbol{x},z)&0\\
   0&G_{y}(\boldsymbol{x},z)
   \end{array}
   \right] }
   \left\{
   \begin{array}{c}
   \gamma_{zx}(\boldsymbol{x},z)\\
   \gamma_{yz}(\boldsymbol{x},z)
   \end{array}
   \right\} 
   \psi(z)\ dz\\
   &=& \left( \displaystyle{\int_0^h  
   \left[
   \begin{array}{cc}
   G_{x}(\boldsymbol{x},z)&0\\
   0&G_{y}(\boldsymbol{x},z)
   \end{array}
   \right] }
   \psi^2(z)\ dz\right)
   \boldsymbol{\nabla}w(\boldsymbol{z})\\
   &=&\boldsymbol{C_2}(\boldsymbol{x})\boldsymbol{\nabla}w(\boldsymbol{x})
   \end{array}
   }$


 * Pro případ $G_x = G_y = G$
  
   $\displaystyle{
   \boldsymbol{q}(\boldsymbol{x}) = C_2(\boldsymbol{x})\boldsymbol{\nabla}w(\boldsymbol{x})
   }$


   kde $C_2$ [Nm$^{−1}$]

  
   $\displaystyle{
   C_2(\boldsymbol{x}) = \int_0^h G(\boldsymbol{x},z)\psi^2(z)\ dz
   }$

### Diferenciální rovnice pružné vrstvy

  * Uvažujeme izotropní a homogenní materiál (vzhledem k souřadnicím
$x$ a $y$)

    $\displaystyle{
C_1(\boldsymbol{x}) = C_1 ,\\
C_2 (\boldsymbol{x}) = C_2 .
}$


  * Po dosazení konstitutivních rovnic do podmínky rovnováhy dostáváme

  $\displaystyle{
    C_2\boldsymbol{\nabla}^T\boldsymbol{\nabla}w(\boldsymbol{x})-C_1w(\boldsymbol{x})+\bar{p}(\boldsymbol{x}) = 0
    }$

    tedy

   $\displaystyle{
    C_2\Delta w(\boldsymbol{x})-C_1w(\boldsymbol{x})+\bar{p}(\boldsymbol{x}) = 0  
    }$


   * Z hlediska matematické terminologie se tato parciální diferenciální rovnice nazývá Helmholtzovou rovnicí.

### Slabá formulace

   * Požadujeme, aby platilo

      $\displaystyle{
      \delta w(\boldsymbol{x})\left( \boldsymbol{\nabla}^T\boldsymbol{C_2}(\boldsymbol{x})\boldsymbol{\nabla}w(\boldsymbol{x}) - C_1(\boldsymbol{x})w(\boldsymbol{x})+\bar{p}(\boldsymbol{x})\right)\ d\boldsymbol{x} = 0
      }$
    
      pro všechny váhové funkce $\delta w(\boldsymbol{x})$.

    
   * Aplikací Gaussovy věty upravíme předchozí rovnost na

      $\displaystyle{
      \begin{array}{rcl}
      0&=&\displaystyle{\int_{\Gamma} \delta w(\boldsymbol{x})\boldsymbol{n}^T(\boldsymbol{x})\boldsymbol{C_2}(\boldsymbol{x})\boldsymbol{\nabla}w(\boldsymbol{x})\ d\boldsymbol{x} - \int_{\Omega}\left(\boldsymbol{\nabla}\delta w(\boldsymbol{x})   \right)^T \boldsymbol{C_2}(\boldsymbol{x})\boldsymbol{\nabla}w(\boldsymbol{x})\ d\boldsymbol{x}}\\
      &-&\displaystyle{ \int_{\Omega} \delta w(\boldsymbol{x})C_1(\boldsymbol{x})w(\boldsymbol{x})\ d\boldsymbol{x} +  \int_{\Omega} \delta w(\boldsymbol{x})\bar{p}(\boldsymbol{x})\ d\boldsymbol{x}}
      \end{array}
      }$


   * Slabé řešení w(x) tedy splňuje pro všechna $\delta w(\boldsymbol{x})$

      $\displaystyle{
      \int_{\Omega} \delta w(\boldsymbol{x})C_1(\boldsymbol{x})w(\boldsymbol{x})\ d\boldsymbol{x} + 
      \int_{\Omega}\left(\boldsymbol{\nabla}\delta w(\boldsymbol{x})   \right)^T \boldsymbol{C_2}(\boldsymbol{x})\boldsymbol{\nabla}w(\boldsymbol{x})\ d\boldsymbol{x} = \int_{\Gamma} \delta w(\boldsymbol{x})\bar{q}(\boldsymbol{x})\ d\boldsymbol{x} +  \int_{\Omega} \delta w(\boldsymbol{x})\bar{p}(\boldsymbol{x})\ d\boldsymbol{x}
      }$

### Galerkinovská aproximace
   
   * Aproximace neznámých $w(\boldsymbol{x})$ a jejich gradientů $\nabla^Tw(\boldsymbol{x})$

      $\displaystyle{
      w(\boldsymbol{x})\approx \boldsymbol{N}(\boldsymbol{x})\boldsymbol{r}\\
      \boldsymbol{\nabla}^T w(\boldsymbol{x}) \approx \boldsymbol{\nabla}^T\boldsymbol{N}(\boldsymbol{x})\boldsymbol{r} = \boldsymbol{B}(\boldsymbol{x})\boldsymbol{r}
      }$


   * Aproximace váhové funkce $\delta w(x)$ a jejího gradientu $\nabla T \delta w(x)$

      $\displaystyle{
        \delta w(\boldsymbol{x})\approx \boldsymbol{N}(\boldsymbol{x})\boldsymbol{\delta r}\\
      \boldsymbol{\nabla}^T \delta w(\boldsymbol{x}) \approx \boldsymbol{\nabla}^T\boldsymbol{N}(\boldsymbol{x})\boldsymbol{\delta r} = \boldsymbol{B}(\boldsymbol{x})\boldsymbol{\delta r}
      }$


   * Aproximace slabé formulace

      $\displaystyle{
      \int_{\Omega} (\boldsymbol{N}(\boldsymbol{x})\boldsymbol{\delta r})^TC_1(\boldsymbol{x})\boldsymbol{N}(\boldsymbol{x})\boldsymbol{r}\ d\boldsymbol{x} + 
      \int_{\Omega}\left(\boldsymbol{B}(\boldsymbol{x})\boldsymbol{\delta r}\right)^T \boldsymbol{C_2}(\boldsymbol{x})\boldsymbol{B}(\boldsymbol{x})\boldsymbol{r}\ d\boldsymbol{x} = \int_{\Gamma} (\boldsymbol{N}(\boldsymbol{x})\boldsymbol{\delta r})^T\bar{q}(\boldsymbol{x})\ d\boldsymbol{x} +  \int_{\Omega} (\boldsymbol{N}(\boldsymbol{x})\boldsymbol{\delta r})^T\bar{p}(\boldsymbol{x})\ d\boldsymbol{x}
      }$

     pro všechna $\delta r$.
 
   * Soustava lineárních rovnic

      $\displaystyle{\boldsymbol{K}\boldsymbol{r}=\boldsymbol{R}=\boldsymbol{R_q}+\boldsymbol{R_p}}$

      kde

        $\displaystyle{
        \begin{array}{rcl}
      \boldsymbol{K}&=&\displaystyle{ \int_{\Omega} \left(\boldsymbol{N}^T(\boldsymbol{x})C_1(\boldsymbol{x})\boldsymbol{N}(\boldsymbol{x}) +  \boldsymbol{B}^T(\boldsymbol{x}) \boldsymbol{C_2}(\boldsymbol{x})\boldsymbol{B}(\boldsymbol{x})\right)
      \ d\boldsymbol{x}}\\
      \boldsymbol{R_q}&=& \displaystyle{\int_{\Gamma} \boldsymbol{N}^T(\boldsymbol{x})\bar{q}(\boldsymbol{x})\ d\boldsymbol{x}}\\
      \boldsymbol{R_p}&=&\displaystyle{\int_{\Omega} \boldsymbol{N}^T(\boldsymbol{x})\bar{p}(\boldsymbol{x})\ d\boldsymbol{x}}
      \end{array}
      }$

## Příklad 

Modelujte interakci dvou základových pasů zatížených rovnoměrným spojitým zatížením jako desku na pružném podloží.

<img src = "Figures/ex3.png" width = 500>

<i>
    Pozn.: Pro zachování kompatibility stupňů volnosti je uvažováno působení desky na celé řešené oblasti. Pro prvky, kde deska nepůsobí, je tuhost deskového prvku oslabena výrazným snížením tloušťky desky.
    </i>

```octave
lenX = 13;
lenY = 10;

nx = 39;
ny = 40;iMPROVED FORMATTING

nn = (nx+1)*(ny+1);
ne = nx*ny;

[conn, lm, ev, xy] = inputData (lenX, lenY, nx, ny);
```

```octave
neq = nn*3;
peq = 0;

E = 30.e9;
v = 0.15;
h = 0.1;
C1 = 1000;
C2 = 1000;
q = -100;

k = zeros (neq+peq);
f = zeros (neq, 1);
 
for i=1:ne
 xe = [xy(conn(i,1),1), xy(conn(i,2),1), xy(conn(i,3),1), xy(conn(i,4),1)];
 ye = [xy(conn(i,1),2), xy(conn(i,2),2), xy(conn(i,3),2), xy(conn(i,4),2)];
 
 
 % ev = 0 ... podlozi
 % ev = 1 ... deska
 if(ev(i) == 0)
   % platek(...) - funkce pro vypocet matice tuhosti desky na pruznem podlozi
   ke = platek(E,v,h/1000,C1,C2, xe, ye); % pouze podlozi, tuhost desky snizena /1000
 else
   ke = platek(E,v,h,C1,C2, xe, ye); % deska
 end
 
 k(lm(i,:), lm(i,:))=k(lm(i,:),lm(i,:))+ke;
end

for i=1:8
  e1 = (12*ny+8+i);
  e2 = (24*ny+8+i);
  
  for j=1:4
    f(lm(e1,(j-1)*3+1)) = f(lm(e1,(j-1)*3+1)) + q/4;
    f(lm(e2,(j-1)*3+1)) = f(lm(e2,(j-1)*3+1)) + q/4;
    
  end
end

u = k\f;

w = zeros(nn, 1);

hold on;
coeff = 1;
for i=1:ne
 x = [xy(conn(i,1),1), xy(conn(i,2),1);
      xy(conn(i,4),1), xy(conn(i,3),1)];
 y = [xy(conn(i,1),2), xy(conn(i,2),2);
      xy(conn(i,4),2), xy(conn(i,3),2)];
 w = coeff*[u(lm(i,1)),u(lm(i,4));u(lm(i,10)),u(lm(i,7))];
 subplot(1,2,1);
 surf(x,y,w); view(0,90); hold on;  axis([0 13 0 22 -0.5 0]); grid off
 subplot(1,2,2);
 Y = y;
 Y(:,:) = 0;
 surf(x,Y,w,gradient(w)); view(0,0); hold on;  axis([0 13 0 10 -0.25 0])
end
```

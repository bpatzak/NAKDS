# Stacionární vedení tepla v 1D

Úloha stacionárního (ustáleného) vedení tepla se zabývá případem, kdy došlo k ustálení teploty v tělese pro zadané okrajové podmínky a tyto okrajové podmínky se dále nemění v čase. 

### Diferenciální rovnice problému

Základní veličiny:
   * Teplota $T(x)$ \[K\] (stavová veličina),
   * Teplo $Q(x)$ \[J\] (tepelná energie) - část vnitřní energie tělesa, kterou těleso přijme nebo odevzdá při tepelné výměně druhému tělesu (změna stavu),
   * Gradient teploty $\frac{dT(x)}{dx}$ - změna teploty ve směru souřadné osy $x$, právě na této změně závisí výměna tepelné energie (ne na samotné hodnotě $T$)
   * Tepelný tok $q(x)$ \[Wm$^{-2}$\] - množství tepla $\Delta Q(x)$, které projde plochou $\Delta A(x)$ s normálou $n(x)$ za čas $\Delta t$ (pro ustálený stav $\Delta t = 1$ s)
   
   $\displaystyle{
   q(x) = \frac{\Delta Q(x)}{\Delta A(x) \Delta t}n(x) 
   }$

Diferenciální rovnice úlohy vedení tepla plyne z <b>bilance energie</b>:

<img src = "Figures/bilance3.png" width=300>

$\displaystyle{
q_x(x)A(x) + \bar{Q}\left(x+\frac{\Delta x}{2}\right) = q_x(x+\Delta x)A(x+\Delta x),
}$


pro $\Delta x\rightarrow 0$ 

$\displaystyle{
-\frac{d}{dx} (q_x(x)A(x)) + \bar{Q}(x)A(x) = 0.
}$


Pro konstantní plochu průřezu

$\displaystyle{
-\frac{d q_x}{dx}(x) + \bar{Q}(x) = 0.
}$


#### Konstitutivní rovnice (Fourierův zákon)

$\displaystyle{
q(x) = -\lambda(x)\frac{dT}{dx}(x),
}$

kde $\lambda$ je součinitel tepelné vodivosti \[Wm$^{-1}$K$^{-1}$\].

#### Okrajové podmínky

Kladný předepsaný okrajový tok uvažujeme ve směru vnější normály 

$\displaystyle{
q_x(x)n(x) = \bar{q}_x(x).
}$


  * <b>Dirichletova okrajová podmínka</b> - předepsaná teplota na hranici
  
  $\displaystyle{
  T(x) = \bar{T}(x),\qquad \mbox{pro }x\in \Gamma_T.
  }$
  
  
  * <b>Neumannova okrajová podmínka</b> - předepsaný okrajový tok
  
  $\displaystyle{
  -\lambda(x)\frac{dT}{dx}(x)n(x) = \bar{q}_x(x),\qquad \mbox{pro }x\in \Gamma_{qp}.
  }$
  
  
  * <b>Smíšená okrajová podmínka</b> - přestup tepla  
     (Convection/Mixed/Robin/Newton BC) 
 
    $\displaystyle{
    \bar{q}_x(x) = \alpha(x)\left(T(x)-T_0(x)\right),\qquad \mbox{pro }x\in \Gamma_{qc},
    }$   
  
      kde $\alpha$ je tzv. film coefficient \[Jm$^{-2}$K$^{-1}$s$^{-1}$\] a $T_0$ je teplota okolního prostředí.  
  
  
  * <b>Radiace </b>
   
   $\displaystyle{
   \bar{q}(x) = \varepsilon(x)\sigma(x)\left(T^4(x) - T^4_{\infty}\right),\qquad \mbox{pro }x\in \Gamma_{qr},
   }$
   
   kde $\varepsilon$ \[-\] je povrchové vyzařování vztažené k černému tělesu ($0<\varepsilon<1$), $\sigma  = 5.67\times 10^{-8}$ \[Wm$^{-2}$K$^{-4}$\] je Stefan-Boltzmanova konstanta a $T_{\infty}$ je teplota okolního prostředí.

### Slabé řešení


Hledáme řešení $T(x)$ rovnice takové, které splňuje rovnici

$\displaystyle{
0 = \int_{\Omega} \delta T(x) \left(  \frac{d}{dx}\left(\lambda(x)\frac{d T}{dx}(x)\right)+\bar{Q}(x) \right) \ dx
}$

pro všechna $\delta T $ splňující kinematické i statické okrajové podmínky. Použitím integrace per-partes získáme


$\displaystyle{
0 = \int_{\Gamma}\delta T(x) \lambda(x)\frac{d T}{dx}(x)n(x) \ dx-\int_{\Omega} \frac{d\delta T}{dx}(x) \lambda(x)\frac{d T}{dx}(x)\ dx + \int_{\Omega}  \delta T(x) \bar{Q}(x) \ dx,
}$

kde $n$ je vnější normála k hranici $\Gamma$. Hraniční integrál můžeme rozdělit na část hranice $\Gamma_T$, kde působí kinematické (Dirichletovy) okrajové podmínky, a na část $\Gamma_q$, kde působí silové okrajové podmínky (předepsaný okrajový tok)


$\displaystyle{
\int_{\Gamma}\delta T(x) \lambda(x)\frac{d T}{dx}(x)n(x) \ dx  = \int_{\Gamma_T}\underbrace{\delta T(x)}_{=0} \lambda(x)\frac{d T}{dx}(x)n(x) \ dx + \int_{\Gamma_q}\delta T(x) \underbrace{\lambda(x)\frac{d T}{dx}(x)n(x)}_{=-\bar{q}_x} \ dx,
}$


Část s předepsaným okrajovým tokem můžeme dále rozdělit na část, na které je předepsána Neumannova okrajová podmínka a na část s předepsanou smíšenou okrajovou podmínkou (přestup tepla)

$\displaystyle{
\int_{\Gamma_q}\delta T(x) \bar{q}_x \ dx = \int_{\Gamma_{qp}}\delta T(x) \bar{q}_x \ dx  + \int_{\Gamma_{qc}}\delta T(x) \alpha(x)(T(x)-T_0(x)) \ dx.
}$


Dosazením získáme

$\displaystyle{ \int_{\Omega} \frac{d\delta T}{dx}(x) \lambda(x)\frac{d T}{dx}(x)\ dx + \int_{\Gamma_{qc}}\delta T(x) \alpha(x)T(x) \ dx = \int_{\Gamma_{qc}}\delta T(x) \alpha(x)T_0(x)\ dx -\int_{\Gamma_{qp}}\delta T(x) \bar{q}_x \ dx  + \int_{\Omega}  \delta T(x)\bar{Q}(x) \ dx. }$




### Diskretizovaná úloha

Předpokládejme aproximaci neznámé teploty ve tvaru $T(x)=N^e(x)r^e$ a váhové funkce $\delta T(x)=N^e(x)\delta r^e$. Po dosazení do slabého řešení dostáváme

$\displaystyle{\displaystyle{ \left(\boldsymbol{K}^e_{\Omega} + \boldsymbol{K}^e_\Gamma\right)r^e =
f^e_{\Gamma_c} + f^e_{\Gamma_p} + f^e_{\Omega}    }}$


* Levá strana
    * Matice vodivosti $\displaystyle{K^e_{\Omega}}$
    
      $\displaystyle{\boldsymbol{K}^e_{\Omega} = \int_{\Omega} \boldsymbol{B}^{eT}\lambda\  \boldsymbol{B}^{e} \ ds }$
      
    * Příspěvek do levé strany od přestupu tepla $\displaystyle{K^e_{\Gamma}}$
    
     $\displaystyle{\boldsymbol{K}^e_{\Gamma} =\int_{\Gamma_c} \boldsymbol{N}^{eT}\alpha^e \boldsymbol{N}^e\ ds}$


* Pravá strana
    * Přestup tepla  $\displaystyle{f^e_{\Gamma_c}}$
    
      $\displaystyle{ f^e_{\Gamma_c} = \int_{\Gamma_c}\boldsymbol{N}^{eT}\alpha^e T_0^e(x)  \ ds}$

    * Předepsaný tok  $\displaystyle{f^e_{\Gamma_p}}$

      $\displaystyle{f^e_{\Gamma_p} = - \int_{\Gamma_p}\boldsymbol{N}^{eT}  \bar{q}^e(x)  \ ds}$

    * Vnitřní zdroj  $\displaystyle{ f^e_{\Omega} }$
    
      $\displaystyle{ f^e_{\Omega} = \int_{\Omega} \boldsymbol{N}^{eT}  \bar{Q}^e(x) \ ds    }$

### Analytické řešení


* Diferenciální rovnice vedení tepla

$\qquad\qquad\displaystyle{\frac{d}{dx} \left( \lambda (x)\frac{dT(x)}{dx}\right) + \bar{Q}(x) = 0}$

* Pro $\lambda(x) = konst. = \lambda$

$\qquad\qquad\displaystyle{\lambda \frac{d^2T(x)}{dx^2} = -\bar{Q}(x)}$

* Integrací získáme

$\qquad\qquad\displaystyle{\lambda \frac{dT(x)}{dx} = -\bar{Q}(x)x + C_1}$


$\qquad\qquad\displaystyle{\lambda T(x) = -\bar{Q}(x)\frac{x^2}{2} + C_1x + C_2}$

* Okrajové podmínky:

$\qquad\qquad\displaystyle{
\begin{array}{lll}
\bar{T}(0) = 0 &\qquad\qquad \lambda T(0) = C_2 & \Rightarrow\qquad C_2 = 0 \\
\bar{q}(4) = 5 &\qquad\qquad q(4) =  -\lambda \frac{dT(x)}{dx} = {Q}(4)\cdot 4 - C_1&  \Rightarrow\qquad C_1 = {Q}(4)\cdot 4 - q(4) = {S\over A}\cdot 4 - q(4) = 4 {5\over{0.1}}-5 = 195
\end{array}
}$

* Analytické řešení

$\qquad\qquad\displaystyle{T(x) = -\frac{1}{\lambda }\bar{Q}(x)\frac{x^2}{2} +\frac{1}{\lambda } 195x}$

$\qquad\qquad\displaystyle{T(x) = -\frac{1}{2}{5\over{0.1}} \frac{x^2}{2} + \frac{1}{2}195x}$

#### Řešení n prvky

```octave
n = 3
l = 4/n;

A = 0.1;
lambda = 2;
x1 = 0; x2 = 4/n;

ki = (lambda/l)*[1 -1; -1 1];
fo = (5/0.1)*[(x2*x2-x2^2/2)/l (x2^2/2)/l];
fgp = 5*[0 -x2/l];

K = zeros (n+1);
F = zeros (n+1, 1);
for i=1:n
    loc = [i i+1];
    K(loc, loc) += ki;
    F(loc)+= fo';
end
F([n n+1]) += fgp';
u = K(2:n+1, 2:n+1)\(F(2:n+1,1));
U = [0 ; u]

#plot analytical solution 
hold on;
x = 0:0.1:4;
plot (x, -12.5*x.^2+97.5*x, "b;u;")

#plot obtained solution
for i=1:n
    plot ([(i-1)*l i*l], [U(i) U(i+1)], "b--")
end
```

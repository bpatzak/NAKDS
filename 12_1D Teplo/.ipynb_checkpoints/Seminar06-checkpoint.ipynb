# Stacionární vedení tepla v 1D



### Diskretizovaná úloha
$$\displaystyle{ \left(\boldsymbol{K}^e_{\Omega} + \boldsymbol{K}^e_\Gamma\right)r^e =
f^e_{\Gamma_c} + f^e_{\Gamma_p} + f^e_{\Omega}    }$$

* Levá strana
    * Matice vodivosti $\displaystyle{K^e_{\Omega}}$
$$\displaystyle{\boldsymbol{K}^e_{\Omega} = \int_{\Omega} \boldsymbol{B}^{eT}\lambda\  \boldsymbol{B}^{e} \ ds }$$

    * Příspěvek do levé strany od přestupu tepla $\displaystyle{K^e_{\Gamma}}$
$$\displaystyle{\boldsymbol{K}^e_{\Gamma} =\int_{\Gamma_c} \boldsymbol{N}^{eT}\alpha^e \boldsymbol{N}^e\ ds}$$

* Pravá strana
    * Přestup tepla  $\displaystyle{f^e_{\Gamma_c}}$
$$\displaystyle{ f^e_{\Gamma_c} = \int_{\Gamma_c}\boldsymbol{N}^{eT}\alpha^e T_0^e(x)  \ ds}$$

    * Předepsaný tok  $\displaystyle{f^e_{\Gamma_p}}$
$$\displaystyle{f^e_{\Gamma_p} = - \int_{\Gamma_p}\boldsymbol{N}^{eT}  \bar{q}^e(x)  \ ds}$$

    * Vnitřní zdroj  $\displaystyle{ f^e_{\Omega} }$
$$\displaystyle{ f^e_{\Omega} = \int_{\Omega} \boldsymbol{N}^{eT}  Q^e(x) \ ds    }$$


### Okrajové podmínky

 * Dirichletova okrajová podmínka - předepsaná teplota na hranici
$$\displaystyle{  T(x) = \bar{T}(x),\qquad \qquad \mbox{pro } x\in \Gamma_T}$$


 * Neumannova okrajová podmínka - předepsaný tok na hranici 
$$\displaystyle{  q_x(x)n(x) = \bar{q_x}(x),\qquad \qquad \mbox{pro } x\in \Gamma_{{qp}}}$$


 * Smíšená okrajová podmínka - přestup tepla
$$\displaystyle{  \bar{q_x}(x) = \alpha(x)(T(x)-T_0(x)),\qquad \qquad \mbox{pro } x\in \Gamma_{qc}}$$


 * Radiace - přestup tepla
$$\displaystyle{  \bar{q_x}(x) = \varepsilon(x)\sigma(x)(T^4(x)-T^4_{\infty}(x)),\qquad \qquad \mbox{pro } x\in \Gamma_{qc}}$$
($\displaystyle{\varepsilon}$ - povrchové vyzařování, 
$\sigma$ - Stefan-Boltzmanova konstanta)

### Lineární aproximace
*  Matice interpolačních funkcí
<img src="Figures/basis.png" width="300" align="right">

$
\qquad \displaystyle{\boldsymbol{N}^e = \left[\frac{x_2-x}{l^e}, \frac{x-x_1}{l^e} \right]}
$

*  Geometrická matice (derivace interpolačních funkcí)

$
\qquad \displaystyle{\boldsymbol{B}^e = \left[\frac{-1}{l^e}, \frac{1}{l^e} \right]} 
$

* Matice vodivosti

$
\qquad \displaystyle{\boldsymbol{K}^{e} = \int_{\Omega} \boldsymbol{B}^{eT}\lambda\  \boldsymbol{B}^{e} \ ds = \int_{\Omega} \left[ 
\begin{array}{r}
{-1/l^e}\\
{1/l^e} 
\end{array}
\right] \lambda \left[-1/l^e\quad 1/l^e \right]\ ds = 
\frac{\lambda}{l^e}
\left[
\begin{array}{rr}
 1 & -1\\
-1 & 1 
\end{array}
\right]}
$

* Vnitřní zdroj

$\qquad \displaystyle{ f^e_{\Omega} = \int_{\Omega} \boldsymbol{N}^{eT} Q^e(x)\ ds  =
\int_{x_i}^{x^j} \left[ 
\begin{array}{r}
{(x_j - x)/l^e}\\
{(x - x_i)/l^e} 
\end{array}
\right] \frac{S}{A^e} dx=\left[ 
\begin{array}{c}
{\left(x_jx - {{x^2}\over{2}}\right)/l^e}\\
{\left({x^2\over2} - x_ix\right)/l^e} 
\end{array}
\right]_{x_i}^{x^j}
\frac{S}{A^e}
}$


* Předepsaný tok  $\displaystyle{f^e_{\Gamma_p}}$

$\qquad \displaystyle{f^e_{\Gamma_p} = - \int_{\Gamma_p}\boldsymbol{N}^{eT}\ \bar{q}^e(x) \ d{\Gamma_p}=-\int_{\Gamma_p}\left[ 
\begin{array}{c}
{(x_j - x)/l^e}\\
{(x - x_i)/l^e} 
\end{array}
\right] \bar{q}^e(x) \ d{\Gamma_p}=-\left.\left[ 
\begin{array}{c}
{(x_j - x)/l^e}\\
{(x - x_i)/l^e} 
\end{array}
\right]  \bar{q}(4) \right|_{x=4}
}$


### Řešení jedním prvkem

* Vnitřní zdroj

$\qquad \displaystyle{ f^1_{\Omega} =

{\left({x^2\over2} - x_1x\right)/l_1}

\begin{array}{c}

```octave
l1 = 4;
A = 0.1;
lambda = 2;
x1 = 0; x2 = 4;

k1 = (lambda/l1)*[1 -1; -1 1];

K = k1;
fo = (5/0.1)*[(x2*x2-x2^2/2)/l1 (x2^2/2)/l1];
fgp = 5*[0; -1];
f1 = fo + fgp;
F = f1;

u = K(2,2)\F(2);
U = [0 u]

hold on;
x = 0:0.1:4;
plot (x, -12.5*x.^2+97.5*x, "b;u;")
plot ([0 l1], [U(1) U(2)], "b--;uc;")
```

### Řešení n prvky

```octave
n = 3
l = 4/n;

A = 0.1;
lambda = 2;
x1 = 0; x2 = 4/n;

ki = (lambda/l)*[1 -1; -1 1];
fo = (5/0.1)*[(x2*x2-x2^2/2)/l (x2^2/2)/l];
fgp = 5*[0 -x2/l];

K = zeros (n+1);
F = zeros (n+1, 1);
for i=1:n
    loc = [i i+1];
    K(loc, loc) += ki;
    F(loc)+= fo';
end
F([n n+1]) += fgp';
u = K(2:n+1, 2:n+1)\(F(2:n+1,1));
U = [0 ; u]

#plot analytical solution 
hold on;
x = 0:0.1:4;
plot (x, -12.5*x.^2+97.5*x, "b;u;")

#plot obtained solution
for i=1:n
    plot ([(i-1)*l i*l], [U(i) U(i+1)], "b--")
end
```

### Řešení dvěma prvky s lineární aproximací

* Matice vodivosti

$
\qquad \displaystyle{\boldsymbol{K}^{1} = 
\frac{\lambda_1}{l_1}
\left[
\begin{array}{rr}
 1 & -1\\
-1 & 1 
\end{array}
\right] = 
\frac{0.5}{2}
\left[
\begin{array}{rr}
 1 & -1\\
-1 & 1 
\end{array}
\right]
}$

$
\qquad \displaystyle{\boldsymbol{K}^{2} = 
\frac{\lambda_2}{l_2}
\left[
\begin{array}{rr}
 1 & -1\\
-1 & 1 
\end{array}
\right] = \frac{2}{4}
\left[
\begin{array}{rr}
 1 & -1\\
-1 & 1 
\end{array}
\right]
}$

* Příspěvek do matice soustavy od přestupu tepla

$
\qquad \displaystyle{\boldsymbol{K}_{\Gamma}^{1} = \int_{\Gamma_c} \boldsymbol{N}^{eT}\alpha^e \boldsymbol{N}^e\ ds= \left[ 
\begin{array}{c}
{(x_2-x)/l_1}\\
{(x-x_1)/l_1} 
\end{array}
\right]
\alpha
\left.
\left[
{{x_2 - x}\over{l_1}},\ 
{{x - x_1}\over{l_1}}
\right]
\right|_{x_1}=
4\left[ 
\begin{array}{cc}
1&0\\
0&0 
\end{array}
\right]
}$

* Přestup tepla

$\qquad \displaystyle{ f^1_{\Gamma_c} = \int_{\Gamma_c}\boldsymbol{N}^{eT}\alpha^e T_0(x)  \ ds =
\left.
\left[ 
\begin{array}{c}
{(x_2-x)/l_1}\\
{(x-x_1)/l_1} 
\end{array}
\right]
\alpha T_0
\right|_{x_1}=4\cdot20\left\{ 
\begin{array}{r}
1\\
0 
\end{array}
\right\} = \left\{ 
\begin{array}{r}
80\\
0 
\end{array}
\right\}
}$


#### Globální soustava rovnic

$ \displaystyle{
\left[
\begin{array}{rrr}
 4.25& -0.25 & 0\\
-0.25 & 0.75 & -0.5\\
0 & -0.5 & 0.5
\end{array}
\right]
\left\{
\begin{array}{r}
 T_1\\
 T_2\\
 0
\end{array}
\right\}= \left\{
\begin{array}{r}
 80\\
 0\\
 r_1
\end{array}
\right\}
}$


```octave
l1 = 2; l2 = 4;
A = 0.1;
lambda1 = 0.5; lambda2 = 2;
x1 = 0; x2 = 2; x3 = 6;
alpha = 4;
T0 = 20;

loc1 = [1 2];
loc2 = [2 3];

k1 = (lambda1/l1)*[1 -1; -1 1];
k1 += alpha*[1 0; 0 0];
k2 = (lambda2/l2)*[1 -1; -1 1];

fgc1 = alpha*T0*[1; 0];

K = zeros(3,3);
K(loc1,loc1) += k1;
K(loc2,loc2) += k2

F = zeros(3,1);
F(loc1) += fgc1

u = K(1:2,1:2)\F(1:2);
U = [u; 0]

hold on;
plot ([x1 x2], [U(1) U(2)], "b-;e1;")
plot ([x2 x3], [U(2) U(3)], "r-;e2;")
```

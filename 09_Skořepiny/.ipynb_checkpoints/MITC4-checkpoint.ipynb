# Skořepinový prvek MITC4

Formulace tohoto prvku vychází z prvku pro třídimenzionální analýzu vylepšeného tak, aby nedocházelo ke smykovému uzamknutí. Název představovaného čtyřúhelníkového prvku MITC4 plyne z anglického spojení Mixed Interpolation of Tensorial Components, do češtiny volně přeloženo jako smíšená interpolace tenzorových složek. Po praktické stránce z tohoto přístupu vyplývá, že je pro příčné smykové deformace použita jiná interpolace než pro zbylé složky vektoru deformací. Tento přístup byl navržen v článku A Continuum Mechanics Based Four-Node Shell Element for General Non-Linear Analysis (E. Dvorkin, K.J. Bathe, Eng. Comput.,1984, Vol. 1, 1983). 

### Základní vztahy
Geometrii prvku lze popsat následovně

$\displaystyle{
\begin{array}{lr}
x_i = \sum_{k=1}^4 h_k x_i^k + \frac{r_3}{2}\sum_{k=1}^4 a_k h_k V_{ni}^k,
&
\hspace{18cm}
(1)
\end{array}}$

<img src="figures/mitc.png" width=500/>



kde $h_k(r_1, r_2)$ jsou dvoudimenzionální interpolační fuknkce odpovídající uzlu $k$, $r_i$ jsou přirozené souřadnice, $x_i$ vyjadřuje souřadnice kartézského systému pro libovolný bod, $x_i^k$ odpovídá kartézským souřadnicím $i$-tého uzlu, $V_{ni}^k$ jsou složky řídícího vektoru v~uzlu a $a_k$ je tloušťka skořepiny v uzlu měřená podél řídícího vektoru. Pro posunutí libovolného bodu platí

$\displaystyle{
\begin{array}{lr}
u_i = \sum_{k=1}^4 h_k u_i^k + \frac{r_3}{2}\sum_{k=1}^4 a_k h_k (-V_{2i}^k \alpha_k + V_{1i}^k \beta_k),
&
\hspace{15cm}
(2)
\end{array}}
$

kde $u_i$ jsou posuny ve směru os $x_i$ a $\alpha$ a $\beta$ jsou rotace řídícího vektoru $V_{n}^k$ kolem $V_{1}^k$ a $V_{2}^k$, kde

$\displaystyle{V_{1}^k =\frac{\pmb e_2\times V_n^k}{\|\pmb e_2\times V_n^k\|},}$

$\displaystyle{V_{2}^k =V_n^k\times V_1^k,}$

Vektor $\pmb e_2$ je bázovým vektorem kartézského systému souřadnic. Vektor neznámých posunutí takto formulovaného prvku má tvar

$\displaystyle{
\pmb r_e = \{u_1^1, u_2^1, u_3^1, \alpha_1, \beta_1, u_1^2, u_2^2, u_3^2, \alpha_2, \beta_2, u_1^3, u_2^3, u_3^3, \alpha_3, \beta_3, u_1^4, u_2^4, u_3^4, \alpha_4, \beta_4\}^T.
}$

Pro interpolaci posunutí a geometrie jsou využity bilineární aproximační funkce, které mají na čtyřúhelníku tvar

$\displaystyle{
\begin{array}{lr}
\begin{array}{l}
 h_1 = \frac{1}{4}(1+r_1)(1+r_2),\\
 h_2 = \frac{1}{4}(1-r_1)(1+r_2),\\
 h_3 = \frac{1}{4}(1-r_1)(1-r_2),\\
 h_4 = \frac{1}{4}(1+r_1)(1-r_2).\\
\end{array}
&
\hspace{19.5cm}
(3)
\end{array}
}
$

Pro derivace $h_{k,i}$ interpolačních funkcí podle $r_i$ platí

$\displaystyle{
\begin{array}{lr}
{\begin{array}{ll}
h_{1,1} = \frac{\partial h_1}{r_1} =  \frac{1}{4}(1+r_2), \qquad & \qquad h_{1,2} = \frac{\partial h_1}{r_2} =  \frac{1}{4}(1+r_1),\\
h_{2,1} = \frac{\partial h_2}{r_1} =  -\frac{1}{4}(1+r_2), \qquad & \qquad h_{2,2} = \frac{\partial h_2}{r_2} =  \frac{1}{4}(1-r_1),\\
h_{3,1} = \frac{\partial h_3}{r_1} =  -\frac{1}{4}(1-r_2), \qquad & \qquad h_{3,2} = \frac{\partial h_3}{r_2} =  -\frac{1}{4}(1-r_1),\\
h_{4,1} = \frac{\partial h_4}{r_1} =  \frac{1}{4}(1-r_2), \qquad & \qquad h_{4,2} = \frac{\partial h_4}{r_2} =  -\frac{1}{4}(1+r_1).\\
\end{array}}
&
\hspace{12cm}
(4)
\end{array}
}$

### Sestavení prvku
Hlavním problémem takto formulovaného prvku je smykové zatuhnutí pro tenké konstrukce. Použitím interpolace posunutí dle (2) vychází příčné smykové deformace nenulové i pro namáhání konstantním ohybovým momentem. Z tohoto důvodu bylo navrženo nepočítat tyto deformace z posunutí vyjádřených ve vztahu (2), ale interpolovat tyto složky takovým způsobem, který by umožnil obrdržení konstantní aproximace smykových složek deformace podél směrů $r_1$ a $r_2$, tedy

$\displaystyle{
\begin{array}{lr}
\begin{array}{l}
\tilde\varepsilon_{13} = \frac{1}{2}(1+r_2)\tilde\varepsilon_{13}^A + \frac{1}{2}(1-r_2)\tilde\varepsilon_{13}^C,\\
\tilde\varepsilon_{23} = \frac{1}{2}(1+r_1)\tilde\varepsilon_{23}^D + \frac{1}{2}(1-r_1)\tilde\varepsilon_{23}^B,
\end{array}
&
\hspace{17.5cm}
(5)
\end{array}
}$

kde $\sim$ značí vyjádření v systému přirozených souřadnic a složky  $\tilde\varepsilon_{13}^A$, $\tilde\varepsilon_{23}^B$,   $\tilde\varepsilon_{13}^C$ a $\tilde\varepsilon_{23}^D$ odpovídají hodnotám příslušných deformací ve středech hran prvku spočtených z posunutí, tj. například

$\displaystyle{
\tilde\varepsilon_{13}^A = \left.\left(\frac{\partial u}{\partial r_3}+\frac{\partial w}{\partial r_1}\right)\right|_A.
}$

Vzhledem k tomu, že celková geometrická matice je vyjádřena v kartézském systému souřadnic, je nutné vztahy pro výpočet příčných smykových deformací transformovat do tohoto systému.

Je uvažován prvek se střednicovou plochou ležící v rovině $x,y$, uvažovaná poloha prvku v lokálním souřadném systému je znázorněna na obrázku. 

<img src="figures/mitc2d.png" width=300/>

V systému přirozených souřadnic $r_1, r_2$ a $r_3$ je vhodné zavést tzv. kovariantní a kontravariantní báze. Pro kovariantní bázové vektory $\pmb g_i$ $(i = 1,2,3)$ platí 

$\displaystyle{
\pmb g_i = \frac{\partial\pmb x}{\partial r_i}.
}$

Kontravariantní bázové vektory  $\pmb g^j$ $(j = 1,2,3)$ lze dopočítat ze vztahu

$
\displaystyle{\pmb g_i\pmb g^j = \delta_i^j,
}$

kde $\delta_i^j$ je tzv. Kroneckerovo delta,  $\delta_i^j = 1$ pro $i=j$ a $\delta_i^j = 0$ pro $i\ne j$.
Pro složky $\tilde\varepsilon_{ij}$ lze psát

$\displaystyle{
\begin{array}{lr}
\begin{array}{l}
\tilde \varepsilon_{ij} = \frac{1}{2}\left(\frac{\partial\pmb u}{\partial r_i}\pmb g_j + \frac{\partial\pmb u}{\partial r_j}\pmb g_i\right),\\
\tilde \varepsilon_{ij} = \frac{1}{2}\left(\frac{\partial\pmb u}{\partial r_i}\frac{\partial\pmb x}{\partial r_j} + \frac{\partial\pmb u}{\partial r_j}\frac{\partial\pmb x}{\partial r_i}\right).
\end{array}
&
\hspace{19.5cm}
(6)
\end{array}
}$

Dosazením z (1) a (2) do rovnice (6) a napočítáním konkrétních hodnot interpolačních funkcí a jejich derivací v bodech $A,B,C,D$ lze vyjádřit vztahy pro  $\tilde\varepsilon_{13}^A$, $\tilde\varepsilon_{23}^B$,   $\tilde\varepsilon_{13}^C$ a $\tilde\varepsilon_{23}^D$, a tedy

$\displaystyle{
\begin{array}{lr}
\begin{array}{rl}
\tilde\varepsilon_{13}^A =& \frac{1}{8} \left[  (\pmb u^1 - \pmb u^2)\cdot\frac{1}{2}(a_1V_n^1 + a_2V_n^2)  +\right.\\    
  &\quad \left.(\pmb x^1 - \pmb x^2)\cdot\frac{1}{2}(a_1(-V_2^1\alpha_1+V_1^1\beta_1) + a_2(-V_2^2\alpha_2+V_1^2\beta_2) )    \right],\\
\tilde\varepsilon_{13}^C =& \frac{1}{8} \left[  (\pmb u^4 - \pmb u^3)\cdot\frac{1}{2}(a_3V_n^3 + a_4V_n^4)  +\right.\\    
  &\quad \left.(\pmb x^4 - \pmb x^3)\cdot\frac{1}{2}(a_3(-V_2^3\alpha_3+V_1^3\beta_3) + a_4(-V_2^4\alpha_4+V_1^4\beta_4) )    \right],\\
\tilde\varepsilon_{23}^B =& \frac{1}{8} \left[  (\pmb u^1 - \pmb u^4)\cdot\frac{1}{2}(a_1V_n^1 + a_4V_n^4)  +\right.\\    
  &\quad \left.(\pmb x^1 - \pmb x^4)\cdot\frac{1}{2}(a_1(-V_2^1\alpha_1+V_1^1\beta_1) + a_4(-V_2^4\alpha_4+V_1^4\beta_4) )    \right],\\
\tilde\varepsilon_{23}^D =& \frac{1}{8} \left[  (\pmb u^2 - \pmb u^3)\cdot\frac{1}{2}(a_2V_n^2 + a_3V_n^3)  +\right.\\    
  &\quad \left.(\pmb x^2 - \pmb x^3)\cdot\frac{1}{2}(a_2(-V_2^2\alpha_2+V_1^2\beta_2) + a_3(-V_2^3\alpha_3+V_1^3\beta_3) )    \right].\\
\end{array}
&
\hspace{11.5cm}
(7)
\end{array}
}$

Výsledné vztahy pro $\tilde\varepsilon_{13}$ a $\tilde\varepsilon_{23}$ lze vyjádřit dosazením vztahů (7) do rovnic (5). Pro transformaci do kartézského systému souřadnic lze využít vztahu

$\displaystyle{
\tilde\varepsilon_{ij}\pmb g^i\pmb g^j = \varepsilon_{kl}\pmb e_k\pmb e_l, 
}$

kde $\varepsilon_{kl}$ jsou složky tenzoru deformací v kartézském souřadném systému s bázovými vektory $\pmb e_k$ a $\pmb e_l$. Uvážením symetrie tenzoru deformací lze vyjádřit složky vektoru deformací $\gamma_{xz}$ a $\gamma_{yz}$, a to následovně

$\displaystyle{
\begin{array}{lr}
\begin{array}{l}
\gamma_{xy} =2\tilde\varepsilon_{13}(\pmb g^1 \cdot\pmb e_1 )(\pmb g^3 \cdot\pmb e_3 ) + 2\tilde\varepsilon_{23}(\pmb g^2 \cdot\pmb e_1 )(\pmb g^3 \cdot\pmb e_3 ),\\
\gamma_{yz} =2\tilde\varepsilon_{13}(\pmb g^1 \cdot\pmb e_2)(\pmb g^3 \cdot\pmb e_3 ) + 2\tilde\varepsilon_{23}(\pmb g^2 \cdot\pmb e_2 )(\pmb g^3 \cdot\pmb e_3 )
\end{array}
&
\hspace{15cm}
(8)
\end{array}
}$

Výsledné vztahy pro příčné smykové deformace lze obdržet dosazením z (5) a (7) do (8). 

Pro ostatní složky vektoru deformace lze využít vztahy plynoucí z teorie 3D pružnosti, tedy

$\displaystyle{
\begin{array}{lr}
\begin{array}{l}
\varepsilon_{x} = \frac{\partial u_1}{\partial x_1},\\
\varepsilon_{y} = \frac{\partial u_2}{\partial x_2},\\
\gamma_{xy} = \frac{\partial u_1}{\partial x_2} + \frac{\partial u_2}{\partial x_1}.
\end{array}
&
\hspace{21.25cm}
(9)
\end{array}
}$

Pro derivace posunů $u_1$ a $u_2$ podle přirozených souřadnic platí

$\displaystyle{
\begin{array}{lr}
\begin{array}{l}
\frac{\partial u_i}{\partial r_1} = \sum_{k=1}^4 h_{k,1} u_i^k + \frac{r_3}{2}\sum_{k=1}^4 a_k h_{k,1} (-V_{2i}^k \alpha_k + V_{1i}^k \beta_k),\\
\frac{\partial u_i}{\partial r_2} = \sum_{k=1}^4 h_{k,2} u_i^k + \frac{r_3}{2}\sum_{k=1}^4 a_k h_{k,2} (-V_{2i}^k \alpha_k + V_{1i}^k \beta_k),\\
\frac{\partial u_i}{\partial r_3} =  \frac{1}{2}\sum_{k=1}^4 a_k h_{k} (-V_{2i}^k \alpha_k + V_{1i}^k \beta_k),
\end{array}
&
\hspace{14.5cm}
(10)
\end{array}
}$

kde $h_{k,i}$ jsou derivace interpolačních funkcí. Z věty o derivaci složené funkce plyne

$\displaystyle{
\begin{array}{lr}
\left\{
\begin{array}{c}
\frac{\partial}{\partial x_1}\\
\frac{\partial}{\partial x_2}\\
\frac{\partial}{\partial x_3}
\end{array} \right\}=\pmb J^{-1}
\left\{
\begin{array}{c}
\frac{\partial}{\partial r_1}\\
\frac{\partial}{\partial r_2}\\
\frac{\partial}{\partial r_3}
\end{array} \right\} ,
&
\hspace{20cm}
(11)
\end{array}
}$

kde $\pmb J$ je tzv. Jakobián 

$\displaystyle{
\pmb J = 
\left[
\begin{array}{ccc}
\frac{\partial x_1}{\partial r_1}  &\frac{\partial x_2}{\partial r_1}  & \frac{\partial x_3}{\partial r_1} \\
\frac{\partial x_1}{\partial r_2}  &\frac{\partial x_2}{\partial r_2}  & \frac{\partial x_3}{\partial r_2} \\
\frac{\partial x_1}{\partial r_3}  &\frac{\partial x_2}{\partial r_3}  & \frac{\partial x_3}{\partial r_3} 
\end{array} \right] .
}$

Výsledné vztahy pro složky vektoru deformací $\varepsilon_{x}$, $\varepsilon_{y}$, $\varepsilon_{xy}$ lze vyjádřit dosazením z (10) a (11) do (9).

### Geometrická matice

Z ppředchozích přednášek víme, že geometrická matice $\pmb B$ zprostředkovává vztah mezi vektorem deformací a vektorem neznámých posunutí

$\displaystyle{
\pmb \varepsilon = \pmb B \pmb r_e.
}$

Geometrickou matici pro tento prvek lze tedy sestavit vytknutím jednotlivých stupňů volnosti z výše uvedených formulací složek vektoru deformace.


### Matice tuhosti

Celková matice tuhosti je opět dána vztahem

$\displaystyle{
\pmb K = \int_{V} {\pmb B}^T \pmb D \pmb B dV,
}$

kde konstitutivní matice $\pmb D$ zohledňuje vlastnosti skořepin. Při praktickém výpočtu byla použita Gaussova [numerická integrace](../Doplňky/Numerická%20integrace/Numerická%20integrace.ipynb) s osmi (2x2x2) integračními body.

#### Transformace do prostoru
Vzhledem k tomu, že předvedená formulace prvku uvažovala střednicovou plochu v rovině x,y, je opět nutné provést transformaci do prostoru.  Pro převod vektoru posunutí a matice tuhosti z lokálního systému do globálního platí

$\displaystyle{
\pmb r_L = \pmb A \pmb r_G,
}$

$\displaystyle{
\pmb K_G = \pmb A^T \pmb K_L \pmb A.
}$

kde

$\displaystyle{{\pmb A} =  
\left[\pmb {\begin{array}{cccccc}
 a & 0 & 0 & 0 & 0 & 0\\
0 & b & 0 & 0 & 0 & 0 \\
0 & 0 & a & 0 & 0 & 0 \\
0 & 0 & 0 & b & 0 & 0 \\
0 & 0 & 0 & 0 & a & 0 \\
0 & 0 & 0 & 0 & 0 & b
\end{array}} \right].
}$

Pro převod rotací je zde použita submatice $\pmb b$, která zohledňuje pouze dvě rotace ve vektoru posunutí v lokálním systému. V globálním souřadném systému je ve výsledku uvažováno šest stupňů volnosti v každém uzlu. Pro submatice $\pmb a$ a $\pmb b$ platí

$\displaystyle{{\pmb a} =  
\left[ {\begin{array}{ccc}
 cos(x,X)& cos(x,Y)& cos(x,Z)\\
 cos(y,X)& cos(y,Y)& cos(y,Z)\\
 cos(z,X)& cos(z,Y)& cos(z,Z)\\
\end{array}} \right],
}$

$\displaystyle{{\pmb b} =  
\left[ {\begin{array}{ccc}
 cos(x,X)& cos(x,Y)& cos(x,Z)\\
 cos(y,X)& cos(y,Y)& cos(y,Z)\\
\end{array}} \right].
}$

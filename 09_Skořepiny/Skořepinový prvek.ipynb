# Skořepinový prvek

Jednou z možností, jak zkonstruovat plochý prvek pro analýzu skořepinových konstrukcí, je kombinace prvku pro řešení ohýbaných konstrukcí a prvku pro analýzu membránového působení. V této přednášce je tímto způsobem sestrojen trojúhelníkový prvek s pěti stupni volnosti v každém uzlu. Část přenášející ohybové účinky zde reprezentuje [DKT](../08_Desky/Discrete%20Kirchhoff%20Triangle.ipynb) prvek, který byl představen v rámci deskových prvků. Pro přenos membránových účinků zatížení je použit stěnový prvek s lineární aproximací (viz kapitola [Rovinná napjatost](../07_Rovinné%20úlohy/include/Rovinná%20napjatost.ipynb)). 

<img src="figures/stenaDKT.png" width = 750>

Vzhledem k tomu, že každý z prvků, ze kterých se při formulaci vychází, závisí na jiných stupňích volnosti, lze výsledné vztahy pro skořepinový prvek získat velice snadno. Je třeba sestavit celkovou matici tuhosti, která zahrnuje oba stavy. Taková matice vznikne lokalizací matice tuhosti úlohy rovinné napjatosti $\boldsymbol{K}_{RN}$ ($\boldsymbol{r}_{RN} = u,v$) a matice deskového prvku $\boldsymbol{K}_{DKT}$ ($\boldsymbol{r}_{DKT} = w, \varphi_x, \varphi_y$). 
Trojúhelníkovému skořepinovému prvku pak odpovídá vektor neznámých posunutí uvažující pět stupňů volnosti v každém uzlu, který budeme řadit následovně

$
\displaystyle{
\pmb r_e = \{u_1, v_1, w_1, \varphi_{x1}, \varphi_{y1}, u_2, v_2, w_2, \varphi_{x2}, \varphi_{y2}, u_3, v_3,, w_3, \varphi_{x3}, \varphi_{y3}\}^T
}
$

a s ohledem na toto řazení budeme lokalizovat  $\boldsymbol{K}_{RN}$ a $\boldsymbol{K}_{DKT}$ do výsledné lokální matice tuhosti skořepinového prvku  $\boldsymbol{K}_L$.

### Transformace do prostoru
Až doposud byl prvek formulován v rovině. Pro použití prvku pro analýzu skořepinových konstrukcí je nutné provést transformaci do prostoru. Pro převod vektoru z lokálního systému do globálního platí

$\displaystyle{
\pmb r_L = \pmb T \pmb r_G,
}$

kde
$\displaystyle{{\pmb T} =  
\left[\pmb {\begin{array}{cccccc}
 T_u & 0 & 0 & 0 & 0 & 0\\
0 & T_\varphi & 0 & 0 & 0 & 0 \\
0 & 0 & T_u & 0 & 0 & 0 \\
0 & 0 & 0 & T_\varphi & 0 & 0 \\
0 & 0 & 0 & 0 & T_u & 0 \\
0 & 0 & 0 & 0 & 0 & T_\varphi
\end{array}} \right].
}$

Submatice $\pmb T_u$ je složená ze směrových kosínů mezi osami $x, y, z$ lokálního a $X, Y, Z$  globálního systému.
Pro převod rotací je zde použita submatice $\pmb T_\varphi$, která zohledňuje existenci pouze dvou rotačních stupňů volnosti ve vektoru posunutí v lokálním systému. V globálním souřadném systému je ve výsledné matici uvažováno šest stupňů volnosti v každém uzlu. Pro submatice $\pmb T_u$ a $\pmb T_\varphi$ platí

$\displaystyle{{\pmb T_u} =  
\left[ {\begin{array}{ccc}
 cos(x,X)& cos(x,Y)& cos(x,Z)\\
 cos(y,X)& cos(y,Y)& cos(y,Z)\\
 cos(z,X)& cos(z,Y)& cos(z,Z)\\
\end{array}} \right],
}$

$\displaystyle{{\pmb T_\varphi} =  
\left[ {\begin{array}{ccc}
 cos(x,X)& cos(x,Y)& cos(x,Z)\\
 cos(y,X)& cos(y,Y)& cos(y,Z)\\
\end{array}} \right].
}$


Globální matici tuhosti lze pak získat ze vztahu

$\displaystyle{
\pmb K_G = \pmb T^T \pmb K_L \pmb T.
}$

# Úvod do teorie pružnosti

$$
\newcommand{\del}[2]{\mbox{$\displaystyle\frac{#1}{#2}$}}
$$


Cílem prostorové úlohy pružnosti je určit v tělese objemu $\Omega$ s hranicí $\Gamma$ tři základní pole:
 * pole posunů  $\boldsymbol u= \{u, v, w\}^T$
 * pole deformace $\boldsymbol \varepsilon = \{\varepsilon_x, \varepsilon_y, \varepsilon_z, \gamma_{yz}, \gamma_{zx}, \gamma_{xy}\}^T $,
 * pole napětí $\boldsymbol \sigma= \{\sigma_x, \sigma_y, \sigma_z, \tau_{yz}, \tau_{zx}, \tau_{xy}\}^T$. 

Pro řešení úlohy máme k dispozici tři sady rovnic - geometrické, statické a fyzikální.
    
### Geometrické rovnice

Geometrické rovnice popisují vztahy mezi polem posunutí a polem deformace
$
{\boldsymbol \varepsilon = \boldsymbol\partial^T \boldsymbol u},
$


$$
\displaystyle{
\left\{
\begin{array}{c}
\varepsilon_x \\
\varepsilon_y \\
\varepsilon_z \\
\gamma_{yz} \\
\gamma_{zx} \\
\gamma_{xy}
\end{array}
\right\}
 =
 \left[
\begin{array}{ccc}
\frac{\partial}{\partial x} & 0 & 0  \\
0 & \frac{\partial}{\partial y} & 0  \\
0 & 0 & \frac{\partial}{\partial z}   \\
0 & \frac{\partial}{\partial z} &  \frac{\partial}{\partial y} \\
\frac{\partial}{\partial z} & 0 & \frac{\partial}{\partial x} \\
\frac{\partial}{\partial y} &  \frac{\partial}{\partial x} & 0
\end{array}
\right]
\left\{
\begin{array}{c}
u\\
v\\
w
\end{array}
\right\}
}
$$

### Statické rovnice

Statické rovnice, zvané také Cauchyho rovnice nebo podmínky rovnováhy, popisují vztahy mezi polem napětí a polem objemových sil
$
{\boldsymbol\partial \boldsymbol \sigma + \boldsymbol{ \bar{X}} = \boldsymbol 0},
$

<img src="Figures/equlibrium-x.png" width=450>

$$
\displaystyle{
 \left[
\begin{array}{cccccc}
\frac{\partial}{\partial x} & 0 & 0 & 0 & \frac{\partial}{\partial z} &  \frac{\partial}{\partial y} \\
0 & \frac{\partial}{\partial y} & 0 & \frac{\partial}{\partial z} & 0 & \frac{\partial}{\partial x}\\
0 & 0 & \frac{\partial}{\partial z} &  \frac{\partial}{\partial y} &  \frac{\partial}{\partial x} & 0
\end{array}
\right]
\left\{
\begin{array}{c}
\sigma_x \\
\sigma_y \\
\sigma_z \\
\tau_{yz} \\
\tau_{zx} \\
\tau_{xy}
\end{array}
\right\}
 +\left\{
\begin{array}{c}
\bar{X}\\
\bar{Y}\\
\bar{Z}
\end{array}
\right\}
=
\left\{
\begin{array}{c}
0\\
0\\
0
\end{array}
\right\}
}
$$

### Fyzikální rovnice

Fyzikální rovnice popisují vztah mezi polem deformace a polem napětí. O těchto rovnicích často mluvíme jako o konstitutivních vztazích, pro lineárně pružný materiál platí

$
\boldsymbol\sigma = \boldsymbol D \boldsymbol \varepsilon,
$

kde $\boldsymbol D$ je matice tuhosti daného materiálu. Pro lineární izotropní materiál má tato matice tvar
$$
\boldsymbol D =
\frac{E}{(1+\nu)(1-2\nu)}
\left[
\begin{array}{cccccc}
1-\nu & \nu & \nu & 0 & 0 & 0\\
\nu & 1-\nu & \nu & 0 & 0 & 0\\
\nu & \nu & 1-\nu & 0 & 0 & 0\\
 0 & 0 & 0 & \frac{1-2\nu}{2} & 0 & 0\\
 0 & 0 & 0 & 0 & \frac{1-2\nu}{2} & 0\\
 0 & 0 & 0 & 0 & 0 & \frac{1-2\nu}{2}
\end{array}
\right].
$$

Inverzní vztah je zprostředkován maticí poddajnosti materialu $\boldsymbol {C}$, která je inverzí matice tuhosti $\boldsymbol{D}$

$
\boldsymbol\varepsilon = \boldsymbol C \boldsymbol\sigma,
$

$$\displaystyle{
\boldsymbol{C} = \boldsymbol{D}^{-1}=
\left[
\begin{array}{cccccc}
\frac{1}{E} & \frac{-\nu}{E}& \frac{-\nu}{E} & 0 & 0 & 0\\
\frac{-\nu}{E} & \frac{1}{E}& \frac{-\nu}{E} & 0 & 0 & 0\\
\frac{-\nu}{E} & \frac{-\nu}{E}& \frac{1}{E} & 0 & 0 & 0\\
 0 & 0 & 0 & \frac{1}{G} & 0 & 0\\
 0 & 0 & 0 & 0 & \frac{1}{G} & 0\\
 0 & 0 & 0 & 0 & 0 & \frac{1}{G}
\end{array}
\right].
}$$

### Okrajové podmínky

Nedílnou součástí formulace problému jsou okrajové podmínky. Hranici $\Gamma$ rozdělíme na část $\Gamma_u$, kde jsou předepsané kinematické okrajové podmínky, a na část  $\Gamma_p$ s předepsanými statickými (silovými) okrajovými podmínkami.

#### Kinematické okrajové podmínky

Na hranici $\Gamma_u$ s předepsanými posuny $\boldsymbol{\bar{u}}$ platí

$
\boldsymbol u - \boldsymbol{\bar{u}} = \boldsymbol 0.
$

#### Statické okrajové podmínky

Na hranici $\Gamma_p$ s předepsanými povrchovými silami $\boldsymbol{\bar{p}}$ platí

$
\boldsymbol n \boldsymbol \sigma - \boldsymbol{\bar{p}} =\boldsymbol 0,
$

kde $\boldsymbol n$ je matice směrových kosinů $n_x, n_y, n_z$ vnější normály k povrchu
$$
\left[ \begin{array}{cccccc}
n_x & 0 & 0 & 0 & n_z & n_y\\
0 & n_y & 0 & n_z & 0 & n_x\\
0 & 0 & n_z & n_y & n_x & 0 
\end{array} \right]
\left\{ \begin{array}{c}
\sigma_x \\
\sigma_y \\
\sigma_z \\
\tau_{yz} \\
\tau_{zx} \\
\tau_{xy}
\end{array} \right\} -
\left\{ \begin{array}{c}
\bar{p}_x\\
\bar{p}_y\\
\bar{p}_z
\end{array} \right\} =
\left\{ \begin{array}{c}
0\\
0\\
0
\end{array} \right\}
$$

### Diferenciální rovnice problému

Budeme se zabývat deformační variantou MKP, primární neznámé tedy budou posuny $\boldsymbol u$, deformace a napětí dopočítáme následně. Hledáme $\boldsymbol u$ splňující kinematické i statické okrajové podmínky takové, aby splňovalo podmínky rovnováhy

$
{\boldsymbol\partial \boldsymbol \sigma + \boldsymbol{ \bar{X}} = \boldsymbol 0},
$

dosazením z konstitutivních vztahů získáme 

$
{\boldsymbol\partial (\boldsymbol D \boldsymbol \varepsilon) + \boldsymbol{ \bar{X}} = \boldsymbol 0}
$

a následně aplikací geometrických rovnic plyne

$
{\boldsymbol\partial (\boldsymbol D \boldsymbol\partial^T \boldsymbol u) + \boldsymbol{ \bar{X}} = \boldsymbol 0}
$

Řešení $\boldsymbol u$ splňující tyto rovnice nazýváme silným řešením.

## Jednorozměrná úloha - tažený-tlačený prut

Budeme uvažovat prut ležící v ose $x$ a přijmeme následující předpoklady:

<img src="Figures/1dbar_equlibrium" width = 500>


 * zatížení působí pouze v ose prutu,
 * ve směrech $y, z$ se prut může volně rozpínat,
 * jedinou nenulovou složkou napětí je $\sigma_x$.
 * průřez je homogenní z izotropního materiálu (napětí je tedy konstantní po průřezu)
 * prut je zatížen objemovým zatížením $X$ jenž je konstantní po průřezu, jeho výslednici na průřezu označíme $f_x(x) = A(x)X(x)$


Dosazením třetího předpokladu do konstitutivních vztahů vyjádříme složky vektoru deformace
$$
\displaystyle{
\left\{
\begin{array}{c}
\varepsilon_x \\
\varepsilon_y \\
\varepsilon_z \\
\gamma_{yz} \\
\gamma_{zx} \\
\gamma_{xy}
\end{array}
\right\}
=
\left[
\begin{array}{cccccc}
\frac{1}{E} & \frac{-\nu}{E}& \frac{-\nu}{E} & 0 & 0 & 0\\
\frac{-\nu}{E} & \frac{1}{E}& \frac{-\nu}{E} & 0 & 0 & 0\\
\frac{-\nu}{E} & \frac{-\nu}{E}& \frac{1}{E} & 0 & 0 & 0\\
 0 & 0 & 0 & \frac{1}{G} & 0 & 0\\
 0 & 0 & 0 & 0 & \frac{1}{G} & 0\\
 0 & 0 & 0 & 0 & 0 & \frac{1}{G}
\end{array}
\right]
}
\left\{
\begin{array}{c}
\sigma_x \\
0 \\
0 \\
0 \\
0 \\
0
\end{array}
\right\}.
$$

Získáme tři nenulové složky deformace, které jsou na sobě závislé, složky $\varepsilon_y, \varepsilon_z$ můžeme tedy vyjádřit pomocí $\varepsilon_x$

$
\displaystyle{
\varepsilon_x = \frac{1}{E}\sigma_x,\\
\varepsilon_y = -\frac{\nu}{E}\sigma_x = -\nu \varepsilon_x,\\
\varepsilon_z = -\frac{\nu}{E}\sigma_x = -\nu \varepsilon_x.
}$

Pro vyjádření deformace $\varepsilon_x$ použijeme geometrické rovnice 

$\displaystyle{
\varepsilon_x = \frac{du(x)}{dx},
}$

posun $u(x)$ bude tedy naší primární neznámou. 

### Diferenciální rovnice



Hledáme řešení $u(x)$ dostatečně hladké, tak aby splňovalo 


 Pro $x \in \Omega$:

$\quad\displaystyle{
\frac{d}{\ dx}
\left(
E(x) A(x) \frac{\ du}{\ dx}(x) 
\right)
+
b(x) = 0,
}$

 pro $x \in \Gamma^u$: $\displaystyle u(x) = \overline{u}(x)$,

 pro $x \in \Gamma^t$: $\displaystyle A(x)\sigma(x) n(x) = E(x)A(x) \frac{\ du}{\ dx}(x) n(x) = \overline{t}(x)$.

Takové pole posunutí $u(x)$ které splňuje tyto rovnice se nazývá <b>silné řešení</b>.

### Silné řešení

Hledáme $u(x)$, které splňuje

$\displaystyle{EA\left(\frac{d^2 u(x)}{\ dx^2}\right) + 1,\ \ \ x\in(0,2),\ \ u(0)=0,\ N(2)=AE\frac{\ du(2)}{\ dx}=0}$

Integrací postupně dostaneme

$\displaystyle{EA\frac{\ du(x)}{\ dx} + 1x + C_1 = 0,\ \ \ \ EA u(x) + 1\frac{x^2}{2} + C_1 x + C_2= 0,}$


integrační konstanty $C_1$ a $C_2$ určíme z okrajových podmínek

$\displaystyle{ u(0)=0: C_2 = 0,\ \ \ \ N(2)=0: AE\frac{\ du(2)}{\ dx}=-C_1 - 2=0: C_1=-2}.$

Konečně tedy dostáváme

$\displaystyle{ u(x) = \frac{1}{EA} (2x-x^2/2),\ \ \ N(x) =  AE\frac{\ du(x)}{\ dx} = 2-x}.$

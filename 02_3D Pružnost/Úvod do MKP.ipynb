
### Metoda vážených reziduí

<i>
Připomeneme si integraci per partes 

$\displaystyle{
\begin{array}{rcl}
\displaystyle{\int_0^L\frac{\ df}{\ dx}(x) g(x)\ dx}
& = &\displaystyle{\left[f(x) g(x)\right]_0^L -
\int_0^L f(x)\frac{\ dg}{\ dx}(x)\ dx },
\\ \displaystyle{\int_\Omega
\frac{\ df}{\ dx}(x) g(x) \ dx}
& = &\displaystyle{
\int_\Gamma f(x) g(x) n(x) \ dx - \int_\Omega f(x) \frac{\ dg}{\ dx}(x) \ dx}.
\end{array}
}$
</i>


 Definujme reziduum řešení pro danou funkci $v(x)$

$\displaystyle{
\begin{array}{rcl}
x \in \Omega & : & 
r\left(v(x)\right) = 
\frac{d}{\ dx}
\bigl(
E(x) A(x) \frac{\ dv}{\ dx}(x) 
\bigr)
+
b(x) 
\\
x \in \Gamma^u & : & 
r\left(v(x)\right) = \overline{u}(x) - v(x) 
\\
x \in \Gamma^t & : & 
r\left(v(x)\right) 
= \overline{t}(x) - E(x)A(x) \frac{\ dv}{\ dx}(x) n(x) 
\end{array}
}$

Pokud je $v(x)$ řešením problému ($v(x) \equiv u(x)$) máme $r(x) \equiv 0$.

Idea <b>metody vážených reziduí</b>: Zvolme libovolnou funkci $\delta u(x)$ (tzv. váhová funkce (test, weight function)) a spočtěme

$$
\int_\Omega \delta u(x) r(v(x)) \ dx
+
\int_\Gamma \delta u(x) r(v(x)) \ dx.
$$

Jestliže hodnota integrálu bude rovna nule pro <b> všechny</b> váhové funkce $\delta
u(x)$, potom $v(x)$ je řešením problému.


#### Slabé řešení

Hledáme řešení $u(x)$ rovnice takové, které splňuje rovnici

$\displaystyle{
0 = \int_{\Omega} \delta u \left(  \frac{d}{dx}\left(EA\frac{d u}{dx}\right)+f_x \right) \ dx
}$

pro všechna $\delta u $ splňující kinematické okrajové podmínky ($\delta u = 0$). Použitím integrace per-partes získáme


$\displaystyle{
0 = \int_{\Gamma}\delta u \left( EA\frac{d u}{dx}\right)n \ dx-\int_{\Omega} \frac{d(\delta u)}{dx} \left( EA\frac{d u}{dx}\right) \ dx + \int_{\Omega}  \delta u \bar{f_x} \ dx,
}$

kde $n$ je vnější normála k hranici $\Gamma$. Hraniční integrály můžeme rozdělit na část hranice $\Gamma_u$, kde jsou předepsány kinematické okrajové podmínky, a na část $\Gamma_p$, kde jsou aplikovány silové okrajové podmínky

$\displaystyle{
0 =  \int_{\Gamma_u}\underbrace{\delta u}_{= 0} \left( EA\frac{d u}{dx}\right)n \ dx +
\int_{\Gamma_p}\delta u \underbrace{\left( EA\frac{d u}{dx}\right)n}_{= \bar{N}} \ dx
-\int_{\Omega} \frac{d(\delta u)}{dx} \left( EA\frac{d u}{dx}\right) \ dx + \int_{\Omega}  \delta u \bar{f_x} \ dx,
}$

Úpravou výrazů dostaneme finální tvar rovnic

$\displaystyle{
\int_{\Omega} \frac{d(\delta u)}{dx} EA \frac{d u}{dx} \ dx = \left[\delta u \bar{N}\right]_{\Gamma_p} + \int_{\Omega}  \delta u \bar{f_x} \ dx.
}
$

### Analogie metody vážených reziduí a Principu minima potenciální energie

Ze všech kinematicky přípustných stavů pružného tělesa nastává ten, který dává potenciální energii systému minimální hodnotu

$\displaystyle{
\Pi = E_i+E_e = min,
}$

kde

$\displaystyle{\begin{array}{rcl} 
\displaystyle{E_i}&=&\displaystyle{\frac{1}{2}\int_\Omega\sigma\varepsilon\,d\Omega,}\\
\displaystyle{E_e}&=&\displaystyle{-\int_\Omega ub\,d\Omega-\int_{\Gamma^t}u\bar{t}\,d\Gamma}.
\end{array}
}$

$\Pi$ je funkcionál (funkce funkcí). Jak se bude měnit hodnota $\Pi$, když se bude měnit $u(x)$? Malou změnu funkce budeme nazývat její variací a vyjádříme ji jako $\delta u(x)=\xi w(x)$, kde $w(x)$ je libovolná funkce respektující kinematické okrajové podmínky a $\xi$ malé nezáporné číslo. 

Odpovídající změna funkcionálu, která se nazývá jeho variací je definována jako

$\displaystyle{
\delta\Pi = \Pi(u(x)+\xi w(x))-\Pi(u(x))\equiv\Pi(u(x)+\delta u(x))-\Pi(u(x)).
}$


My hledáme minimum, tedy variace musí být nulová a tedy $\delta\Pi=0$.

<img src ="Figures/fig6c.png" width=500/>


Pro náš případ

$\displaystyle{
\begin{array}{rcl}
\delta\Pi&=&\displaystyle{\frac{1}{2}\int AE \left(\frac{du}{dx}+\xi\frac{dw}{dx}\right)^2\,dx-\frac{1}{2}\int AE \left(\frac{du}{dx}\right)^2\,dx}\\
&&\displaystyle{-\int (u+\xi w)b\,dx-\int (u+\xi w)\bar{t}\,d\Gamma+\int (u)b\,dx+\int (u)\bar{t}\,d\Gamma}\\
&=&\displaystyle{\frac{1}{2}\int AE\left(\left(\frac{du}{dx}\right)^2+2\xi\frac{du}{dx}\frac{dw}{dx}+\xi^2\left(\frac{dw}{dx}\right)^2\right)\,dx}\\
&&\displaystyle{ -\frac{1}{2}\int AE\left(\frac{du}{dx}\right)^2\,dx-\xi\int wb\,dx-\xi\left.(w\bar{t})\right\vert_\Gamma}
\end{array}
}$

$\displaystyle{
\begin{array}{rcl}
\displaystyle{\delta\Pi}&=&\displaystyle{\xi\int AE \frac{dw}{dx}\frac{du}{dx}\,dx-\xi\int wb\,dx-\xi(w\bar{t})\vert_\Gamma}
\end{array}
}$

Hledáme minimum, tedy variace musí být nulová ($\delta\Pi=0$). Dosazením z předchozích výrazů a podělením $\xi$ máme

$\begin{array}{rcl}
\displaystyle{\delta\Pi\xi} &=& \displaystyle{\int_\Omega AE \frac{dw}{dx}\frac{du}{dx}\,dx-\int_\Omega wb\,dx-(w\bar{t})\vert_\Gamma = 0}\\
\displaystyle{\delta\Pi} &=&\displaystyle{ \int_\Omega AE \frac{d\delta u}{dx}\frac{du}{dx}\,dx-\int_\Omega \delta ub\,dx-(\delta u\bar{t})\vert_\Gamma = 0}\\
\end{array}
$

To lze dále zjednodušit. Pokud označíme $\delta\varepsilon=\frac{dw}{dx}$ ($\delta\varepsilon$ budeme nazývat virtuální deformací) a dosazením konstitutivních vztahů obdržíme

$\displaystyle{
\delta\Pi = \int_\Omega A \sigma\delta\varepsilon\,dx-\int_\Omega \delta ub\,dx-(\delta u\bar{t})\vert_\Gamma = 0
}$

A to je známý <b>"Princip virtuálních posunutí"</b>.

Všimněte si ekvivalence odvozených výrazů a výrazů získaných aplikací metody vážených reziduí.

### Diskretizovaná úloha

Konstrukci rozdělíme na $n$ dílků (prvků). Neznámé posuny $u$ budeme na každém prvku aproximovat jako součin zvolených interpolačních funkcí $N_i$ a příslušných (neznamých) vah $u_i$ (často nazývaných uzlové hodnoty). Stejně budeme aproximovat i průběh váhové funkce $\delta u$

$\displaystyle{
u \approx \sum N^e_i u_i = \boldsymbol{N}^e\boldsymbol {r}^e, 
\quad 
\delta u \approx \sum N^e_i \delta u_i = \boldsymbol{N}^e\boldsymbol {w}^e, 
}$

kde $\boldsymbol{N}^e$ je matice interpolačních funkcí $N_i$ prvku, $\boldsymbol {r}^e$ je vektor uzlových hodnot posunů na prvku $e$ a $\boldsymbol {w}^e$ je vektor uzlových hodnot váhových funkcí $\delta u$.
Pro derivace pak platí

$\displaystyle{
\frac{du}{dx} = \frac{d}{dx}(\sum N^e_i u_i) = \sum \frac{dN^e_i}{dx} u_i = \boldsymbol{B}^e\boldsymbol {r}^e, 
\quad
\frac{d(\delta u)}{dx} = \boldsymbol{B}^e\boldsymbol {w}^e.
}$

kde matice $\boldsymbol{B}^e$ obsahuje derivace interpolačních funkcí.

Integrál slabého řešení na celé oblasti rozdělíme na sumu integrálů přes jednotlivé prvky a po úpravě získáme 

$\displaystyle{
\sum_{e=1}^n\left\{
\int_{\Omega} \boldsymbol{B}^e\boldsymbol{w}^e EA  \boldsymbol{B}^e\boldsymbol {r}^e \ dx = \left[\boldsymbol{N}^e\boldsymbol {w}^e\bar{N}\right]_{\Gamma_p} + \int_{\Omega} \boldsymbol{N}^e\boldsymbol {w}^e \bar{f_x} \ dx \right\},\\
\sum_{e=1}^n\left\{
\boldsymbol{w}^e \left(\int_{\Omega} \boldsymbol{B}^e EA  \boldsymbol{B}^e\ dx\ \boldsymbol {r}^e \right) =\boldsymbol{w}^e \left( \left[\boldsymbol{N}^e \bar{N}\right]_{\Gamma_p} + \int_{\Omega} \boldsymbol{N}^e \bar{f_x} \ dx\right) \right\},\\
\sum_{e=1}^n\left\{
\underbrace{\int_{\Omega} \boldsymbol{B}^e EA  \boldsymbol{B}^e \ dx}_{\boldsymbol{K^e}}\ \boldsymbol{r}^e  =\underbrace{ \left[\boldsymbol{N}^e \bar{N}\right]_{\Gamma_p}}_{\boldsymbol{f}_{\Gamma}^e} + \underbrace{ \int_{\Omega} \boldsymbol{N}^e \bar{f_x} \ dx}_{\boldsymbol{f}_{\Omega}^e}\right\}.
}
$

### Lokalizace
Rozšířením lokálních vektorů $\boldsymbol{u^e}$ do globálního vektoru $\boldsymbol{u}$ uzlových hodnot můžeme psát

$
\boldsymbol{K^gr}=\boldsymbol{f^g}.
$

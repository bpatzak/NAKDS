Cílem prostorové úlohy pružnosti je určit v tělese objemu $\Omega$ s hranicí $\Gamma$ tři základní pole:
<!-- * pole posunů  $\boldsymbol u(\boldsymbol x)= \{u(\boldsymbol x), v(\boldsymbol x), w(\boldsymbol x)\}^T$ ,-->
 * pole posunů  $\boldsymbol u= \{u, v, w\}^T$
 * pole deformace $\boldsymbol \varepsilon = \{\varepsilon_x, \varepsilon_y, \varepsilon_z, \gamma_{yz}, \gamma_{zx}, \gamma_{xy}\}^T $,
 * pole napětí $\boldsymbol \sigma= \{\sigma_x, \sigma_y, \sigma_z, \tau_{yz}, \tau_{zx}, \tau_{xy}\}^T$. 

Pro řešení úlohy máme k dispozici tři sady rovnic - geometrické, statické a fyzikální.
    
### Geometrické rovnice

Geometrické rovnice popisují vztahy mezi polem posunutí a polem deformace
$
{\boldsymbol \varepsilon = \boldsymbol\partial^T \boldsymbol u},
$


$$
\displaystyle{
\left\{
\begin{array}{c}
\varepsilon_x \\
\varepsilon_y \\
\varepsilon_z \\
\gamma_{yz} \\
\gamma_{zx} \\
\gamma_{xy}
\end{array}
\right\}
 =
 \left[
\begin{array}{ccc}
\frac{\partial}{\partial x} & 0 & 0  \\
0 & \frac{\partial}{\partial y} & 0  \\
0 & 0 & \frac{\partial}{\partial z}   \\
0 & \frac{\partial}{\partial z} &  \frac{\partial}{\partial y} \\
\frac{\partial}{\partial z} & 0 & \frac{\partial}{\partial x} \\
\frac{\partial}{\partial y} &  \frac{\partial}{\partial x} & 0
\end{array}
\right]
\left\{
\begin{array}{c}
u\\
v\\
w
\end{array}
\right\}
}
$$

Statické rovnice, zvané také Cauchyho rovnice nebo podmínky rovnováhy, popisují vztahy mezi polem napětí a polem objemových sil
$
{\boldsymbol\partial \boldsymbol \sigma + \boldsymbol{ \bar{X}} = \boldsymbol 0},
$

$$
\displaystyle{
 \left[
\begin{array}{cccccc}
\frac{\partial}{\partial x} & 0 & 0 & 0 & \frac{\partial}{\partial z} &  \frac{\partial}{\partial y} \\
0 & \frac{\partial}{\partial y} & 0 & \frac{\partial}{\partial z} & 0 & \frac{\partial}{\partial x}\\
0 & 0 & \frac{\partial}{\partial z} &  \frac{\partial}{\partial y} &  \frac{\partial}{\partial x} & 0
\end{array}
\right]
\left\{
\begin{array}{c}
\sigma_x \\
\sigma_y \\
\sigma_z \\

\tau_{xy}
\end{array}
\right\}
 +\left\{
\begin{array}{c}
\bar{X}\\
\bar{Y}\\
\bar{Z}
\end{array}
\right\}

\left\{
\begin{array}{c}

0\\
0
\end{array}
\right\}
}
$$

### Fyzikální rovnice

Fyzikální rovnice popisují vztah mezi polem deformace a polem napětí. O těchto rovnicích často mluvíme jako o konstitutivních vztazích, pro lineárně pružný materiál platí

$
\boldsymbol\sigma = \boldsymbol D \boldsymbol \varepsilon,
$

kde $\boldsymbol D$ je matice tuhosti daného materiálu. Pro lineární izotropní materiál má tato matice tvar
$$
\boldsymbol D =
\frac{E}{(1+\nu)(1-2\nu)}
\left[
\begin{array}{cccccc}
1-\nu & \nu & \nu & 0 & 0 & 0\\
\nu & 1-\nu & \nu & 0 & 0 & 0\\
\nu & \nu & 1-\nu & 0 & 0 & 0\\
 0 & 0 & 0 & \frac{1-2\nu}{2} & 0 & 0\\
 0 & 0 & 0 & 0 & \frac{1-2\nu}{2} & 0\\
 0 & 0 & 0 & 0 & 0 & \frac{1-2\nu}{2}
\end{array}
\right].
$$

Inverzní vztah je zprostředkován maticí poddajnosti materialu $\boldsymbol {C}$, která je inverzí matice tuhosti $\boldsymbol{D}$

$
\boldsymbol\varepsilon = \boldsymbol C \boldsymbol\sigma,
$

$$\displaystyle{
\boldsymbol{C} = \boldsymbol{D}^{-1}=
\left[
\begin{array}{cccccc}
\frac{1}{E} & \frac{-\nu}{E}& \frac{-\nu}{E} & 0 & 0 & 0\\
\frac{-\nu}{E} & \frac{1}{E}& \frac{-\nu}{E} & 0 & 0 & 0\\
\frac{-\nu}{E} & \frac{-\nu}{E}& \frac{1}{E} & 0 & 0 & 0\\
 0 & 0 & 0 & \frac{1}{G} & 0 & 0\\
 0 & 0 & 0 & 0 & \frac{1}{G} & 0\\
 0 & 0 & 0 & 0 & 0 & \frac{1}{G}
\end{array}
\right].
}$$

### Okrajové podmínky

Nedílnou součástí formulace problému jsou okrajové podmínky. Hranici $\Gamma$ rozdělíme na část $\Gamma_u$, kde jsou předepsané kinematické okrajové podmínky, a na část  $\Gamma_p$ s předepsanými statickými (silovými) okrajovými podmínkami.

#### Kinematické okrajové podmínky

Na hranici $\Gamma_u$ s předepsanými posuny $\boldsymbol{\bar{u}}$ platí

$
\boldsymbol u - \boldsymbol{\bar{u}} = \boldsymbol 0.
$

#### Statické okrajové podmínky

Na hranici $\Gamma_p$ s předepsanými povrchovými silami $\boldsymbol{\bar{p}}$ platí

$
\boldsymbol n \boldsymbol \sigma - \boldsymbol{\bar{p}} =\boldsymbol 0,
$

kde $\boldsymbol n$ je matice směrových kosinů $n_x, n_y, n_z$ vnější normály k povrchu
$$
\left[
\begin{array}{cccccc}
n_x & 0 & 0 & 0 & n_z & n_y\\
0 & n_y & 0 & n_z & 0 & n_x\\
0 & 0 & n_z & n_y & n_x & 0 
\end{array}
\right]
\left\{
\begin{array}{c}
\sigma_x \\
\sigma_y \\
\sigma_z \\
\tau_{yz} \\
\tau_{zx} \\
\tau_{xy}
\end{array}
\right\}
-
\left\{
\begin{array}{c}
\bar{p}_x\\
\bar{p}_y\\
\bar{p}_z
\end{array}
\right\}
=
\left\{
\begin{array}{c}
0\\
0\\
0
\end{array}
\right\}
$$

### Diferenciální rovnice problému

Budeme se zabývat deformační variantou MKP, primární neznámé tedy budou posuny $\boldsymbol u$, deformace a napětí dopočítáme následně. Hledáme $\boldsymbol u$ splňující kinematické i statické okrajové podmínky takové, aby splňovalo podmínky rovnováhy

$
{\boldsymbol\partial \boldsymbol \sigma + \boldsymbol{ \bar{X}} = \boldsymbol 0},
$

dosazením z konstitutivních vztahů získáme 

$
{\boldsymbol\partial (\boldsymbol D \boldsymbol \varepsilon) + \boldsymbol{ \bar{X}} = \boldsymbol 0}
$

a následně aplikací geometrických rovnic plyne

$
{\boldsymbol\partial (\boldsymbol D \boldsymbol\partial^T \boldsymbol u) + \boldsymbol{ \bar{X}} = \boldsymbol 0}
$

Řešení $\boldsymbol u$ splňující tyto rovnice nazýváme silným řešením.

### Metoda vážených reziduí

<i>
Připomeneme si integraci per partes 

$\displaystyle{
\begin{array}{rcl}
\displaystyle{\int_0^L
\frac{\ df}{\ dx}(x)
g(x)
\ dx}
& = &\displaystyle{
\left[
f(x) g(x)
\right]_0^L - \int_0^L
f(x)
\frac{\ dg}{\ dx}(x)
\ dx },
\\\displaystyle{
\int_\Omega
\frac{\ df}{\ dx}(x)
g(x)
\ dx}
& = &\displaystyle{
\int_\Gamma
f(x) g(x) n(x)
\ dx - \int_\Omega
f(x)
\frac{\ dg}{\ dx}(x)
\ dx}.
\end{array}
}$
</i>


 Definujme reziduum řešení pro danou funkci $v(x)$

$\displaystyle{
\begin{array}{rcl}
x \in \Omega & : & 
r\left(v(x)\right) = 
\frac{d}{\ dx}
\bigl(
E(x) A(x) \frac{\ dv}{\ dx}(x) 
\bigr)
+
b(x) 
\\
x \in \Gamma^u & : & 
r\left(v(x)\right) = \overline{u}(x) - v(x) 
\\
x \in \Gamma^t & : & 
r\left(v(x)\right) = \overline{t}(x) - E(x)A(x) \frac{\ dv}{\ dx}(x) n(x) 
\end{array}
}$

Pokud je $v(x)$ řešením problému ($v(x) \equiv u(x)$) máme $r(x) \equiv 0$.

Idea <b>metody vážených reziduí</b>: Zvolme libovolnou funkci $\delta u(x)$ (tzv. váhová funkce (test, weight function)) a spočtěme

$$
\int_\Omega \delta u(x) r(v(x)) \ dx
+
\int_\Gamma \delta u(x) r(v(x)) \ dx.
$$

Jestliže hodnota integrálu bude rovna nule pro <b> všechny</b> váhové funkce $\delta
u(x)$, potom $v(x)$ je řešením problému.


#### Slabé řešení

Hledáme řešení $u(x)$ rovnice takové, které splňuje rovnici

$\displaystyle{
0 = \int_{\Omega} \delta u \left(  \frac{d}{dx}\left(EA\frac{d u}{dx}\right)+f_x \right) \ dx
}$

pro všechna $\delta u $ splňující kinematické okrajové podmínky ($\delta u = 0$). Použitím integrace per-partes získáme


$\displaystyle{
0 = \int_{\Gamma}\delta u \left( EA\frac{d u}{dx}\right)n \ dx-\int_{\Omega} \frac{d(\delta u)}{dx} \left( EA\frac{d u}{dx}\right) \ dx + \int_{\Omega}  \delta u \bar{f_x} \ dx,
}$

kde $n$ je vnější normála k hranici $\Gamma$. Hraniční integrály můžeme rozdělit na část hranice $\Gamma_u$, kde jsou předepsány kinematické okrajové podmínky, a na část $\Gamma_p$, kde jsou aplikovány silové okrajové podmínky

$\displaystyle{
0 =  \int_{\Gamma_u}\underbrace{\delta u}_{= 0} \left( EA\frac{d u}{dx}\right)n \ dx +
\int_{\Gamma_p}\delta u \underbrace{\left( EA\frac{d u}{dx}\right)n}_{= \bar{N}} \ dx
-\int_{\Omega} \frac{d(\delta u)}{dx} \left( EA\frac{d u}{dx}\right) \ dx + \int_{\Omega}  \delta u \bar{f_x} \ dx,
}$

Úpravou výrazů dostaneme finální tvar rovnic

$\displaystyle{
\int_{\Omega} \frac{d(\delta u)}{dx} EA \frac{d u}{dx} \ dx = \left[\delta u \bar{N}\right]_{\Gamma_p} + \int_{\Omega}  \delta u \bar{f_x} \ dx.
}
$

### Analogie metody vážených reziduí a Principu minima potenciální energie

Ze všech kinematicky přípustných stavů pružného tělesa nastává ten, který dává potenciální energii systému minimální hodnotu

$\displaystyle{
\Pi = E_i+E_e = min,
}$

kde

$\displaystyle{\begin{array}{rcl}
\displaystyle{E_i}&=&\displaystyle{\frac{1}{2}\int_\Omega\sigma\varepsilon\,d\Omega,}\\
\displaystyle{E_e}&=&\displaystyle{-\int_\Omega ub\,d\Omega-\int_{\Gamma^t}u\bar{t}\,d\Gamma}.
\end{array}
}$

$\Pi$ je funkcionál (funkce funkcí). Jak se bude měnit hodnota $\Pi$, když se bude měnit $u(x)$? Malou změnu funkce budeme nazývat její variací a vyjádříme ji jako $\delta u(x)=\xi w(x)$, kde $w(x)$ je libovolná funkce respektující kinematické okrajové podmínky a $\xi$ malé nezáporné číslo. 

Odpovídající změna funkcionálu, která se nazývá jeho variací je definována jako

$\displaystyle{
\delta\Pi = \Pi(u(x)+\xi w(x))-\Pi(u(x))\equiv\Pi(u(x)+\delta u(x))-\Pi(u(x)).
}$


My hledáme minimum, tedy variace musí být nulová a tedy $\delta\Pi=0$.

<img src ="Figures/fig6c.png" width=500/>


Pro náš případ

$\displaystyle{
\begin{array}{rcl}
\delta\Pi&=&\displaystyle{\frac{1}{2}\int AE \left(\frac{du}{dx}+\xi\frac{dw}{dx}\right)^2\,dx-\frac{1}{2}\int AE \left(\frac{du}{dx}\right)^2\,dx}\\
&&\displaystyle{-\int (u+\xi w)b\,dx-\int (u+\xi w)\bar{t}\,d\Gamma+\int (u)b\,dx+\int (u)\bar{t}\,d\Gamma}\\
&=&\displaystyle{\frac{1}{2}\int AE\left(\left(\frac{du}{dx}\right)^2+2\xi\frac{du}{dx}\frac{dw}{dx}+\xi^2\left(\frac{dw}{dx}\right)^2\right)\,dx}\\
&&\displaystyle{ -\frac{1}{2}\int AE\left(\frac{du}{dx}\right)^2\,dx-\xi\int wb\,dx-\xi\left.(w\bar{t})\right\vert_\Gamma}
\end{array}
}$

$\displaystyle{
\begin{array}{rcl}
\displaystyle{\delta\Pi}&=&\displaystyle{\xi\int AE \frac{dw}{dx}\frac{du}{dx}\,dx-\xi\int wb\,dx-\xi(w\bar{t})\vert_\Gamma}
\end{array}
}$

Hledáme minimum, tedy variace musí být nulová ($\delta\Pi=0$). Dosazením z předchozích výrazů a podělením $\xi$ máme

$\begin{array}{rcl}
\displaystyle{\delta\Pi\xi} &=& \displaystyle{\int_\Omega AE \frac{dw}{dx}\frac{du}{dx}\,dx-\int_\Omega wb\,dx-(w\bar{t})\vert_\Gamma = 0}\\
\displaystyle{\delta\Pi} &=&\displaystyle{ \int_\Omega AE \frac{d\delta u}{dx}\frac{du}{dx}\,dx-\int_\Omega \delta ub\,dx-(\delta u\bar{t})\vert_\Gamma = 0}\\
\end{array}
$

To lze dále zjednodušit. Pokud označíme $\delta\varepsilon=\frac{dw}{dx}$ ($\delta\varepsilon$ budeme nazývat virtuální deformací) a dosazením konstitutivních vztahů obdržíme

$\displaystyle{
\delta\Pi = \int_\Omega A \sigma\delta\varepsilon\,dx-\int_\Omega \delta ub\,dx-(\delta u\bar{t})\vert_\Gamma = 0
}$

A to je známý <b>"Princip virtuálních posunutí"</b>.

Všimněte si ekvivalence odvozených výrazů a výrazů získaných aplikací metody vážených reziduí.

### Diskretizovaná úloha

Konstrukci rozdělíme na $n$ dílků (prvků). Neznámé posuny $u$ budeme na každém prvku aproximovat jako součin zvolených interpolačních funkcí $N_i$ a příslušných (neznamých) vah $u_i$ (často nazývaných uzlové hodnoty). Stejně budeme aproximovat i průběh váhové funkce $\delta u$

$\displaystyle{
u \approx \sum N^e_i u_i = \boldsymbol{N}^e\boldsymbol {r}^e, 
\quad 
\delta u \approx \sum N^e_i \delta u_i = \boldsymbol{N}^e\boldsymbol {w}^e, 
}$

kde $\boldsymbol{N}^e$ je matice interpolačních funkcí $N_i$ prvku, $\boldsymbol {r}^e$ je vektor uzlových hodnot posunů na prvku $e$ a $\boldsymbol {w}^e$ je vektor uzlových hodnot váhových funkcí $\delta u$.
Pro derivace pak platí

$\displaystyle{
\frac{du}{dx} = \frac{d}{dx}(\sum N^e_i u_i) = \sum \frac{dN^e_i}{dx} u_i = \boldsymbol{B}^e\boldsymbol {r}^e, 
\quad
\frac{d(\delta u)}{dx} = \boldsymbol{B}^e\boldsymbol {w}^e.
}$

kde matice $\boldsymbol{B}^e$ obsahuje derivace interpolačních funkcí.

Integrál slabého řešení na celé oblasti rozdělíme na sumu integrálů přes jednotlivé prvky a po úpravě získáme 

$\displaystyle{
\sum_{e=1}^n\left\{
\int_{\Omega} \boldsymbol{B}^e\boldsymbol{w}^e EA  \boldsymbol{B}^e\boldsymbol {r}^e \ dx = \left[\boldsymbol{N}^e\boldsymbol {w}^e\bar{N}\right]_{\Gamma_p} + \int_{\Omega} \boldsymbol{N}^e\boldsymbol {w}^e \bar{f_x} \ dx \right\},\\
\sum_{e=1}^n\left\{
\boldsymbol{w}^e \left(\int_{\Omega} \boldsymbol{B}^e EA  \boldsymbol{B}^e\ dx\ \boldsymbol {r}^e \right) =\boldsymbol{w}^e \left( \left[\boldsymbol{N}^e \bar{N}\right]_{\Gamma_p} + \int_{\Omega} \boldsymbol{N}^e \bar{f_x} \ dx\right) \right\},\\
\sum_{e=1}^n\left\{
\underbrace{\int_{\Omega} \boldsymbol{B}^e EA  \boldsymbol{B}^e \ dx}_{\boldsymbol{K^e}}\ \boldsymbol{r}^e  =\underbrace{ \left[\boldsymbol{N}^e \bar{N}\right]_{\Gamma_p}}_{\boldsymbol{f}_{\Gamma}^e} + \underbrace{ \int_{\Omega} \boldsymbol{N}^e \bar{f_x} \ dx}_{\boldsymbol{f}_{\Omega}^e}\right\}.
}
$

Rozšířením lokálních vektorů $\boldsymbol{u^e}$ do globálního vektoru $\boldsymbol{u}$ uzlových hodnot můžeme psát

$
\boldsymbol{K^gr}=\boldsymbol{f^g}.
$

### Silné řešení

Hledáme $u(x)$, které splňuje

$\displaystyle{EA\left(\frac{d^2 u(x)}{\ dx^2}\right) + 1,\ \ \ x\in(0,2),\ \ u(0)=0,\ N(2)=AE\frac{\ du(2)}{\ dx}=0}$

Integrací postupně dostaneme

$\displaystyle{EA\frac{\ du(x)}{\ dx} + 1x + C_1 = 0,\ \ \ \ EA u(x) + 1\frac{x^2}{2} + C_1 x + C_2= 0,}$


integrační konstanty $C_1$ a $C_2$ určíme z okrajových podmínek

$\displaystyle{ u(0)=0: C_2 = 0,\ \ \ \ N(2)=0: AE\frac{\ du(2)}{\ dx}=-C_1 - 2=0: C_1=-2}.$

Konečně tedy dostáváme

$\displaystyle{ u(x) = \frac{1}{EA} (2x-x^2/2),\ \ \ N(x) =  AE\frac{\ du(x)}{\ dx} = 2-x}.$

### Slabé řešení - lineární aproximace

Hledáme $u(x)$ (dostatečně integrovatelné), $u(0) = 0$ takové, aby:

$\displaystyle{
\int_0^2
EA\frac{\ d\delta u}{\ dx}(x) \frac{\ du}{\ dx}(x) \ dx
=\int_0^2
\delta u(x) 1
\ dx
}$

pro libolné $\delta u(x)$ (dostatečně integrovatelné), kde $\delta u(0) = 0$. 

Hledané řešení a váhovou funkci vyjádřeme jako:

$\displaystyle{u(x) = ax+b,\ \ \delta u(x)= cx + d}$

   * Z podmínky $u(0)=0$ plyne $b=0$,
   * Z podmínky $\delta u(0) = 0$ pak $d=0$,


Zvolenou aproximaci řešení a váhové funkce dosadíme do slabého řešení

$\displaystyle{\int_0^2ac\ dx-\int_0^2cx\ dx = 0},$

provedením integrace

$\displaystyle{2ac-2c=0}.$

Všimněte si, že každý člen obsahuje konstantu $c$ která je libovolná (vyjadřuje variabilitu volby váhové funkce)

$\displaystyle{c(2a-2)=0}.$

Protože $c\ne 0$, pak výraz v závorce musí být roven $0$

$\displaystyle{2a-2=0\ \Rightarrow\ a=1}.$

Konečně tedy dostáváme přibližné řešení

$\displaystyle{ u(x)=x,\ \ \ N(x)=EA\frac{\ du}{\ dx}= 1}.$

### Prvek s lineární aproximací

#### Lineární aproximace

Lineární aproximační funkce mají pro $x\in(0, l^e)$ tvar

$\displaystyle{
N_1 = 1-\frac{x}{l^e},\\
N_2 = \frac{x}{l^e}
}$

nebo obecně na intervalu $x\in(x_1, x_2)$
<img src="Figures/basis.png" width="270" align="right">

$\displaystyle{
N_1 = \frac{x_2 - x}{x_2-x_1},\\
N_2 = \frac{x - x_1}{x_2-x_1}.
}$

Libovolnou funkci $f(x)$ můžeme potom aproximovat jako lineární kombinaci bázových funkcí $N_i$ a uzlových hodnot $f_i$

$
f(x) = N_1 f_1 + N_2 f_2.
$

Tato aproximace popisuje danou funkci $f(x)$ přesně, pokud je funkce nejvýše prvního stupně (lineární). Analogicky, pomocí derivací bázových funkcí, můžeme aproximovat derivaci funkce. Derivace bázových funkcí pro lineární aproximaci mají tvar

$\displaystyle{
B_1=-\frac{1}{l^e},\\ 
B_2=\frac{1}{l^e}.
}$



#### Diskretizovaná úloha

Odvodili jsme si příspěvek jednolivých prvků do celkové soustavy rovnic

$
\boldsymbol{K^er} = \boldsymbol{f}_{\Gamma}^e +  \boldsymbol{f}_{\Gamma}^e ,
$

kde matice tuhosti $\boldsymbol{K^e}$  a vektory zatížení $\boldsymbol{f}_{\Gamma}^e, \boldsymbol{f}_{\Omega}^e $ mají tvar

$\displaystyle{
\boldsymbol{K^e} = \int_{\Omega} \boldsymbol{B}^e EA  \boldsymbol{B}^e \ dx, \\
\boldsymbol{f}_{\Gamma}^e =\left[\boldsymbol{N}^e \bar{N}\right]_{\Gamma_p},\\
\boldsymbol{f}_{\Omega}^e =\int_{\Omega} \boldsymbol{N}^e \bar{f_x} \ dx.
}$

Při použití lineární aproximace dostaneme

$\displaystyle{
\boldsymbol{K^e} = \int_{\Omega} \left\{ \begin{array}{c}
-\frac{1}{l^e}\\
\frac{1}{l^e}
\end{array}\right\}
 EA  \left\{ -\frac{1}{l^e}\quad \frac{1}{l^e} \right\}  \ dx  = \frac{EA}{l^e}\left[\begin{array}{cc}
1&-1\\
-1&1
\end{array}\right]
\\
\boldsymbol{f}_{\Gamma}^e =\left[\boldsymbol{N}^e \bar{N}\right]_{\Gamma_p},\\
\boldsymbol{f}_{\Omega}^e =\int_{\Omega} \boldsymbol{N}^e \bar{f_x} \ dx.
}$

#### Transformace do 2D

Vztah mezi koncovými silami a posuny v lokální soustavě prutu jsme si odvodili minule.

$
\left\{\begin{array}{c}F_{1x}^{le}\\F_{2x}^{le}\end{array}\right\}=
\left[\begin{array}{cc}k^e& -k^e\\-k^e& k^e\end{array}\right]
\left\{\begin{array}{c}u_{1x}^{le}\\u_{2x}^{le}\end{array}\right\}
$


Nyní tento vztah rozšíříme do 2D. Nejprve rozšı́řı́me vztah pro koncové sı́ly a posuny prutu přidánı́m
lokálních koncových sil kolmých k ose prutu, které budou nulové, nebot jde o tažený-tlačený prut:

$F^l_{1y}= F^l_{2y}= 0:$

$
\left\{\begin{array}{c}F_{1x}^{le}\\F_{1y}^{le}\\F_{2x}^{le}\\F_{2y}^{le}\end{array}\right\}=
\left[\begin{array}{cccc}k^e& 0& -k^e& 0\\0&0&0&0\\-k^e& 0& k^e& 0\\0&0&0&0\end{array}\right]
\left\{\begin{array}{c}u_{1x}^{le}\\u_{1y}^{le}\\u_{2x}^{le}\\u_{2y}^{le}\end{array}\right\}
$


Pro transformaci složek vektoru mezi lokálnı́m a globálnı́m souřadným
systémem platı́:

$
\begin{array}{l}
u_{1x}^{le}&=&u_{1x}^{ge} \cos(\phi^e) + u_{1y}^{ge} \sin(\phi^e)\\
u_{1y}^{le}&=&-u_{1x}^{ge}\sin(\phi^e)+u_{1y}^{ge}\cos(\phi^e)
\end{array}
$



Maticově to pro vektory koncových sil či posunů můžeme zapsat jako

$
{r}^{le} = {T}^e{r}^{ge};\;{F}^{le} = {T}^e {F}^{ge}
$

$
{r}^{ge}=\{u_{1x}^{ge},u_{1y}^{ge},u_{2x}^{ge},u_{2y}^{ge}\}^T;\;\;
{T}^e=\left[\begin{array}{cccc}
\cos(\phi)& \sin(\phi)& 0& 0\\
-\sin(\phi)& \cos(\phi)& 0& 0\\
0&0&\cos(\phi)& \sin(\phi)\\
0&0&-\sin(\phi)& \cos(\phi)
\end{array}\right]
$

Pozn.: Transformačnı́ matice T je ortogonálnı́; tedy inverznı́ matice
je matice transponovaná

$
({T}^e)^T{T}^e = {T}^e({T}^e)^T = {I};\;\;\;({T}^e)^T = ({T}^e)^{-1}
$

$
\begin{array}{l}
{F}^{ge} &=& ({T}^e)^T{F}^{le}\\
&=&({T}^e)^T{K}^{le}{r}^{le}\\
&=&\underbrace{({T}^e)^T{K}^{le}{T}^e}_{{K}^{ge}}{r}^{ge}
\end{array}
$

Nynı́ můžeme již snadno odvodit vztah mezi globálnı́mi koncovými
silami a posuny

$
{K}^{ge}={{EA}\over{l}}\left[\begin{array}{cccc}
c^2&cs&-c^2&-cs\\
cs&s^2&-cs& -s^2\\
-c^2&-cs&c^2&cs\\
-cs&-s^2&cs&s^2
\end{array}\right];\;\;\begin{array}{c}c=\cos(\phi^e)\\s=\sin(\phi^e)\end{array}
$

## Grafický výstup

Matlab umožňuje snadnou tvorbu grafů. Pro vykreslení nějaké funkce je vhodná funkce <code>plot</code>. Do jednoho grafu můžeme nakreslit libovolný počet funkcí, musíme však použít příkaz <code>hold on</code>, který zajistí uchování i předchozí funkce. Volitelnými argumenty funkce <code>plot</code> je barva, vzory bodů a čar (naleznete v nápovědě).

Pro tisk bodů lze použít funkci <code>scatter</code>. 

Grafický výstup je v Matlabu kreslen do zvláštních oken (figures), nové okno otevřeme příkazem <code>figure</code>, pokud chceme všechna otevřená okna pozavírat použijeme příkaz <code>close all</code>. Okno s grafem lze rozčlenit na podokna pomocí funkce <code>subplot</code>.

Rozsah hodnot na osách lze nastavit příkazem <code>axis</code>, osy popíšeme pomocí příkazů <code>xlabel, ylabel</code>. Graf lze nadepsat pomocí příkazu <code>title</code>.

```octave

subplot(1, 2, 1)
x = [0:0.05:1] * pi;
y = cos(x);
plot(x,y);

hold on;
y = sin(x);
plot(x,y, '--r*');

subplot(1, 2, 2)
scatter(x,y, 'k'); 
axis([0 pi -1 1]); 
xlabel('x'); ylabel('sin(x)'); 
title('Example 2')
```

Existují další funkce, které jsou vhodné pro použití ve 2D (<code>patch</code>), případně ve 3D (<code>mesh, surf</code>). Jejich použití ukážeme, pokud na ně narazíme v dalších cvičeních v průběhu semestru.

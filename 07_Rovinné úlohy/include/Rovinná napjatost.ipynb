# Rovinná napjatost

Rovnice pro popis úlohy rovinné napjatosti plynou ze základních vztahů pro 3D pružnost, které jsou zredukovány na základě předpokladů pro rovinnou napjatost. Pro úlohu v rovině $xy$ předpokládáme

 * zatížení působí pouze v rovině $xy$,
 * ve směru $z$ se prut může volně rozpínat,
 * jediné nenulové složky napětí jsou $\sigma_x, \sigma_y, \tau_{xy}$.
 * průřez je homogenní z izotropního materiálu (napětí je tedy konstantní po průřezu)
 <!-- * prut je zatížen objemovým zatížením $X$ jenž je konstantní po průřezu, jeho výslednici na průřezu označíme $f_x(x) = A(x)X(x)$/!-->

Dosazením třetího předpokladu do konstitutivních vztahů vyjádříme složky vektoru deformace

$
\displaystyle{
\left\{
\begin{array}{c}
\varepsilon_x \\
\varepsilon_y \\
\varepsilon_z \\
\gamma_{yz} \\
\gamma_{zx} \\
\gamma_{xy}
\end{array}
\right\}
=
\left[
\begin{array}{cccccc}
\frac{1}{E} & \frac{-\nu}{E}& \frac{-\nu}{E} & 0 & 0 & 0\\
\frac{-\nu}{E} & \frac{1}{E}& \frac{-\nu}{E} & 0 & 0 & 0\\
\frac{-\nu}{E} & \frac{-\nu}{E}& \frac{1}{E} & 0 & 0 & 0\\
 0 & 0 & 0 & \frac{1}{G} & 0 & 0\\
 0 & 0 & 0 & 0 & \frac{1}{G} & 0\\
 0 & 0 & 0 & 0 & 0 & \frac{1}{G}
\end{array}
\right]
}
\left\{
\begin{array}{c}
\sigma_x \\
\sigma_y \\
0 \\
0 \\
0 \\
\tau_{xy}
\end{array}
\right\}.
$

Získáme čtyři nenulové složky deformace, složku  $\varepsilon_z$ však můžeme vyjádřit pomocí $\varepsilon_x$ a  $\varepsilon_y$

$
\displaystyle{
\varepsilon_x = \frac{1}{E}\sigma_x -\frac{\nu}{E}\sigma_y,\\
\varepsilon_y = -\frac{\nu}{E}\sigma_x + \frac{1}{E}\sigma_y,\\
\varepsilon_z = -\frac{\nu}{E}\sigma_x-\frac{\nu}{E}\sigma_y = \ldots = -\frac{\nu}{1-\nu}(\varepsilon_x+\varepsilon_y)\\
\gamma_{xy} = \frac{1}{G}\tau_{xy}.
}$

Pro vyjádření deformací $\varepsilon_x,\varepsilon_y$ a $\gamma_{xy}$ použijeme geometrické rovnice 

<img src="Figures/defs3.png" width = 750>

$\displaystyle{
\varepsilon_x = \frac{du(x)}{dx},\\
\varepsilon_y = \frac{dv(x)}{dy},\\
\gamma_{xy} = \frac{du(x)}{dy}+\frac{dv(x)}{dx},
}$

posuny $u(x)$ a $v(x)$ budou tedy naše primární neznámé. 


Pro řešení úlohy máme opět k dispozici tři sady rovnic - geometrické, statické a fyzikální.
   

### Geometrické rovnice

$\displaystyle{
\begin{array}{rcl}
\left\{
  \begin{array}{c}
    \varepsilon_{x}( \boldsymbol{x} ) \\
    \varepsilon_{y}( \boldsymbol{x} ) \\
    \gamma_{xy}( \boldsymbol{x} ) 
  \end{array}
\right\}
& = &
\left[
\begin{array}{cc}
\displaystyle{\frac{\partial}{\partial x} }&         0 \\
        0 & \displaystyle{\frac{\partial}{\partial y}} \\
\displaystyle{\frac{\partial}{\partial y}} & \displaystyle{\frac{\partial}{\partial x} }
\end{array}
\right]
\left\{
\begin{array}{c}
u( \boldsymbol{x} ) \\
v( \boldsymbol{x} ) 
\end{array}
\right\} \\
\boldsymbol{\varepsilon}( \boldsymbol{x} ) & = & \boldsymbol{\partial}^T \boldsymbol{u}( \boldsymbol{x} )
\end{array}
}$


Složka deformace $\varepsilon_z \not= 0$ dopočítává z konstitutivních rovnic.

### Podmínky rovnováhy - statické rovnice

 * Vektor (nezávislých složek) napětí
$\displaystyle{\boldsymbol{\sigma}(\boldsymbol{x}) = \{
\sigma_{x}(\boldsymbol{x}), \sigma_{y}(\boldsymbol{x}), \tau_{xy}(\boldsymbol{x})
\}^T}$ 

Statické rovnice: $\boldsymbol{x} \in \Omega$

$\displaystyle{
\begin{array}{rcl}
 \left[
    \begin{array}{ccc}
   \displaystyle{   \frac{\partial}{\partial x}  }  &
      0            &\displaystyle{
      \frac{\partial}{\partial y}}
    \\
      0            &\displaystyle{
      \frac{\partial}{\partial y} }   &\displaystyle{
      \frac{\partial}{\partial x}}
    \end{array}
  \right]
\left\{
\begin{array}{c}
\sigma_{x}(\boldsymbol{x}) \\
\sigma_{y}(\boldsymbol{x}) \\
\tau_{xy}(\boldsymbol{x})
\end{array}
\right\}
+
\left\{
\begin{array}{c}
\overline{X}(\boldsymbol{x}) \\
\overline{Y}(\boldsymbol{x}) 
\end{array}
\right\}
& = &
\left\{
\begin{array}{c}
0 \\
0 
\end{array}
\right\} \\
\boldsymbol{\partial} \, \boldsymbol{\sigma}( \boldsymbol{x} ) + \boldsymbol{\overline{X}} & = &
\boldsymbol{0}  
\end{array}
}$


Pro rovinnou napjatost je složka napětí $\sigma_{z} = 0$.

### Konstitutivní rovnice - lineární izotropní materiál

$\displaystyle{
\begin{array}{rcl}
\left\{
 \begin{array}{c}
\sigma_{x} \\
\sigma_{y} \\
\tau_{xy}
\end{array}
\right\}
\kern -1mm%
&=& 
\kern -1mm
\displaystyle{\frac{E}{1-\nu^2}}
\left[
  \begin{array}{ccc}
1 &\nu & 0\\[3mm]
\nu & 1 & 0\\[3mm]
0 & 0 &\displaystyle{ \frac{1-\nu}{2}}
\end{array}\right]
\left\{\begin{array}{l}
\varepsilon_{x}\\
\varepsilon_{y}\\
\gamma_{xy}
\end{array}\right\}\\
\varepsilon_z \kern -1mm&=&\kern -1mm \displaystyle{-\frac{\nu}{1-\nu}(\varepsilon_x+\varepsilon_y)}
\end{array}
}$


### Okrajové podmínky


Kinematické okrajové podmínky: $\boldsymbol{x} \in \Gamma_u$: $\boldsymbol{u}(
\boldsymbol{x} ) - \boldsymbol{\overline{u}}( \boldsymbol{x} ) = \boldsymbol{0}$

Statické okrajové podmínky: $\boldsymbol{x} \in \Gamma_p$

$\displaystyle{
\begin{array}{rcl}
\left[
  \begin{array}{cccccc}
    n_{x}(\boldsymbol{x}) &
    0     &
    n_{y}(\boldsymbol{x}) 
    \\ 
    0     &
    n_{y}(\boldsymbol{x}) &
    n_{x}(\boldsymbol{x}) 
  \end{array}
\right]
\left\{
  \begin{array}{c}
    \sigma_{x}( \boldsymbol{x} ) \\
    \sigma_{y}( \boldsymbol{x} ) \\
    \tau_{xy}( \boldsymbol{x} ) 
  \end{array}
\right\}
-
\left\{
  \begin{array}{c}
    \overline{p}_x( \boldsymbol{x} ) \\
    \overline{p}_y( \boldsymbol{x} ) 
  \end{array}
\right\}
& = &
\left\{
  \begin{array}{c}
    0 \\
    0 
  \end{array}
\right\} \\
\boldsymbol{n}(\boldsymbol{x}) \boldsymbol{\sigma}( \boldsymbol{x} ) -
  \boldsymbol{\overline{p}}(\boldsymbol{x}) & = & \boldsymbol{0}
\end{array}
}$


### Slabé řešení

Divergenční (Clapeyronův) teorém: 
$\displaystyle{
\int_\Omega \boldsymbol{\sigma}^T\boldsymbol{\partial}^T\boldsymbol{u}d\Omega=\int_\Gamma\boldsymbol{u}^T\boldsymbol{n}\boldsymbol{\sigma}d\Gamma-\int_\Omega\boldsymbol{u}^T\boldsymbol{\partial\sigma}d\Omega
}$


Aplikací Clapeyronova teorému na úlohu rovinné napjatosti 

$\displaystyle{
\begin{array}{rcl}
0 & = &\displaystyle{ \int_\Omega \boldsymbol{\delta u}( \boldsymbol{x} )^T \left(
\boldsymbol{\partial} \, \boldsymbol{\sigma}( \boldsymbol{x} ) + \boldsymbol{\overline{X}}
\right)\ d\Omega }\\
& \stackrel{\mbox{Clapeyron}}{=} & \displaystyle{
\int_{\Gamma_u} \overbrace{\boldsymbol{\delta u}(\boldsymbol{x})^T}^{\boldsymbol{0}}
\boldsymbol{n}( \boldsymbol{x}) \boldsymbol{\sigma}( \boldsymbol{x})\ d\Gamma_u
+
\int_{\Gamma_p} \boldsymbol{\delta u}(\boldsymbol{x})^T \overbrace{\boldsymbol{n}( \boldsymbol{x})
\boldsymbol{\sigma}( \boldsymbol{x})}^{\boldsymbol{\overline{p}}}\ d\Omega 
 -  \int_\Omega \left( \boldsymbol{\partial}^T \boldsymbol{\delta u}( \boldsymbol{x})
\right) ^T \boldsymbol{\sigma}( \boldsymbol{x} )\ d\Omega 
+
\int_\Omega \boldsymbol{\delta u}( \boldsymbol{x} )^T \boldsymbol{\overline{X}}( \boldsymbol{x}
)\ d\Omega }
\end{array}
}$


Přisoudíme-li váhové funkci $\boldsymbol{\delta u}( \boldsymbol{x})$ fyzikální
smysl <i> virtuálního posunu</i>, můžeme člen $\boldsymbol{\partial}^T
\boldsymbol{\delta u}( \boldsymbol{x})$ identifikovat jako virtuální deformaci $\boldsymbol{\delta
\varepsilon}(\boldsymbol{x})$.

$\displaystyle{
\begin{array}{rcl}
\displaystyle{\int_\Omega \boldsymbol{\delta \varepsilon}(\boldsymbol{x})^T \boldsymbol{\sigma}( \boldsymbol{x} )
\ d\Omega} & = & \displaystyle{\int_{\Gamma_p} \boldsymbol{\delta u}( \boldsymbol{x} ) ^T
\boldsymbol{\overline{p}}(\boldsymbol{x})\ d\Gamma_p + \int_\Omega \boldsymbol{\delta u}(
\boldsymbol{x} )^T \boldsymbol{\overline{X}}( \boldsymbol{x} ) \ d\Omega} \\
\displaystyle{\delta W_{\rm int} }& = & \displaystyle{\delta W_{\rm ext}}.
\end{array}
}$


Metodu vážených reziduí lze tedy chápat jako <em> zobecnění principu
virtuálních posunů </em>.

### Galerkinovská aproximace


Provedeme diskretizaci řešené oblasti na jednotlivé prvky a neznámé posuny  budeme aproximovat pomocí bázových funkcí $\boldsymbol{N}^e$ a uzlových hodnot $\boldsymbol{r}^e$

$\displaystyle{
\boldsymbol{u}^e( \boldsymbol{x} ) \approx \boldsymbol{N}^e( \boldsymbol{x} )  \boldsymbol{r}^e,
}$


Aproximace polí deformací a napětí je realizována pomocí matice derivací bázových funkcí (geometrické matice) $\boldsymbol{B}^e$. (Pro jednoduchost dále předpokládejme, že vektor počáteční deformace $\boldsymbol{\varepsilon}_0=\boldsymbol{0}$.)

$\displaystyle{
\begin{array}{rcl}
\boldsymbol{\varepsilon}^e( \boldsymbol{x} ) & \approx & \boldsymbol{B}^e( \boldsymbol{x}) \left( \boldsymbol{r}^e \right), \\
\boldsymbol{\sigma}^e( \boldsymbol{x} ) & \approx & \boldsymbol{D}^e( \boldsymbol{x}) \left(
\boldsymbol{B}^e( \boldsymbol{x}) \left( \boldsymbol{r}^e\right) \right). 
\end{array}
}$


Obdobně pro váhové funkce
$\displaystyle{
\begin{array}{rcl}
\boldsymbol{\delta u}^e( \boldsymbol{x} ) \approx \boldsymbol{N}^e(\boldsymbol{x}) \boldsymbol{\delta r^e}, &&
\boldsymbol{\delta \varepsilon^e}( \boldsymbol{x} ) \approx \boldsymbol{B}^e(\boldsymbol{x}) \boldsymbol{\delta r^e}.
\end{array}
}$


Po dosazení do slabé formulace podmínek rovnováhy dostáváme

$\displaystyle{
\boldsymbol{\delta r}^T \displaystyle{\Bigg\{\sum_{e=1}^{n}\boldsymbol{L}^{eT}\bigg[\overbrace{
\int_\Omega \boldsymbol{B}^e(\boldsymbol{x})^T \boldsymbol{D}^e(\boldsymbol{x}) \boldsymbol{B}^e( \boldsymbol{x}
)\ d\Omega}^{\boldsymbol{K^e}}
\boldsymbol{L}^e\boldsymbol{r} 
\kern -1mm
- 
\kern -1mm
\overbrace{\int_\Omega 
\kern -1mm
\boldsymbol{N}^e(\boldsymbol{x})^T \boldsymbol{\overline{X}} (\boldsymbol{x}) 
\kern -1mm
\ d\Omega}^{\boldsymbol{f_\Omega^e}}  
\;\;-
\underbrace{\int_{\Gamma_p} 
\kern -1mm
\boldsymbol{N}^e(\boldsymbol{x}) 
^T \boldsymbol{\overline{p}}( \boldsymbol{x}) \ d\Gamma_p}_{\boldsymbol{f_\Gamma^e}} \bigg]\Bigg\} = 0.}
}$


Neznámé uzlové posuny tedy splňují rovnici

$\displaystyle{
\boldsymbol{K} \, \boldsymbol{r} = \boldsymbol{f} = \boldsymbol{f_\Omega} + \boldsymbol{f_\Gamma}  .
}$

## Příklad č. 1 - Tahový patch test

```octave
% funkce pocita plochu trojuhelnikoveho prvku
% In: 
%   xe - vektor x-ovych souradnic uzlu prvku (x1,x2,x3)
%   ye - vektor y-ovych souradnic uzlu prvku (y1,y2,y3)
%
% Out:
%   Ae - plocha prvku
%
function Ae = area_triangle (xe,ye) 

Ae=(1/2)*((xe(2)*ye(3)-xe(3)*ye(2))-(xe(1)*ye(3)-xe(3)*ye(1))+(xe(1)*ye(2)-xe(2)*ye(1))) ;

end

% funkce pocita delky stran trojuhelnikoveho prvku
% In: 
%   xe - vektor x-ovych souradnic uzlu prvku (x1,x2,x3)
%   ye - vektor y-ovych souradnic uzlu prvku (y1,y2,y3)
%
% Out:
%   le - delky hran prvku (3,1)
%
function le = length_triangle (xe,ye) 

le = zeros(3,1) ;
le(1) = sqrt((xe(2)-xe(1))*(xe(2)-xe(1))+(ye(2)-ye(1))*(ye(2)-ye(1)));
le(2) = sqrt((xe(3)-xe(2))*(xe(3)-xe(2))+(ye(3)-ye(2))*(ye(3)-ye(2)));
le(3) = sqrt((xe(1)-xe(3))*(xe(1)-xe(3))+(ye(1)-ye(3))*(ye(1)-ye(3)));
end

% funkce pocita matici tuhosti 2d trojuhelnikoveho stenoveho prvku s linearni aproximaci
% In: 
%   xe - vektor x-ovych souradnic uzlu prvku (x1,x2,x3)
%   ye - vektor y-ovych souradnic uzlu prvku (y1,y2,y3)
%   E - Younguv modul pruznosti
%   nu - Poissonuv soucinitel
%
% Out:
%   ke - matice tuhosti prvku (6,6)
%   dbe - matice na vypocet napeti (3,6)
%   de - materialova matice (3,3)
%   be - mative derivaci bazovych funkci (3,6)
%
function [ke,dbe,de,be] = plane_stress (xe,ye,E,nu) 

Ae = area_triangle(xe,ye) ;

y23 = ye(2)-ye(3); y31 = ye(3)-ye(1); y12 = ye(1)-ye(2);
x32 = xe(3)-xe(2); x13 = xe(1)-xe(3); x21 = xe(2)-xe(1);

be = (1/(2*Ae))*[ y23, 0, y31, 0, y12, 0 ;
                  0, x32, 0, x13, 0, x21 ;
                  x32,y23,x13,y31,x21,y12] ;

de = (E/(1-nu*nu))*[ 1, nu, 0;
                    nu, 1, 0;
                    0, 0, (1-nu)/2 ];

dbe = de*be ;

ke = Ae*be'*de*be;

end

% funkce pro lokalizaci matice tuhosti prvku 
% do matice tuhosti konstrukce
% vyzaduje existenci globalni matice kodovych cisel - lm
% 
%
% Vstup:
%   k - matice tuhosti konstrukce
%   ke- matice tuhosti prvku
%   lm- matice kodovych cisel
%   e - cislo prvku 
%
% Vystup:
%   k - matice tuhosti konstrukce
%
function k = assembly (k,ke,lm,e)

ndof = size(ke,1);
for i=1:ndof
        ia = lm(e,i);
        if ia ~=0 
            for j=1:ndof
                ja=lm(e,j);
                if ja ~=0
                  k(ia,ja)=k(ia,ja)+ke(i,j);
                end   
            end
        end
    end
end

% funkce pro lokalizaci matice prave strany 
% vyzaduje existenci globalni matice kodovych cisel - lm
% 
%
% Vstup:
%   f - matice prave strany
%   fe- matice prave strany prvku
%   lm- matice kodovych cisel
%   e - cislo prvku 
%
% Vystup:
%   f - matice prave strany
%
function f = assemblyf (f,fe,lm,e)

ndof = size(fe,1);
for i=1:ndof
        ia = lm(e,i);
        if ia ~=0 
           f(ia,1)=f(ia,1)+fe(i,1);
        end
    end
end
```

```octave
%pocet prvku
nelem=4;
nnodes=5;
E=10; 
nu = 0.1;

% pole kodovych cisel uzlu
id = [ 5 6; 1 7; 2 8; 9 10; 3 4 ];

%pole uzlovych cisel prvku
ide = [1 2 5; 2 3 5; 3 4 5; 4 1 5];

% pocet neznamych
neq = 10;

% pole souradnic
x =[0.0, 6.0, 6.0, 0.0, 2.0];
y =[0.0, 0.0, 3.0, 3.0, 1.0];

ke = zeros(6,6);
dbe = zeros(3,6);
be = zeros(3,6);
de = zeros(3,3);

k = zeros(neq,neq);
f = zeros(neq,1);

for i = 1:nelem
    xe = [x(ide(i,:))] ;
    ye = [y(ide(i,:))] ;
    [ke,dbe,de,be] = plane_stress(xe,ye,E,nu) ;
    lm = [id(ide(i,1),:), id(ide(i,2),:), id(ide(i,3),:)];
    k=assembly(k,ke,lm,1);
end

f = [ 1.5, 1.5, 0.0, 0.0] ;

%reseni posunu
kuu=k(1:4,1:4)
fuu=f(:)
u=kuu\fuu;
u

ug = zeros(neq,1);
ug(1:4)=u(:)

nodal_u = [];
% Vypis posunu podle uzlu
for i=1:size(id,1)
    nodal_u = [nodal_u;ug(id(i,1)),ug(id(i,2))];
end
nodal_u
```

}
}$


```octave
% funkce pocita napeti na hrnach 2d trojuhelnikoveho stenoveho prvku s linearni aproximaci
% In: 
%   xe - vektor x-ovych souradnic uzlu prvku (x1,x2,x3)
%   ye - vektor y-ovych souradnic uzlu prvku (y1,y2,y3)
%   sig - matice napeti v tezisti (3,1)
%
% Out:
%   pse - napeti na hranach (6,1)
%
function pse = boundary_stress (xe,ye,sig) 

pse = zeros(6,1) ;

A = [ sig(1), sig(3);
      sig(3), sig(2)] ;

y23 = ye(2)-ye(3); y31 = ye(3)-ye(1); y12 = ye(1)-ye(2);
x32 = xe(3)-xe(2); x13 = xe(1)-xe(3); x21 = xe(2)-xe(1);

le = length_triangle (xe,ye) ;

c1 = -y12/le(1) ;
s1 = -x21/le(1) ;

T = [ c1, s1; -s1, c1 ];

temp = T*A*[c1;s1];

pse(1) = temp(1,1) ;
pse(2) = temp(2,1) ;

c1 = -y23/le(2) ;
s1 = -x32/le(2) ;

T = [ c1, s1; -s1, c1 ];

temp = T*A*[c1;s1];

pse(3) = temp(1,1) ;
pse(4) = temp(2,1) ;

c1 = -y31/le(3) ;
s1 = -x13/le(3) ;

T = [ c1, s1; -s1, c1 ];

temp = T*A*[c1;s1];

pse(5) = temp(1,1) ;
pse(6) = temp(2,1) ;

end

%reseni relativnich deformaci a napeti

for i = 1:nelem
    xe = [x(ide(i,:))] ;
    ye = [y(ide(i,:))] ;
    [ke,dbe,de,be] = plane_stress(xe,ye,E,nu) ;
    lm = [id(ide(i,1),:), id(ide(i,2),:), id(ide(i,3),:)] ;
    ul = [ug(lm)] ;
    eps = be*ul;
    sig = dbe*ul;
    pse = boundary_stress (xe,ye,sig);
    disp('element number: ')
    disp(i)
    disp('deformation')
    disp(eps')
    disp('stress')
    disp(sig')
    disp('boundary stress')
    disp(pse')
    disp(' ')
end

% Plot deformed configuration over the initial configuration
for i = 1:nnodes
    xnew(i) = x(i) + ug(id(i,1));
    ynew(i) = y(i) + ug(id(i,2));
end

figure; 
for i = 1:nelem
    xe = [x(ide(i,:))] ;
    ye = [y(ide(i,:))] ;
    plot(xe,ye,'k');hold on;
    xe = [xnew(ide(i,:))] ;
    ye = [ynew(ide(i,:))] ;
    plot(xe,ye,'r');hold on;
end
title('Initial and deformed structure'); xlabel('X'); ylabel('Y');
```

### Analytické řešení:

Předpokládáme, že je aproximace posunutí na prvku lineární (implikuje konstantní relativní deformace a konstantní napětí):

$\displaystyle{
u=ax, v=0
}$


Z geometrických rovnic:

$\displaystyle{
\varepsilon_x = \frac{\partial u}{\partial x} = a
}$

$\displaystyle{
\varepsilon_y = \frac{\partial v}{\partial y} = 0
}$

$\displaystyle{
\gamma_{xy} = \frac{\partial u}{\partial y} + \frac{\partial v}{\partial x} = 0
}$


Z konstitutivních rovnic:

$\displaystyle{
\left\{\begin{array}{c} \sigma_x \\ \sigma_y \\ \tau_{xy} \end{array}\right\}
=
\frac{E}{1-\nu^2}\left[
\begin{array}{ccc}
1 & \nu & 0 \\
\nu & 1 & 0 \\
0 & 0 & \frac{1-\nu}{2}
\end{array}
\right]
\left\{\begin{array}{c} \varepsilon_x \\ \varepsilon_y \\ \gamma_{xy} \end{array}\right\}
}$


Z okrajových podmínek:

$\displaystyle{\sigma_x = 1.0}$


Celkem:

$\displaystyle{
\sigma_x = \frac{E}{1-\nu^2}a=1.0 \implies a=0.099=\varepsilon_x
}$

$\displaystyle{
\sigma_y = \frac{E}{1-\nu^2}\nu a=0.1 
}$

$\displaystyle{
\tau_{xy} = 0
}$

$\displaystyle{
u(6)=ax=0.099 \cdot 6=0.594
}$

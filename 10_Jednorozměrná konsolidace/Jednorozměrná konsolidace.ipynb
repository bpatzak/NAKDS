# Jednorozměrná konsolidace

### Motivace

  *  Jeden z důležitých problémů v geotechnice: 'Jak se bude deformovat zemina působením vnějšího zatížení?'

  *  Analýza deformace porézního materiálu

  *  Časově závislý problém

  *  Typický příklad sdruženého problému

  *  Jednoduchý model (viz. I. Vaníček, 'Mechanika zemin',Vydavatelství ČVUT, Praha, 1996):

 <img src="Figures/hydro_mechanical.png" width = 850>

### Diferenciální rovnice jednorozměrné konsolidace

  *  Základní neznámé

     *  Pole posunutí (sednutí) $\mtrx{u}(\mtrx{x},t)$

     *  Pórový tlak $p(\mtrx{x}, t)$

  *  Předpoklady

     *  Rovnoměrný přírůstek zatížení $\overline{\sigma}$ aplikovaný na $1$ m$^2$
  povrchu zeminy (vynechán symbol $\Delta$ pro úsporu místa)

     *  Zemina uvažována jako porézní médium s póry, které jsou <i>zcela</i> vyplněny 
  <i>nestlačitelným médiem</i> (vodou) (tzv. plně saturovaná zemina)

     *  Zanedbáme setrvačné síly, předpokládáme nevazkou tekutinu

     *  Odezvu budeme hledat na časovém intervalu $[0,T]$

  *  Prostorová závislost jen na prostorové proměnné $z \in [0,L]$, $\Rightarrow \Omega = (0,L)$, $\Gamma = \{0,L\}$

  *  Pole posunutí: $w(z,t)$; $w(z=L,t) = 0$ m

  *  Nenulová složka deformace ($z \in \Omega$)
  
  $\displaystyle{\strain_z(z,t) = \frac{\partial w(z,t)}{\partial z}} $

  *  Koncept efektivních napětí 
  
  $\displaystyle{
\stress_z(z,t) = \stress_z^{\rm  eff}(z,t) - p(z,t),
}$

  kde
     *  $\stress_z^{\rm eff}(z,t)$ je efektivní napětí ve skeletu zeminy
     *  $p(z,t)$ je <i>pórový tlak</i>, $p(z=0,t)=\overline{p}(t)$, ($p > 0$ tlak)

### Podmínky rovnováhy

*  Podmínky rovnováhy ($z \in \Omega$)

  $\displaystyle{
\sigma_z(z+\Delta z,t) \cdot 1\rm{ m}^2 - \sigma_z(z,t) \cdot 1\rm{ m}^2 + \gamma(z) \Delta z \cdot 1\rm{ m}^2
= 0
}$


*  Podělením $\Delta z$ a limitním přechodem $\Delta z \rightarrow 0$:

  $\displaystyle{\begin{array}{rcl}
\displaystyle{\frac{\partial}{\partial z}
\stress_z(z,t) + \gamma_z(z)} & = & 0 \\
\displaystyle{\frac{\partial}{\partial z} \sigma_z^{\rm eff}(z,t)-
\frac{\partial}{\partial z} p(z,t) +
\gamma_z(z)} & = & 0
\end{array}}$


*  Podmínka rovnováhy pro $z=0$ m:

  $\displaystyle{\begin{array}{rcl}
\displaystyle{\stress_z(0,t) \cdot 1 \rm{ m}^2  + 
\overline{\sigma} \cdot 1 \rm{ m}^2} & = & 0 \\
\displaystyle{p(0,t) - \stress_z^{\rm eff}(0,t)} & = & \overline{\sigma}
\end{array}}$


### Objemová bilance

Změna objemu materiálu $\equiv$ objemu vytlačené vody (nestlačitelnost)

  *  Změna objemu materiálu (stlačení, za jednotku času) 
      $\displaystyle{\left(\strain_z(z+\frac{\Delta z}{2}, t + \Delta t) - \strain_z(z+\frac{\Delta z}{2}, t)\right) \Delta z \cdot 1\rm{ m}^2 }$


  *  Změna objemu kapaliny: Rychlost proudění
  $v_z(z,t)$ $[\rm{ms}^{-1}$]

   $\displaystyle{ \left(  v_z(z + \Delta z, t + \frac{\Delta t}{2})  - v_z(z, t + \frac{\Delta t}{2}) \right) 1\rm{ m}^2 \cdot \Delta t
   }$


  *  Bilancí objemových změn, podělením $\Delta t \Delta z \cdot  1 \rm{m}^{2}$ a limitním přechodem $\Delta t \rightarrow 0, \Delta z \rightarrow   0$:

   $\displaystyle{
\frac{\partial}{\partial t} \strain_z(z,t) =
\frac{\partial}{\partial z} v_z(z,t)
}$


  *  Okrajová podmínka: $v_z(L,t) = 0$ ms$^{-1}$


#### Konstitutivní rovnice

  *  Pro skelet zeminy: pro $z \in \Omega$:
    $\displaystyle{ \stress_z^{\rm eff}(z,t)  = \frac{E(z)(1-\nu(z))}{(1+\nu(z))(1-2\nu(z))} \strain_z(z,t) = E_{\rm eod}(z) \strain_z(z,t),}$

   kde $E_{\rm eod}$ je tzv. <i>edometrický</i> modul pružnosti.


  *  Rychlost proudění kapaliny v porézním prostředím udává Darcyho zákon
    $\displaystyle{ v_z(z,t) = -k(z)  \frac{\partial}{\partial z} \left( \frac{p(z,t)}{\gamma_z(z)} + z \right), }$

   kde $k$\[ms$^{-1}$\] je filtrační součinitel zeminy (hydraulická vodivost).


  *  Počáteční podmínky (uvažujeme pouze přitížení/odtížení vzhledem k referenčnímu stavu)
    $\displaystyle{ w(z,t_0) = \overline{w}_0(z) = 0\rm{ m} }$

### Diferenciálné rovnice problému

Hledáme $w(z,t)$ a $p(z,t)$ (dostatečně hladké), takové aby:

  *  pro $z \in \Omega$ a $0 < t \leq T$

  $
  \begin{array}{rcl} \displaystyle{-\frac{\partial}{\partial z} \left(  E_{\rm oed}(z)  \frac{\partial w(z,t)}{\partial z}\right) + \frac{\partial p(z,t)}{\partial z}} & = & 
  \gamma_z(z)     \\\displaystyle{ -\frac{\partial^2 w(z,t)}{\partial z\partial t} -\frac{\partial}{\partial z}     \left( k(z) \frac{\partial}{\partial z} \left( \frac{p(z,t)}{\gamma_z(z)} + z \right) \right)}     
  & = & 0 \end{array}
  $

  *  Okrajové podmínky:

  $\displaystyle{\begin{array}{rcl}    \displaystyle{p(0,t)=\overline{p} }&& v_z(L,t)=0 \\    \displaystyle{\sigma(0,t)-\overline{\sigma}} && w(L,t)=0    \end{array}}$

  *  Počáteční podmínky

  $\displaystyle{w(z,0) = 0}$

#### Slabé řešení

  *  Použime časovou diskretizaci diferenčními schematy

  *  Diferenciální rovnice přenásobíme testovacími funkcemi ($\delta w(z)$ a $\delta
  p(z)$) a provedeme integraci na řešené oblasti

  $\displaystyle{\begin{array}{r}
  \displaystyle{\int_\Omega
  \delta w(z)
  \left(
  -\frac{\partial}{\partial z}
  \left( 
  E_{\rm oed}(z) 
  \frac{\partial w(z,t)}{\partial z}
  \right)
  +
  \frac{\partial p(z,t)}{\partial z}   -
  \gamma_z(z)
  \right)
  d z   =
  0
  \\
  \int_\Omega
  \delta p(z)
  \left(   -
  \frac{\partial}{\partial t}
  \left( \frac{\partial w(z,t)}{\partial z}\right)   -
  \frac{\partial}{\partial z}
  \left(
  k(z) 
  \frac{\partial}{\partial z}
  \left(
  \frac{p(z,t)}{\gamma_z(z)} + z
  \right)
  \right)
  \right)
  d z   =0}
  \end{array}}$

  *  Provedeme integraci per-partes a uvažováním následující MKP aproximace
  
  $\displaystyle{\begin{array}{rcl}
  w^e(z,t) = \mtrx{N}_w^e(z) \mtrx{w}^e(t) &&
  p^e(z,t) = \mtrx{N}_p^e(z) \mtrx{p}^e(t) \\
  \delta w^e(z) = \mtrx{N}_w^e(z) \delta \mtrx{w}^e &&
  \delta p^e(z,t) = \mtrx{N}_p^e(z) \delta \mtrx{p}^e 
  \end{array}}$
}

*  obdržíme následující soustavu rovnic

  $\displaystyle{\begin{array}{rcl}   \displaystyle{\sum_e \left(\delta \mtrx{w}^e\right)\trn \left(
  \mtrx{K}_w^e \mtrx{w}^e(t) + \mtrx{K}^e_{wp} \mtrx{p}^e(t)   - \mtrx{f}_w^e(t) \right)} 
  & = & 0   
  \\ \displaystyle{\sum_e   \left(\delta \mtrx{p}^e\right)\trn   \left(
  \mtrx{K}_{wp}^e \frac{d \mtrx{w}^e(t)}{d t}   + \mtrx{K}_p^e \mtrx{p}^e(t)   -   \mtrx{f}^e_p(t)  \right)} 
  & = &   0  \end{array}}
  $

*  Vyjádření v globálním tvaru 

  $\displaystyle{\begin{array}{rcl}
  \displaystyle{\mtrx{K}_w \mtrx{w}(t) + \mtrx{K}_{wp} \mtrx{p}(t) }
  & = & 
  \displaystyle{\mtrx{f}_w(t) + \mtrx{R}_w(t)}
  \\
  \displaystyle{\mtrx{K}_{wp}\trn \frac{d \mtrx{w}(t)}{d t} +  \mtrx{K}_p \mtrx{p}(t)}
  & = &
  \displaystyle{\mtrx{f}_p(t) + \mtrx{R}_p(t) }
  \\
  \displaystyle{\mtrx{w}(0)} 
  & = &
  \displaystyle{ \mtrx{0}}
  \end{array}}$

*  Rovnice formálně vyjádříme ve stejném tvaru jako v případě nestacionárního vedení tepla

$\displaystyle{\begin{array}{rcrll}
&&
\myMatrix{cc}{
 \mtrx{0} & \mtrx{0} \\
 \mtrx{K}_{wp}\trn & \mtrx{0}
}
\frac{d}{d t}
\myMatrix{c}{
\mtrx{w}(t) \\
\mtrx{p}(t)
}
+
\myMatrix{cc}{
\mtrx{K}_w & \mtrx{K}_{wp} \\
\mtrx{0} & \mtrx{K}_{p}
}
\myMatrix{c}{
\mtrx{w}(t) \\
\mtrx{p}(t)
}
& = & 
\myMatrix{c}{
\mtrx{f}_w(t) \\
\mtrx{f}_p(t)
}
+
\myMatrix{c}{
\mtrx{R}_w(t) \\
\mtrx{R}_p(t)
} 
 \\&&
\mtrx{0}
& = &
\myMatrix{c}{
\mtrx{w}(t) \\
\mtrx{p}(t)
}
\end{array}}$

  *  V kompaktní formě: 

  $\displaystyle{\displaystyle
  \mtrx{P} \frac{d \mtrx{r}(t)}{d t}   +    \mtrx{K} \mtrx{r}(t)   =   \mtrx{f}(t) + \mtrx{R}(t)}$

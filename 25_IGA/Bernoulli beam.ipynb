# B-spline bázové funkce

```octave
% basis functions, p = 2, 1 knospan
ksi = 0:0.01:1;

N2 = [(1-ksi).^2; 2*ksi.*(1-ksi); ksi.^2];
B2 = [2*ksi-2; -4*ksi+2; 2*ksi];

subplot(1,2,1)
plot(ksi, N2(1,:), 'r'); hold on;
plot(ksi, N2(2,:), 'b'); hold on;
plot(ksi, N2(3,:), 'g'); hold on; title('N, p=2')

subplot(1,2,2)
plot(ksi, B2(1,:), 'r'); hold on;
plot(ksi, B2(2,:), 'b'); hold on;
plot(ksi, B2(3,:), 'g'); hold on; title('B, p=2')
```

```octave
% basis functions, p = 3, 1 knotspan
ksi = 0:0.01:1;

N3 = [(1-ksi).^3; 3*ksi.*(1-ksi).^2; 3*ksi.^2.*(1-ksi); ksi.^3];
B3 = [-3*(1-ksi).^2; 3*(1-ksi).^2-6*ksi.*(1-ksi); 6*ksi.*(1-ksi)-3*ksi.^2; 3*ksi.^2];

subplot(1,2,1)
plot(ksi, N3(1,:), 'r'); hold on;
plot(ksi, N3(2,:), 'b'); hold on;
plot(ksi, N3(3,:), 'g'); hold on;
plot(ksi, N3(4,:), 'k'); hold on; title('N, p=3')

subplot(1,2,2)
plot(ksi, B3(1,:), 'r'); hold on;
plot(ksi, B3(2,:), 'b'); hold on;
plot(ksi, B3(3,:), 'g'); hold on;
plot(ksi, B3(4,:), 'k'); hold on; title('B, p=3')
```

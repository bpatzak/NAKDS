# Rovinné nosníky

Uvažujeme nosník v rovině $xz$ a dále přijmeme následující předpoklady:
  * zatížení působí pouze v rovině nosníku
  * nestlačitelnost nosníku po výšce průřezu
  * průřez zůstává i po deformaci rovinný
  
Z prvního předpokladu plyne nezávislost řešení na souřadnici $y$, z druhého a třetího pak konstantní průhyb po výšce průřezu

$$
\begin{array}{rcl}
u(x,y,z)&=&u(x,z),\\
w(x,y,z)&=&w(x).
\end{array}
$$



Existují dvě základní nosníkové teorie:
 * Eulerova (Kirchhoffova, Bernoulli-Navierova) teorie - tenké nosníky, zanedbání vlivu smyku
 * Mindlinova (Timoshenkova, Reissnerova) teorie - tlusté nosníky, vliv smyku
  
Mindlinova teorie, narozdíl od Eulerovy, nepředpokládá zachování kolmosti průřezu k deformované střednici a uvažuje nejen ohybový, ale i smykový účinek. Eulerova teorie je tedy vhodná pro tenké nosníky ($h/L < 1/10$), kde lze vliv smyku zanedbat, pro tlusté nosníky je nezbytné použít Mindlinovu teorii. Právě touto teorií se budeme dále zabývat.

### Mindlinova teorie

<img src="Figures/mindlin.jpg" width="400">
  
Kinematika průřezu (rovina xz)

$\qquad\displaystyle{u(x,z) = u_s(x)+\varphi_y (x)z}$

$\qquad\displaystyle{v = 0}$

$\qquad\displaystyle{w(x,z) = w(x)}$

Nenulové složky deformace

$\qquad\displaystyle{\varepsilon_x = \frac{\partial u}{\partial x} = \frac{\partial u_s}{\partial x} + \frac{\partial \varphi_y}{\partial x}z = \varepsilon_s+\kappa_y z
}$


$\qquad\displaystyle{\gamma_{zx} = \frac{\partial w}{\partial x} + \frac{\partial u}{\partial z} = \frac{\partial w}{\partial x} + \varphi_y
}$

Diferenciální rovnice problému představující podmínky rovnováhy (vodorovnou, momentovou a svislou vyjádřené ve vnitřních silách)
<img src="Figures\equilib3.png" width=600>

$\qquad\displaystyle{
\frac{d}{dx}\left(EA\frac{d u_s}{dx}\right)+f_x=0
}$

$\qquad\displaystyle{
\frac{d}{dx}\left(kGA\left(\frac{dw}{dx}+\varphi_y\right)\right)+f_z =0
}$

$\qquad\displaystyle{
\frac{d}{dx}\left(EI_y\frac{d\varphi_y}{dx}\right)-kGA\left(\frac{dw}{dx}+\varphi_y\right)=0
}$

Okrajové podmínky
 * Kinematické podmínky $\left(\rm{předepsané\ }u_s, w, \varphi_y\right)$
 * Statické podmínky $\left(N(x) = \bar{N}(x), V(x) = \bar{V}(x), M(x) = \bar{M}(x)\right)$



### Slabé řešení 

Budeme požadovat splnění následujících rovnic

$\displaystyle{
0 = \int_{\Omega} \delta u_s \left(  \frac{dN}{dx}+f_x \right) \ dx,
}$

$\displaystyle{
0 = \int_{\Omega} \delta w \left(  \frac{dV}{dx}+f_z \right) \ dx,
}$

$\displaystyle{
0 = \int_{\Omega} \delta \varphi_y \left(  \frac{dM}{dx}-V \right) \ dx,
}$

pro všechna $\delta u_s, \delta w, \delta\varphi$ splňující kinematické okrajové podmínky. Dosazením konstitutivních rovnic

$\displaystyle{
0 = \int_{\Omega} \delta u_s \left(  \frac{d}{dx}\left(EA\frac{d u_s}{dx}\right)+f_x \right) \ dx,
}$

$\displaystyle{
0 = \int_{\Omega} \delta w \left(  \frac{d}{dx}\left(kGA\left(\frac{dw}{dx}+\varphi_y\right)\right)+f_z \right) \ dx,
}$

$\displaystyle{
0 = \int_{\Omega} \delta \varphi_y \left(  \frac{d}{dx}\left(EI_y\frac{d\varphi_y}{dx}\right)-kGA\left(\frac{dw}{dx}+\varphi_y\right) \right) \ dx.
}$

a aplikací integrace per-partes získáme


$\displaystyle{
0 = \int_{\Gamma}\delta u_s \left( EA\frac{d u_s}{dx}\right)n \ dx-\int_{\Omega} \frac{d(\delta u_s)}{dx} \left( EA\frac{d u_s}{dx}\right) \ dx + \int_{\Omega}  \delta u_s \bar{f_x} \ dx,
}$

$\displaystyle{
0 =  \int_{\Gamma} \delta w \left(kGA\left(\frac{dw}{dx}+\varphi_y\right)\right) n\ dx -
 \int_{\Omega} \frac{d(\delta w)}{dx} \left(kGA\left(\frac{dw}{dx}+\varphi_y\right)\right) \ dx+ \int_{\Omega} \delta w \bar{f_z} \ dx,
}$

$\displaystyle{
0 = \int_{\Gamma} \delta \varphi_y  \left(EI_y\frac{d\varphi_y}{dx}\right)
-\int_{\Omega} \frac{d(\delta \varphi_y)}{dx} \left(  EI_y\frac{d\varphi_y}{dx}\right)\ dx
-\int_{\Omega} \delta\varphi_y\left( kGA\left(\frac{dw}{dx}+\varphi_y\right) \right) \ dx.
}$

Hraniční integrály můžeme rozdělit na část hranice $\Gamma_u$, kde působí kinematické okrajové podmínky, a na část $\Gamma_p$, kde působí silové okrajové podmínky. Úpravou výrazů obdržíme finální tvar rovnic





$\displaystyle{
\int_{\Omega} \frac{d(\delta u_s)}{dx} EA \frac{d u_s}{dx} \ dx = \left[\delta u_s \bar{N}\right]_{\Gamma_p} + \int_{\Omega}  \delta u_s \bar{f_x} \ dx,
}
$

$\displaystyle{
\int_{\Omega} \frac{d(\delta w)}{dx} kGA \left(\frac{dw}{dx}+\varphi_y\right) \ dx= \left[\delta w \bar{V}\right]_{\Gamma_p} + \int_{\Omega}  \delta w \bar{f_z} \ dx,
}
$

$
\displaystyle{
\int_{\Omega} \frac{d(\delta \varphi_y)}{dx} EI_y \frac{d \varphi_y}{dx} \ dx + \int_{\Omega} \delta \varphi_y kGA \left(\frac{dw}{dx}+\varphi_y\right) \ dx = \left[\delta \varphi_y \bar{M}\right]_{\Gamma_p}.
}
$

### Diskretizovaná úloha

Konstrukci rozdělíme na $n$ prvků a neznámé na každém prvku budeme aproximovat pomocí lineární kombinace interpolačních funkcí $N_i$ příslušejících každé aproximované neznámé a uzlových hodnot $u_i, w_i, \varphi_{y_i}$. 

$\displaystyle{
u(x) \approx \boldsymbol{N}_u^e\boldsymbol {r}_u^e, 
\quad
w(x) \approx \boldsymbol{N}_w^e\boldsymbol {r}_w^e, 
\quad
\varphi_y(x) \approx \boldsymbol{N}_{\varphi}^e\boldsymbol {r}_{\varphi}^e,
}$
 
derivace pak vyjádříme prostřednictvím derivací interpolačních funkcí $B_i$ a uzlových hodnot

$\displaystyle{
\frac{du(x)}{dx} \approx \sum\frac{dN^e_i}{dx}u^e_i = \boldsymbol{B}_u^e\boldsymbol {r}_u^e, 
\quad
\frac{dw(x)}{dx} \approx \sum\frac{dN^e_i}{dx}w^e_i = \boldsymbol{B}_w^e\boldsymbol {r}_w^e, 
\quad
\frac{d\varphi_y(x)}{dx} \approx \sum\frac{dN^e_i}{dx}\varphi^e_i = \boldsymbol{B}_{\varphi}^e\boldsymbol {r}_{\varphi}^e.
}$
 
Dosazením do slabého řešení získáme výslednou sostavu ve tvaru
 
$\displaystyle{ 
\left[
\begin{array}{rrr}
 \boldsymbol{K_{uu}^e} & 0 & 0 \\
0 & \boldsymbol{K_{ww}^e} & \boldsymbol{K_{w\varphi}^e}\\
0 & \boldsymbol{K_{\varphi w}^e} & \boldsymbol{K_{\varphi \varphi}^e} 
\end{array}
\right]\left\{
\begin{array}{r}
 r_u^e \\
r_w^e\\
r_\varphi^e
\end{array}
\right\}=\left\{
\begin{array}{r}
 R_{u}^e \\
R_{w}^e\\
R_{\varphi}^e
\end{array}
\right\},
}$

kde jednotlivé submatice mají tvar

$\displaystyle{\boldsymbol{K_{uu}^e} = \int\boldsymbol{B}_u^{eT}EA\boldsymbol{B}_u^e\ dx, 
}$

$\displaystyle{\boldsymbol{K_{ww}^e} = \int\boldsymbol{B}_w^{eT}kGA\boldsymbol{B}_w^e \ dx,
}$

$\displaystyle{\boldsymbol{K_{w\varphi}^e} = \int\boldsymbol{B}_w^{eT}kGA\boldsymbol{N}_\varphi^e \ dx,
}$

$\displaystyle{\boldsymbol{K_{\varphi w}^e} = \int\boldsymbol{N}_\varphi^{eT}kGA\boldsymbol{B}_w^{e},
}$

$\displaystyle{\boldsymbol{K_{\varphi\varphi}^e} =\int\boldsymbol{B}_\varphi^{eT}EI\boldsymbol{B}_\varphi^e\,dx+\int\boldsymbol{N}_\varphi^ekGA\boldsymbol{N}_\varphi^e \ dx.
}$

```octave
function Ke = beam2d_stiffness(xz, EA, kGA, EI)

    l = sqrt((xz(3) - xz(1))^2 + (xz(4) - xz(2))^2)
    
    Kuu = EA/l*[1 -1; -1 1];
    [Kww, Kwf, Kff] = beam_stiffness (kGA, EI, l);
     
    O = zeros(2);
    
    Ke = [Kuu O O; O Kww Kwf; O Kwf' Kff]
end
```

```octave
% pole souradnic uzlu
xz=[0, 0; 
    1, 0];
    
% pole kodovych cisel
global lm=[1, 2, 3, 4, 5, 6];

% pole uzlu prvku
elnode=[1 2];

%pocet prvku 
nelem=1;

%prurez
b = 1; E = 1; G = 1; h = 0.1;
EA = E * b * h;
EI = E * 1/12 * b * h^3;
kGA = 5/6 * G * b * h; 

%nulovani vektoru zatizeni, matice tuhosti
f=zeros(6,1);
k=zeros(6);
r=zeros(6,1);

%sestaveni matic tuhosti
for i=1:nelem;
kel=beam2d_stiffness([xz(elnode(i,1),:) xz(elnode(i,2),:)], EA, kGA, EI);
%lokalizace
k(lm(i,:),lm(i,:)) = k(lm(i,:),lm(i,:)) + kel;
end

%vektor zatizeni
f(4)=1;

%reseni posunuti
kuu=k([2,4,6],[2,4,6])
fu=f([2,4,6])
u=kuu\fu;

%dopocteni reakci
k([1,3,5],[2,4,6])
R=k([1,3,5],[2,4,6])*u;
R

%rekonstrukce celeho vektoru posunuti
r=[0;u(1);0;u(2);0;u(3)];
r
```

## Smykové zamykání

Při stejné volbě aproximace průhybu a natočení je aproximace posouvající síly o stupeň vyšší než aproximace momentu. To je v rozporu se Schwedlerovou větou a výsledek je pro štíhlé pruty příliš tuhý - dochází k tzv. smykovému uzamknutí.  

* Posouvající síla $\displaystyle{V(x) = kGA\left(\frac{dw}{dx}+\varphi_y\right)}$ - lineární


* Ohybový moment $\displaystyle{M(x) = EI\frac{d\varphi_y}{dx}}$ - konstantní


* Schwedlerova věta  $\displaystyle{\frac{dM(x)}{dx}-V(x) = 0}$

### Ilustrace smykového zatuhnutí

<img src="Figures/konzola_reseni_plnaintegrace2.jpg" width="750">

Průhyb na konzoli (nosníkové řešení)

$\displaystyle{w = \frac{FL^3}{3EI} + \frac{FL}{kGA}}$

Pro náš příklad


$\displaystyle{
\begin{array}{rcl}
w_{(h = 0.1)} &=& 4\cdot 10^3 + 12,\\
w_{(h = 0.1)} &=& 4\cdot 10^6 + 120.
\end{array}
}$

Výsledný poměr

$\displaystyle{
\begin{array}{rcl}
w_{(h = 0.1)}/w_{(h = 0.1)} &=& 1000 !!!
\end{array}
}$



Problém smykového zamykání lze řešit např. redukovanou integrací, vylepšením aproximace průhybu nebo vedlejší podmínkou.

### Redukovaná integrace

Rozpor se způsoben rozporem v aproximaci momentu (křivosti) a posouvající síly (smyková deformace). Princip redkované integrace je založen na integraci příspěvků od smyku ve slabém řešení tak, jako by to byla konstantní funkce (lineární průběh tedy nahradíme konstantní, průměrnou hodnotou ve stredu prvku). 

Smykové zkosení $\displaystyle{\gamma = \frac{dw}{dx} + \varphi}$ dále tedy uvažujeme jako konstantní, jehož hodnota odpovítá $\displaystyle{\bar{\gamma} = \frac{w_2-w_1}{l} + \frac{\varphi_1+\varphi_2}{2}}$. To je ekvivalentní jednobodové (redukované) integraci, kdy příslušný integrační bod je uprostřed intervalu.

Uvažování konstantního průběhu smykového zkosení ovlivní ty části matice tuhosti, do kterých vstupuje $N_\varphi$. V případě lineární aproximace se změní pouze submatice $\boldsymbol{K_{\varphi\varphi}}$, protože členy v integrálu mají parabolický průběh. Pokud integrujeme lineární funkce, je jednobodová integrace v podstatě plnou integrací.

$\qquad\displaystyle{\boldsymbol{K_{\varphi\varphi}^e} = \frac{EI}{l}\left[
\begin{array}{rr}
1&-1\\
-1&1
\end{array}
\right] +  kGA\left[
\begin{array}{rr}
\boldsymbol{l/4}&\boldsymbol{l/4}\\
\boldsymbol{l/4}&\boldsymbol{l/4}
\end{array}
\right]
}$



### Hierarchická funkce

V případě lineární aproximace pro smykové zkosení platí:
 
$\qquad\displaystyle{\gamma = \frac{dw}{dx} + \varphi = \frac{w_2-w_1}{l} + \varphi_1 + \frac{x}{l}(\varphi_2-\varphi_1) = \rm{konst} +  \frac{x}{l}(\varphi_2-\varphi_1)}$

Aby bylo smykové zkosení alespoň konstantní, je možné vylepšit aproximaci průhybu tak, aby vymizel lineární člen. Toho docílíme, pokud budeme volit: 

$\qquad\displaystyle{w(x) = w^{lin} + \frac{1}{2l}(\varphi_1-\varphi_2)x(x-l)}$

Výsledná aproximace:

$\qquad\displaystyle{u_s(x) = N_1 u_{s1}+ N_2 u_{s2}}$

$\qquad\displaystyle{w(x) = N_1 w_{1}+ N_2 w_{2} + \frac{N_3}{2l}\varphi_1 - \frac{N_3}{2l}\varphi_2}$


$\qquad\displaystyle{u_s(x) = N_1 \varphi_{1}+ N_2 \varphi_{2}}$

Pro výpočet je v tomto případě praktické uspořádat uzlové hodnoty průhybů a natočení do jednoho vektoru $r_{w,\varphi} = \{w_1, \varphi_1, w_2, \varphi_2\}$.

Pro interpolaci průhybu a natočení pak platí

$w(x) = \displaystyle{N_w r_{w,\varphi} = [1-{x\over l}, {x(x-l) \over 2l}, {x\over l}. {-x(x-l)\over 2l}]}$

$\varphi(x) = \displaystyle{N_\varphi r_{w,\varphi} = [0, {1-x \over l}, 0, {x \over l}]}$

Sub-Matice tuhosti pak maji tvar
$K_{ww}=\int B_w^T KGA B_w dx = \displaystyle{\left[\begin{array}{cccc}
{kGA\over l}& 0&-{kGA\over l}&0\\
0&{kGAl\over 12}&0&-{kGAl\over 12}\\
-{kga\over l}&0&{kGA\over l}&0\\
0&-{kGAl\over 12}&0&{kGAl\over 12}
\end{array}
\right]}
$

$K_{w\varphi}=\int B_W^TkGA N_\varphi dx = \displaystyle{\left[\begin{array}{cccc}
0&-{kGA\over 2}&0&-{kGA\over 2}\\
0&-{kGAl\over 12}&0&{kGAl\over 12}\\
0&{kGA\over 2}&0&{kGA\over 2}\\
0&{kGAl\over 12}&0&-{kGAl\over 12}
\end{array}
\right]}
$

$K_{\varphi,\varphi}=\int B_\varphi^TEI B_\varphi dx + \int N_\varphi^T kGA N\varphi dx= \displaystyle{\left[\begin{array}{cccc}
0&0&0&0\\
0& {EI\over l}+{kGAl\over 3}&0& -{EI\over l}+{kGAl\over 6}\\
0&0&0&0\\
0&-{EI\over l}+{kGAl\over 6}&0&  {EI\over l}+{kGAl\over 3}\\
\end{array}
\right]}
$

Po sečtení členů dá stejný výsledek jako redukovaná integrace.

### Vedlejší podmínka

Výpočet smykové síly nahradíme výpočtem z průměrné smykové deformace $\bar{\gamma}$

$
V = kGA \bar{\gamma}
$

a takto spočtenou posouvající sílu dosadíme do diferenciálních rovnic

$
\displaystyle{
\frac{d}{dx}\left(kGA\bar{\gamma}\right) + f_z = 0,\\
\frac{d}{dx}\left(EI\frac{d\varphi}{dx}\right) - kGA\bar{\gamma} = 0.
}
$

Takto upravené rovnice vedou k následujícímu tvaru slabého řešení 

$\displaystyle{
\int_{\Omega} \frac{d(\delta w)}{dx} kGA\bar{\gamma} \ dx= \left[\delta w \bar{V}\right]_{\Gamma_p} + \int_{\Omega}  \delta w \bar{f_z} \ dx,
}
$

$
\displaystyle{
\int_{\Omega} \frac{d(\delta \varphi_y)}{dx} EI_y \frac{d \varphi_y}{dx} \ dx + \int_{\Omega} \delta \varphi_y kGA\bar{\gamma} \ dx = \left[\delta \varphi_y \bar{M}\right]_{\Gamma_p}.
}
$

Do soustavy rovnic navíc musíme přidat rovici, která nám svazuje průměrnou smykovou deformaci se skutečnou $\displaystyle{\gamma = \frac{dw}{dx}+\varphi_y}$, vedlejší podmínka v průměru (s vahou $\delta \lambda$) vyžaduje rovnost obou deformací

$\displaystyle{
\int_{\Omega}\delta \lambda\left(\frac{dw}{dx}+\varphi_y-\bar{\gamma} \right) = 0.
}
$

Poslední rovnici je vhodné vynásobit $kGA$, aby výsledná soustava byla symetrická.

Průměrnou smykovou deformaci uvažujeme jako konstantní, tedy ve tvaru $\bar{\gamma}=N_\gamma \bar\gamma = 1\  \bar\gamma$, podobně aproximuje i váhovou funkci $\delta \lambda = 1\ \delta \lambda$. Po diskretizaci dostaneme 

$\displaystyle{
\underbrace{\int_{\Omega} \boldsymbol{B}_{w}^T kGA\bar{\gamma} \ dx}_{K_{w\bar{\gamma}}}= \left[\delta w \bar{V}\right]_{\Gamma_p} + \int_{\Omega}  \delta w \bar{f_z} \ dx,
}
$

$
\displaystyle{
\underbrace{\int_{\Omega} \boldsymbol{B}_{\varphi}^T EI_y \boldsymbol{B}_{\varphi} \ dx}_{K_{\varphi\varphi}} \ \boldsymbol{r}_{\varphi} + \underbrace{\int_{\Omega} \boldsymbol{N}_{\varphi} kGA\ dx }_{K_{\varphi\bar{\gamma}}} \ \bar{\gamma}= \left[\delta \varphi_y \bar{M}\right]_{\Gamma_p},
}
$

$\displaystyle{
\int_{\Omega}kGA\left(\boldsymbol{B}_{w}\boldsymbol{r}_{w}+\boldsymbol{N}_{\varphi}\boldsymbol{r}_{\varphi}-\bar{\gamma} \right) \ dx = \underbrace{\int_{\Omega} \boldsymbol{B}_{w}kGA\boldsymbol{r}_{w} \ dx}_{K_{\bar{\gamma}w}} +
\underbrace{\int_{\Omega}\boldsymbol{N}_{\varphi}kGA\boldsymbol{r}_{\varphi} \ dx}_{K_{\bar{\gamma}\varphi}} -
\underbrace{\int_{\Omega}kGA\bar{\gamma}  \ dx}_{K_{\bar{\gamma}\bar{\gamma}}}= 0.
}
$

Výsledná soustava

$
\displaystyle{
\left[
\begin{array}{ccc}
0 & 0 & \boldsymbol{K}_{w\bar{\gamma}}\\
0 & \boldsymbol{K}_{\varphi\varphi} & \boldsymbol{K}_{\varphi\bar{\gamma}}\\
\boldsymbol{K}_{\bar{\gamma}w} & \boldsymbol{K}_{\bar{\gamma}\varphi} & \boldsymbol{K}_{\bar{\gamma}\bar{\gamma}}
\end{array}\right]
\left\{
\begin{array}{c}
\boldsymbol{r}_{w}\\
\boldsymbol{r}_{\varphi}\\
\bar{\gamma}
\end{array}\right\}=
\left\{
\begin{array}{c}
\boldsymbol{f}_{w}\\
\boldsymbol{f}_{\varphi}\\
0
\end{array}\right\}.
}  
$

## Oboustranně vetknutý nosník (spojité zatížení, redukovaná integrace)

```octave
function Ke = beam2d_stiffness_RI(xz, EA, kGA, EI)

    l = sqrt((xz(3) - xz(1))^2 + (xz(4) - xz(2))^2);
    
    Kuu = EA/l*[1 -1; -1 1];
    Kww = kGA/l*[1 -1; -1 1];
    Kwf = kGA*[-1/2 -1/2; 1/2 1/2];
    Kff = EI/l*[1 -1; -1 1] + kGA*[l/4 l/4; l/4 l/4];
    
    O = zeros(2);
    
    Ke = [Kuu O O; O Kww Kwf; O Kwf' Kff];
end

l = 3;
n = 11;
fz = 1;

%prurez
b = 1; E = 1; G = 1/2; h = 0.1;
EA = E * b * h;
EI = E * 1/12 * b * h^3;
kGA = 5/6 * G * b * h; 

k = zeros((n+1)*3);
f = zeros((n+1)*3,1);

dx = l/n;
for i = 1:n
    xz = [(i-1)*dx 0 i*dx 0];
    ke = beam2d_stiffness_RI(xz, EA, kGA, EI);
    loc = [(i-1)*3+1 i*3+1 (i-1)*3+2 i*3+2 (i-1)*3+3 i*3+3];
    k(loc,loc) += ke;
    fe = [0 0 fz*dx/2 fz*dx/2 0 0];
    f(loc) += fe';
end

up = [1 2 3 n*3+1 n*3+2 n*3+3];

kuu = k;
kuu(up,:) = [];
kuu(:,up) = [];
fuu = f;
fuu(up) = [];

u = kuu\fuu;

r = [0; 0; 0; u; 0; 0; 0];

for i = 1:n
    df = (r(i*3+3) - r((i-1)*3+3))/dx;
    M = EI * df;
    plot ([(i-1)*dx (i-1)*dx (i)*dx (i)*dx],[0 -M -M 0]); hold on;
end
plot([0 l], [0 0], 'k')
disp('Prubeh momentu')
```

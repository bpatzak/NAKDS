#### Konstrukce interpolačních funkcí na čtyřúhelníku
   
Formulovat aproximační funkce v globálních souřadnicích je snadné pouze pro obdélníkové prvky. Pro obecné čtyřúhelníky se využívá jejich formulace v parametrickém prostoru, kde čtyřúhelník je mapován na jednotkový čtverec v parametrickém prostoru $\xi,\eta$. Zde se interpolační funkce snadno získají jako produkt jednodimenzionálních interpolačních. Například interpolační funkce $N_2(\xi,\eta)$ se získá jako produkt jedno-dimenzionálních funkcí $N_2(\xi)$ a $N_1(\eta)$.
 
 
<img src="Figures/2dshape.png" width=400>
   
 Interpolační funkce mají tedy tvar

$\displaystyle{
      \begin{array}{ll}
        N_1(\xi,\eta)=\del{1}{4}(1-\xi)(1-\eta) & N_2(\xi,\eta)=\del{1}{4}(1+\xi)(1-\eta)\\
        N_3(\xi,\eta)=\del{1}{4}(1+\xi)(1+\eta) & N_4(\xi,\eta)=\del{1}{4}(1-\xi)(1+\eta)
      \end{array}
}$
    
<img src="Figures/shape_functions.png" width=450>
    
Výsledné interpolační funkce můžeme shrnout do kompaktního výrazu
  
$\displaystyle{
      N_i(\xi,\eta)=\del{1}{4}(1+\xi_i\xi)(1+\eta_i\eta).
    }$

Matice bázových funkcí pro aproximaci průhybu a pootočení budou mít shodný tvar

$\displaystyle{
N_w = N_\varphi = \left\{
\begin{array}{cccc}
N_1&
N_2&
N_3&
N_4
\end{array}
\right\}.
}$
 

 
#### Derivace interpolačních funkcí

Transformace mezi parametrickým a skutečným prostorem je založena na myšlence aproximovat geometrii prvku stejně jako neznémé veličiny

$\displaystyle{
    x(\xi,\eta)=\sum N_i(\xi,\eta)x_i^e,\ \ y(\xi,\eta)=\sum N_i(\xi,\eta)y_i^e.
  }$
  
  
<img src="Figures/mapping3.png", width=500/>

Interpolační funkce jsou definované v parametrickém prostoru, umíme tedy snadno spočíst jejich derivace vzhledem ke $\xi$ a $\eta$. Budeme však potřebovat vyjádřit i derivace vzhledem ke skutečným souřadnicím x a y. Podle pravidla o derivaci složených fukcí můžeme psát
   
$\displaystyle{
      \begin{array}{l}
          \pard{N_i}{\xi}=\pard{N_i}{x}\pard{x}{\xi}+\pard{N_i}{y}\pard{y}{\xi}\\
          \pard{N_i}{\eta}=\pard{N_i}{x}\pard{x}{\eta}+\pard{N_i}{y}\pard{y}{\eta}
      \end{array},
    }$
    
v maticovém zápisu
    
$\displaystyle{
      \left\{
      \begin{array}{l}
        \pard{N_i}{\xi}\\
        \pard{N_i}{\eta}
      \end{array}\right\} = 
      \underbrace{
        \left[
          \begin{array}{cc}
            \pard{x}{\xi} & \pard{y}{\xi}\\
            \pard{x}{\eta} & \pard{y}{\eta}
          \end{array}
          \right]
      }_{\mbf{J}}
      \left\{
      \begin{array}{l}
        \pard{N_i}{x}\\
        \pard{N_i}{y}
      \end{array}
      \right\}.
    }$
    
 
Matice Jakobiánu $\mbf{J}$ váže mezi sebou derivace v obou systémech. Hledané derivace v daném bodě podle skutečných souřadnic $x, z$ získáme inverzí předchozí rovnice
   
$\displaystyle{
      \left\{
      \begin{array}{l}
        \pard{N_i}{x}\\
        \pard{N_i}{y}
      \end{array}\right\} = 
      \mbf{J}^{-1}
      \left\{
      \begin{array}{l}
        \pard{N_i}{\xi}\\
        \pard{N_i}{\eta}
      \end{array}
      \right\}.
    }$
    
Substitucí aproximace geometrie do výrazu pro jakobián obdržíme
      
$\displaystyle{
      \mbf{J}=
      \left[
        \begin{array}{ll}
          \sum\pard{N_i}{\xi}x_i^e & \sum\pard{N_i}{\xi}y_i^e\\
          \sum\pard{N_i}{\eta}x_i^e & \sum\pard{N_i}{\eta}y_i^e
        \end{array}
       \right] =
      \left[ 
        \begin{array}{llll}
          \pard{N_1}{\xi} & \pard{N_2}{\xi} & \pard{N_3}{\xi} & \pard{N_4}{\xi}\\
          \pard{N_1}{\eta} & \pard{N_2}{\eta} & \pard{N_3}{\eta} & \pard{N_4}{\eta}
        \end{array}
       \right]
      \left[
        \begin{array}{ll}
          x_1^e& y_1^e\\
          x_2^e& y_2^e\\
          x_3^e& y_3^e\\
          x_4^e& y_4^e\\
        \end{array}
        \right].
    }$
    
 

 Pro sestavení matice tuhosti potřebujeme znát gradienty $\grad N_w$ a $\grad N_\varphi$
 
 $\displaystyle{
 \grad N_w = \grad N_\varphi = 
 \left\{
 \begin{array}{c}
 \frac{\partial}{\partial x}\\
 \frac{\partial}{\partial y}
 \end{array}
 \right\}
  \left\{
\begin{array}{cccc}
N_1&
N_2&
N_3&
N_4
\end{array}
\right\}
=
 \left\{
 \begin{array}{c}
 \frac{\partial N_1}{\partial x}&\frac{\partial N_2}{\partial x}&\frac{\partial N_3}{\partial x}&\frac{\partial N_4}{\partial x}\\
 \frac{\partial N_1}{\partial y}&\frac{\partial N_2}{\partial y}&\frac{\partial N_3}{\partial y}&\frac{\partial N_4}{\partial y}
 \end{array}
 \right\}
=J^{-1}
 \left\{
 \begin{array}{c}
 \frac{\partial N_1}{\partial \xi}&\frac{\partial N_2}{\partial \xi}&\frac{\partial N_3}{\partial \xi}&\frac{\partial N_4}{\partial \xi}\\
 \frac{\partial N_1}{\partial \eta}&\frac{\partial N_2}{\partial \eta}&\frac{\partial N_3}{\partial \eta}&\frac{\partial N_4}{\partial \eta}
 \end{array}
 \right\}.
 }$

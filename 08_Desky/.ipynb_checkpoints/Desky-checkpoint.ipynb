### Předpoklady


<img src="Figures/plate_deformation.png" width="750">

Pro desku v rovině $xy$ přijmeme  následující předpoklady
  * Zatížení desky působí kolmo ke střednicové rovině
  * Příčné stlačení desky zanedbáváme $\Rightarrow w(x,y,z)=w(x,y)$
  * Normály ke střednicové rovině zůstávájí i po deformaci přímé (pseudonormály), jejich pootočení kolem souřadnicových os označíme jako $\psi_x,\psi_y$.
  
Z uvedených předpokladů plynou následující vztahy pro posun libovolného bodu desky

$\displaystyle{
u(x,y,z)=z\psi_y(x,y)\\
v(x,y,y)=-z\psi_x(x,y),\\
w(x,y,z)=w(x,y).}
$


### Deformace
Na základě uvažovaného pole posunutí stanovíme nenulové složky deformačního tenzoru.

#### Membránové složky

$\displaystyle{\begin{array}{}
  \eps_x&=&\ppd{u}{x}=z\ppd{\psi_y}{x}\\
  \eps_y&=&\ppd{v}{y}=-z\ppd{\psi_x}{y}\\
  \gamma_{xy}&=&\ppd{u}{y}+\ppd{v}{x}=z\left(\ppd{\psi_y}{y}-\ppd{\psi_x}{x}\right)\\
\end{array}}$


$\displaystyle{\begin{array}{rcl}
  \left\{\begin{array}{c}
  \eps_x\\
  \eps_y\\
  \gamma_{xy}
  \end{array}\right\} &=& 
  z\ \left[\begin{array}{cc}
      \ppd{}{x} & 0\\
      0 & \ppd{}{y}\\
      \ppd{}{y}& \ppd{}{x}
    \end{array}\right]
  \left[\begin{array}{cc}
      0 & 1\\
      -1& 0
    \end{array}\right]
  \left\{\begin{array}{c}
  \psi_x\\
  \psi_y
  \end{array}\right\}\\[3mm]
  \mbf{\eps}_m&=&\boldsymbol{\kappa}z = z\ \mbf{\partial}^T\mbf{S}\mbf{\psi}
\end{array}}$


#### Smykové složky
$\displaystyle{\begin{array}{rcl}
  \gamma_{xz}&=&\ppd{u}{z}+\ppd{w}{x}=\psi_y+\ppd{w}{x}\\
  \gamma_{yz}&=&\ppd{v}{z}+\ppd{w}{y}=-\psi_x+\ppd{w}{y}\\[3mm]
  \mbf{\gamma}&=&
  \left\{\begin{array}{c}
  \ppd{}{x}\\ \ppd{}{y}
  \end{array}\right\} w + 
  \left[\begin{array}{cc}
      0 & 1\\
      -1& 0
    \end{array}\right]
  \left\{\begin{array}{c}
  \psi_x\\ \psi_y
  \end{array}\right\}\\[3mm]
  \mbf{\gamma} &=& \mbf{\grad}w + \mbf{S}\mbf{\psi}
  \end{array}}$
  

### Napětí
Představme si desku desku rozřezanou na vrstvičky (rovnoběžné se střednicovou rovinou xy).

<img src="Figures/plate_layers.png" width="250">


  $\displaystyle{\begin{array}{rcl}
    \sigma_x&=&\del{E}{1-\nu^2}(\eps_x+\nu\eps_y) = \del{Ez}{1-\nu^2}\left(\ppd{\psi_y}{x}-\nu\ppd{\psi_x}{y}\right)\\
    \sigma_y&=&\del{E}{1-\nu^2}(\eps_y+\nu\eps_x) = \del{Ez}{1-\nu^2}\left(-\ppd{\psi_x}{y}+\nu\ppd{\psi_y}{x}\right)\\
    \tau_{xy}&=&\del{E}{2(1+\nu)}\gamma_{xy}=\del{E}{2(1+\nu)}\left(\ppd{\psi_y}{y}-\ppd{\psi_x}{x}\right)\\
    \tau_{xz}&=&\del{E}{2(1+\nu)}\gamma_{xz}=\del{E}{2(1+\nu)}\left(\psi_y+\ppd{w}{x}\right)\\
    \tau_{yz}&=&\del{E}{2(1+\nu)}\gamma_{yz}=\del{E}{2(1+\nu)}\left(-\psi_x+\ppd{w}{y}\right)
  \end{array}}$

### Měrné vnitřní síly

<img src="Figures/plate_sm.png" width="500">

  $\displaystyle{\begin{array}{rcl}
    m_x&=&\displaystyle{\int_{-\frac{h}{2}}^{\frac{h}{2}} \sigma_x z\,dz}\\
    m_y&=&\displaystyle{\int_{-\frac{h}{2}}^{\frac{h}{2}} \sigma_y z\,dz}\\
    m_{xy}&=&\displaystyle{\int_{-\frac{h}{2}}^{\frac{h}{2}} \tau_{xy} z\,dz}\\
    q_x&=&\displaystyle{\int_{-\frac{h}{2}}^{\frac{h}{2}} \tau_{xz}\,dz}\\
    q_y&=&\displaystyle{\int_{-\frac{h}{2}}^{\frac{h}{2}} \tau_{yz}\,dz}
  \end{array}}$
  

Tyto integrální veličiny jsou vztaženy na jednotku šířky a proto bývají nazývány jako měrné síly (intenzity vnitřních sil).

### Dělení desek 
  *  Kirchhoffova teorie - předpokládá, že nedochází k příčné smykové deformaci $\left(\gamma_{xz}=\gamma_{yz}=0\right)$, a tedy (z rovnic pro smykové složky deformace)
$\psi_x=\ppd{w}{x}$ a $\psi_{y}=\ppd{w}{y}$ (pseudonormála zůstává i po deformaci kolmá ke střednicové ploše). Vhodná pro tenké desky $\left(\del{1}{100}\leq\del{h}{l}\leq\del{1}{10}\right)$,
  *  Mindlinova teorie - průhyb $w$ a natočení pseudonormály $\psi_x, \psi_y$ jsou nezávislé. Vhodná pro tlusté desky $\left(\del{h}{l}\geq\del{1}{10}\right)$,
  *  Membrány $\left(\del{h}{l}\leq\del{1}{100}\right)$ nepřenášejí ohybové účinky
  
<i>Pozn.: Protože v Kirchhoffově teorii nejsou pootočení nezávislá pole, ale odpovídají derivacím průhybu je výsledná diferenciální rovnice čtvrtého řádu. To ve svém důsledku vyžaduje $C_1$ kontinuitu interpolačních funkcí (spojitost funkční hodnoty a prvních derivací), což vyžaduje komplikovanou formulaci. Alternativní možnost představují DKT prvky, vynucující splnění Kirchhoffových předpokladů jen ve vybraných bodech prvku.</i>

### Mindlinova teorie
  
  ####  Momenty
  
$\displaystyle{\begin{array}{rcccccl}
m_x&=&\displaystyle{\int_{-\frac{h}{2}}^{\frac{h}{2}}\sigma_x z\,dz }&=& \displaystyle{\int_{-\frac{h}{2}}^{\frac{h}{2}}\del{Ez}{1-\nu^2}\left(\ppd{\psi_y}{x}-\nu\ppd{\psi_x}{y}\right)\,z\,dz }
&=&\del{Eh^3}{12(1-\nu^2)} \left(\ppd{\psi_y}{x}-\nu\ppd{\psi_x}{y}\right)=D \left(\ppd{\psi_y}{x}-\nu\ppd{\psi_x}{y}\right)
\end{array}}$

Maticově tedy

$\displaystyle{
\left\{
\begin{array}{c}
  m_x\\
  m_y\\
  m_{xy}
\end{array}
\right\} =
\left[
\begin{array}{ccc}
  D & \nu D &  0\\
  \nu D & D & 0 \\
  0 & 0 & \del{(1-\nu)D}{2}
\end{array}
\right]
\mbf{\kappa} = \mbf{D}_m \mbf{\partial}^T\mbf{S}\mbf{\psi}
}$

####   Měrné posouvající síly
  
$\displaystyle{\begin{array}{rcccccl}
q_x&=&\displaystyle{\int_{-\frac{h}{2}}^{\frac{h}{2}}\tau_{xz}\,dz} &=&\displaystyle{ \int_{-\frac{h}{2}}^{\frac{h}{2}}k \tau \del{E}{2(1+\nu)}\left(\psi_y+\ppd{w}{x}\right)\,dz} 
&=&\displaystyle{k\del{Eh}{2(1+\nu)} \left(\psi_y+\ppd{w}{x}\right)}
\end{array}}$

Maticově tedy

$
\left\{
\begin{array}{c}
  q_x\\
  q_y
\end{array}
\right\} =
\left[
\begin{array}{ccc}
  kGh & 0 \\
  0 & kGh
\end{array}
\right]
\left\{
\begin{array}{c}
  \psi_{y}+\ppd{w}{x}\\
  -\psi_{x}+\ppd{w}{y}
\end{array}
\right\} = \mbf{D}_s(\mbf{\grad}w+\mbf{S}\mbf{\psi})
$

### Statické rovnice
Uvažujme pouze zatížení ve smeru osy z. Podmínky rovnováhy pak mají tvar:

$\displaystyle{\begin{array}{}
  \ppd{\sigma_x}{x}+\ppd{\tau_{xy}}{y}+\ppd{\tau_{xz}}{z} = 0\\
  \ppd{\tau_{xy}}{x}+\ppd{\sigma_y}{y}+\ppd{\tau_{yz}}{z} = 0\\
  \ppd{\tau_{zx}}{x}+\ppd{\tau_{yz}}{y} + \bar{Z} = 0
\end{array}}$

Po přenásobení podmínek rovnováhy souřadnicí $z$ a jejich integrací po výšce průřezu dostaneme

$\displaystyle{\int_{-\frac{h}{2}}^{\frac{h}{2}}\left(\ppd{\sigma_x}{x}+\ppd{\tau_{xy}}{y}+\underbrace{\ppd{\tau_{zx}}{z}}_{*}\right)\,z\,dz = 0}$

Označený (\*) člen upravíme integrací per-partes

$\displaystyle{\int_{-\frac{h}{2}}^{\frac{h}{2}}{\ppd{\tau_{zx}}{z}}\,z\,dz=\underbrace{\left[\tau_{zx}\,z\right]_{-\frac{h}{2}}^{\frac{h}{2}}}_{=0}-\int_{-\frac{h}{2}}^{\frac{h}{2}}\tau_{zx}\,dz}$

Uplatníme definici měrných sil a  nakonec dostaneme 

$\displaystyle{
  \ppd{}{x}m_x+\ppd{}{y}m_{xy}-q_x = 0
}$

Obdobně postupujeme i v případě druhé a třetí (tu jen integrujeme, nenásobíme z) rovnice a obdržíme

$\displaystyle{
\ppd{}{x}m_{xy}+\ppd{}{y}m_y-q_y = 0
}$

$\displaystyle{
\ppd{}{x}q_x+\ppd{}{x}q_y+\bar{f_z} = 0
}$

#### Maticový zápis podmínek rovnováhy

  *  Momentové podmínky
  
$\displaystyle{\begin{array}{rcl}
  \left[
  \begin{array}{ccc}
    \ppd{}{x} & 0 & \ppd{}{y}\\
    0 & \ppd{}{y} & \ppd{}{x}
  \end{array}
  \right] \left\{
  \begin{array}{c}
   m_x\\
   m_y\\
   m_{xy}
  \end{array}
  \right\} - \left\{
  \begin{array}{c}
   q_x\\
   q_y
  \end{array}
  \right\} &=& \left\{
  \begin{array}{c}
    0\\0
  \end{array}
  \right\}\\
  \mbf{\partial}\mbf{m}-\mbf{q} &=& \mbf{0}
\end{array}}$


  *  Silová podmínka
  
$\displaystyle{\begin{array}{rcl}
  \left\{
  \begin{array}{cc}
    \ppd{}{x}& \ppd{}{y}
  \end{array}
\right\}
\left\{
\begin{array}{c}
  q_x\\
  q_y
\end{array}
\right\} + \bar{f_z} &=& 0\\
\mbf{\grad}^T\mbf{q}+\bar{f_z} &=& 0
\end{array}}$

### Slabé řešení
Vyjdeme z maticového zápisu podmínek rovnováhy. Pro libovolné $\delta\varphi, \delta w$ splňující $\delta\varphi=0\ {\rm na}\ \Gamma_{mu}$ a $\delta w = 0\ {\rm na}\ \Gamma_{wu}$ platí

$\displaystyle{\begin{array}{}
  \int_{\Omega}(\mbf{S}\mbf{\delta\varphi})^T (\mbf{\partial m}-\mbf{q})\,dx &=& 0\\
  \int_{\Omega}\delta w^T (\mbf{\grad^T q}+\bar{f_z})\,dx &=& 0
\end{array}}$

První rovnici upravíme použitím Clapeyronova teorému

$\displaystyle{\begin{array}{rclcl}
  0 &=& \int_{\Omega}(\mbf{S}\mbf{\delta\varphi})^T (\mbf{\partial m}-\mbf{q})\,dx\\
  &=& \int_{\Gamma_m}(\mbf{S}\mbf{\delta\varphi})^T \overbrace{\mbf{n\cdot m}}^{\bar{\bf{m}}\ {\rm na}\ \Gamma_m}\,dx - \int_{\Omega}(\mbf{\partial}^T\mbf{S\delta\varphi})^T\ \mbf{m}\, dx
  - \int_{\Omega}\mbf{S\delta\varphi}^T\ \mbf{q}\, dx
\end{array}}$

Silovou podmínku upravíme pomocí Gaussova-Greenova vztahu

$\displaystyle{\begin{array}{rclcl}
  0 &=& \int_{\Omega}\delta w^T (\mbf{\grad^T q}+\bar{f_z})\,dx\\
  &=& \int_{\Gamma_q}\delta w\ \overbrace{\mbf{n}^T\cdot\mbf{q}}^{\bar{q}\ {\rm na}\ \Gamma_q}\,dx - \int_{\Omega}(\mbf{\grad}\delta w)^T\mbf{q}\,dx
+ \int_{\Omega}\delta w \bar{f_z}\,dx
\end{array}}$


 *  Výsledné vztahy
 
    $\displaystyle{\begin{array}{rcl}
      \int_{\Omega}(\mbf{\partial}^T\mbf{S\delta\varphi})^T\ \mbf{m}\, dx + \int_{\Omega}(\mbf{S}\mbf{\delta\varphi})^T\mbf{q}\,dx &=& \int_{\Gamma_m}(\mbf{S\delta\varphi})^T\bar{\mbf{m}}\,dx\\
      \int_{\Omega}(\mbf{\grad}\delta w)^T\mbf{q}\,dx &=& \int_{\Gamma_q}\delta w \bar{q}\,dx + \int_{\Omega}\delta w \bar{f_z}\,dx
    \end{array}}$
    
    
  *  Postup diskretizace je obdobný jako u ostatních úloh
  
  $w(x,y) = \mbf{N}_w \mbf{r}_w,\;\;\{\varphi_x,\varphi_y\}^t=\mbf{N}_{\psi} \mbf{r}_{\varphi}$

   a podobně i testovací funkce.
   Následně můžeme psát

   $\displaystyle{\begin{array}{rcl}
 \sum_e \delta\mbf{r}^T_{\varphi} \left(\int(\mbf{\partial}^T\mbf{S}\mbf{N}_{\psi})^T \mbf{D}_m\ \mbf{\partial}^T\mbf{S}\mbf{N}_{\psi}\, d\Omega^e\,\mbf{r}_{\varphi} \right.
 \left.\;\;\; +\int \left(\mbf{S}\mbf{N}_{\psi}\right)^T \mbf{D}_s\ (\mbf{\grad}\mbf{N}_w\mbf{r}_w+\mbf{S}\mbf{N}_{\psi}\mbf{r}_{\varphi})\,d\Omega^e - \mbf{f}_m\right) &=& 0\\
 \sum_e\delta\mbf{r}^T_w \left(\int(\mbf{\grad}\mbf{N}_w)^T\mbf{D}_s\ (\mbf{\grad N_w}\mbf{r}_w+\mbf{S}\mbf{N}_{\psi}\mbf{r}_{\varphi})\,d\Omega^e - \mbf{f}_w\right) &=& 0\\
\end{array}}$

Protože váhové funkce $\delta\varphi, \delta w $ mohou být voleny libovolně a jsou tedy různé od nuly, stejně jako vektory jejich uzlových hodnot  $\delta\mbf{r}^T_\varphi \delta\mbf{r}^T_w$, výrazy v závorkách musí být rovny nule.

#### Matice tuhosti
Výrazy v závorkách můžeme zapsat v kompaktní formě

$\displaystyle{
  \sum_e\left(
  \left[
    \begin{array}{ll}
      \mbf{K}_{\varphi\varphi}^{e,1} + \mbf{K}_{\varphi\varphi}^{e,2} & \mbf{K}_{\varphi w}^e\\[3mm]
      (\mbf{K}^e_{\varphi w})^T& \mbf{K}^e_{ww}
    \end{array}
  \right]
  \left\{
  \begin{array}{l}
    \mbf{r}^e_{\varphi}\\[3mm]
    \mbf{r}^e_{w}
  \end{array}
  \right\}
  - 
  \left\{
  \begin{array}{l}
    \mbf{f}^e_\varphi\\[3mm]
    \mbf{f}^e_w
  \end{array}
  \right\}
  \right) = 0
}$

Kde jednotlivé submatice tvořící matici tuhosti deskového prvku mají tvar

  *  $\mbf{K}^{e,1}_{\varphi\varphi}=\int(\mbf{\partial^T SN_{\varphi}})^T \mbf{D}_m\ \mbf{\partial^T S N_{\varphi}}, d\Omega^e$
  *  $\mbf{K}^{e,2}_{\varphi\varphi}=\int\left(\mbf{S}\mbf{N}_{\psi}\right)^T \mbf{D}_s\ \mbf{S}\mbf{N}_{\psi}\,d\Omega^e$
  *  $\mbf{K}^e_{\varphi w} = \int\left(\mbf{S}\mbf{N}_{\psi}\right)^T \mbf{D}_s\ \mbf{\grad N_w}\,d\Omega^e$
  *  $\mbf{K}^e_{ww} = \int(\mbf{\grad N_w})^T\mbf{D}_s\ \mbf{\grad N_w}\,d\Omega^e$

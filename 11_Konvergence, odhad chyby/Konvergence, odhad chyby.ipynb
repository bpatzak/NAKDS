Metoda konečných prvků je založena na diskretizaci původní spojité konstrukce soustavou prvků (nebo obecněji na diskretizaci slabé formulace řídících rovnic) $\Rightarrow$ výsledkem je přibližné řešení. Jeho přesnost závisí na
   * volbě typu konečného prvku,
   * velikosti jednotlivých prvků,
   * průběhu slabého řešení.


Konvergence (Cauchyho koncepce): Řekněme, že posloupnost reálných čísel $a_n$ konverguje k limitě $a$, pokud pro libovolné $\epsilon > 0$ můžeme najít takové $n_0$, že pro každé $n > n_0$ platí $| a - a_n | \leq \epsilon$. Potom píšeme

$\displaystyle{\begin{array}{rcl}
a_n \rightarrow a & 
\mbox{nebo přesněji} & 
{\lim}_{n \rightarrow \infty} a_n = a.
\end{array}}$

Tato definice jinými slovy tvrdí, že dokážeme posloupností $a_n$ aproximovat limitu $a$ s libovolnou (danou) přesností $\epsilon>0$. Pro aplikaci v MKP je důležité, zda lze slabé řešení dané úlohy $u^{ex}$ aproximovat s libovolnou přesností konečněprvkovým řešením $u^{mkp}_n$

$\displaystyle{
\boldsymbol{u}^{mkp}_{n}(\boldsymbol{x}) \stackrel{?}{\rightarrow} \boldsymbol{u}^\ex(\boldsymbol{x})  
}$


V původní definici konvergence se operuje s reálnými čísly, pro použití v MKP je však důležitá konvergence funkcí. Proto zavádíme tzv. energetickou normu funkce. Například pro tažený-tlačený prut vypočteme energetickou normu funkce posunu $u$ 

$\displaystyle{
\norm{u( x )}^2 = \int_\Omega E( x ) A( x )
\left( 
\der{u( x )}{x} 
\right)^2 \ dx,
}$

která má fyzikální význam vnitřní energie konstrukce, udělíme-li jí daný posun $u$. Zkoumáme tedy, zda platí 

$\displaystyle{
\norm{u^{mkp}_n({x})} \stackrel{?}{\rightarrow}
\norm{u^{ex}({x})}.
}$

V metodě konečných prvků většinou jednotlivá řešení neparametrizujeme počtem prvků $n$, ale typickým rozměrem prvku $h$. V ideálním případě by mělo platit 

$\displaystyle{
\lim_{h \rightarrow 0} \norm{u^{mkp}_h( x )} 
{\rightarrow}
\norm{u^\ex( x )},
}$

tedy pro libovolnou zvolenou přesnost $\epsilon >0$ jsme schopni najít takovou velikost prvku $h$, že platí

$\displaystyle{
\norm{u^\fem_h( x ) - u^\ex( x ) } < \epsilon.
}$ 

Slabé řešení jsme tedy schopni aproximovat s libovolnou přesností v
energetické normě.

Pro zajištění této podmínky musí bázové funkce splňovat následující podmínky
   * dostatečnou hladkost: neznámé funkce musí být aproximovány bázovými funkcemi, které mají spojité derivace řádu o jedna vyššího, než se objevuje ve slabém řešení, 
   * spojitost: aproximované funkce musí být spojité jak uvnitř prvku, tak na hranicích mezi prvky,
   * úplnost (pro úlohy teorie pružnosti): aproximační funkce musí být schopny
       * reprezentovat přemístění prvku jako tuhého tělesa bez vzniku deformací,
       * popsat stav konstantní deformace – vyloučení parazitních posunů (spurious modes).

Prvek, jehož bázové funkce splňují podmínky jak spojitosti, tak úplnosti se nazývá konformní. Pak platí

$\displaystyle{
\norm{u^{mkp}_h( x ) - u^\ex( x ) } \searrow 0
}$

a hovoříme o tzv. monotónní konvergenci. Pokud je splněna podmínka úplnosti, ale není spojena podmínka spojitosti, prvek se nazývá nekonformní. Řešení pak konverguje, ale konvergence nemusí být monotónní. U nekonformních prvků je rigorózní analýza splnění podmínky úplnosti velmi komplikovaná, proto je často využíván tzv. patch test.


### Adaptivní techniky v MKP

Předchozí analýza specifikovala podmínky, za kterých řešení MKP konverguje ke správnému výsledku, nebylo však řečeno jak rychle. Rychlost konvergence lze ovlivnit:
   * zjemňováním sítě h → 0 – tzv. h-konvergence,
   * zvyšováním stupně polynomické aproximace p – tzv. p-konvergence MKP,
   * kombinací obou přístupů – hp-konvergence.

Z výpočetního hlediska je výhodné provádět zjemňování sítě resp. zvyšování stupně polynomu pouze tam, kde přibližné řešení dobře nevystihuje přesné řešení → adaptivní varianta MKP. Pro libovolnou adaptivní techniku je nutné znát chybu přibližného řešení

$\begin{array}{clr}
e( x ) = {u}^\fem( {x} ) - {u}^\ex( {x} ) & \mbox{ respektive } &
\norm{e( x )} = \norm{u^\fem( x ) - u^\ex( x )}
\end{array}$

Názornější veličinou je relativní chyba řešení

$\displaystyle{
\eta = \frac{\norm{e}}{\norm{u}}
}$


Přesné řešení $u^\ex$ ovšem není obecně známé, je nutno se spokojit „pouze" s odhadem chyby $^0\norm{e}$ nebo relativní chyby $^0\eta$.

K posouzení výstižnosti odhadu chyby slouží tzv. index účinnosti odhadu

$\displaystyle{
\vartheta = \frac{^0\norm{e}}{\norm{e}}.
}$


Metody odhadu, pro které platí

$\displaystyle{
\lim_{h \rightarrow 0} \vartheta = 1
}$

se nazývají (asymptoticky) efektivní.

# Odhad chyby pomocí ZZ metody

ZZ metoda byla navržena Zienkiewiczem a Zhuem (O. C. Zienkiewicz and J. Z. Zhu, A simple error estimator and adaptive procedure for practical engeneering analysis, International Journal for Numerical Methods in Engineering 24 (1987), 337–357.). Tato metoda je vhodná pro $h$-adaptivní variantu metody konečných prvků, je velmi jednoduchá na výpočet, řešení odhadu vychází ze známých hodnot uzlových posunů $\boldsymbol{r}$.

Z MKP řešení plyne

$\boldsymbol{u}^{MKP}(\boldsymbol{x})=\boldsymbol{N}(\boldsymbol{x})\boldsymbol{r}$

$\boldsymbol{\sigma}^{MKP}(\boldsymbol{x})=\boldsymbol{DB}(\boldsymbol{x})\boldsymbol{r}$, 

kde $\boldsymbol{B}$ obsahuje derivace interpolačních funkcí. Předpokládáme, že skutečnému průběh napětí je bližší průběh

$\boldsymbol{\sigma}^*(\boldsymbol{x})=\boldsymbol{N}(\boldsymbol{x})\boldsymbol{r}_{\sigma}$

Koeficienty $\boldsymbol{r}_\sigma$ určíme tak, aby chyba mezi přibližným napětím $\boldsymbol{\sigma}^{MKP}$
a „vylepšeným napětím" $\boldsymbol{\sigma}^*$ ve smyslu nejmenších čtverců byla co
nejmenší

$\displaystyle\int_\Omega\left(\boldsymbol{\sigma}^*-\boldsymbol{\sigma}^{MKP}\right)^T\left(\boldsymbol{\sigma}^*-\boldsymbol{\sigma}^{MKP}\right)\ d\Omega \to min,$

 a tedy

$\begin{array}{rcl}
\displaystyle{\frac{\partial}{\partial {\boldsymbol{r}_\sigma}} 
\int_\Omega 
 \left( 
  \boldsymbol{N}( \boldsymbol{x} ) {\boldsymbol{r}_\sigma} - 
  \boldsymbol{B}( \boldsymbol{x} ) \boldsymbol{r}
 \right)\trn
 \left( 
  \boldsymbol{N}( \boldsymbol{x} ) {\boldsymbol{r}_\sigma} - 
  \boldsymbol{B}( \boldsymbol{x} ) \boldsymbol{r} 
 \right)
\ d\boldsymbol{x} }
& = & {0} \\
\displaystyle{\int_\Omega
 \boldsymbol{N}( x )\trn
 \left( 
  \boldsymbol{N}( x ) {\boldsymbol{r}_\sigma} - 
  \boldsymbol{B}( x ) \boldsymbol{r} 
 \right)
\ d\boldsymbol{x} }
& = & {0} 
\end{array}$

Uzlové hodnoty vylepšených napětí plynou z řešení lineární soustavy rovnic

$\displaystyle{
\begin{array}{rcl}
\displaystyle{\left(\int_\Omega \boldsymbol{N^TN}\ d\Omega\right) \boldsymbol{r}_\sigma}& = &\displaystyle{\left(\int_\Omega \boldsymbol{N^T\sigma}^{MKP} d\Omega\right)}\\
\boldsymbol{Ar}_\sigma& = &\boldsymbol{b}
\end{array}
}$


## Příklad 1 : tažený tlačený prut

<img src="figures/1dbar_seminar.png">
A dále budeme uvažovat $f_x = 1$, $l=3$, $EA=1$

### Analytické řešení:
Staticky určitá kce, průběh napětí určíme z podmínky rovnováhy
$\displaystyle{\sigma(x)=\frac{f_x(l-x)}{A},\ \ x\in(0, l)}$

Pusun určíme integrací deformace:
$\displaystyle{u(x) = \int_0^x \frac{\sigma(s)}{E} ds = \left[\frac{f_x ls-\frac{f_xs^2}{2}}{EA}\right]_0^x = -\frac{f_x}{EA}\left(\frac{-x^2}{2}-lx\right)}$

### Řešení MKP


```octave
#pocet prvku na delku
n = 5


l = 3/n;

E = 1;
A = 1;

# mkp reseni
ki = (E*A/l)*[1 -1; -1 1];
fi = [l^2/(2*l) l^2/(2*l)];
K = zeros (n+1);
F = zeros (n+1, 1);
for i=1:n
    loc = [i i+1];
    K(loc, loc) += ki;
    F(loc)+= fi';
endfor
u = K(2:n+1, 2:n+1)\(F(2:n+1,1));
U = [0 ; u]

#plot analytical solution 
hold on;
x = 0:0.1:3;
plot (x, -x.^2/2+3*x, "b;u;", x, -x+3, "r;N;")

sig = zeros(n,1);
#evaluate stress and plot obtained solution
for i=1:n
    eps = (U(i+1)-U(i))/l;
    sig(i) = E*eps;
    N = A*E*eps;
    plot ([(i-1)*l i*l], [U(i) U(i+1)], "b--")
    plot ([(i-1)*l i*l], [N N], "r--")
endfor

```

## Určení vyhlazeného napětí $\sigma^*$
$\displaystyle{\boldsymbol{A}^e=\int_l}\left[\begin{array}{r}
N_1\\
N_2
\end{array}\right]
\left[\begin{array}{rr}
N_1 &
N_2
\end{array}\right]\ dx = \left[\begin{array}{rr}
l/3& l/6\\
l/6& l/3
\end{array}\right] =
\left[\begin{array}{rr}
l/2& \\
& l/2
\end{array}
\right]
$



$
\displaystyle{\boldsymbol{b}^e=\int {N}^T\sigma^e\ dx = }\left\{\begin{array}{r}
\sigma^el/2 \\
\sigma^el/2
\end{array}\right\}
$


Pro uzlové hodnoty vyhlazeného napětí pak platí
$Ar_\sigma^ = b$

kde $A$ se získá lokalizací $A_i$, $b$ lokalizaci $b_i$.

```octave
# evaluate recovered stresses
Ai = [l/2, 0; 0, l/2];
bi = [l/2;l/2];
A = zeros(n+1);
B = zeros(n+1,1);
for i=1:n
    loc = [i i+1];
    A(loc, loc) += Ai;
    B(loc)+= sig(i)*bi;
endfor
sig2 = A\B;

#plot analytical solution 
hold on;
x = 0:0.1:3;
plot (x, -x+3, "r;Sigma;")

#plot recovered stresses
hold on;
for i=1:n
    plot ([(i-1)*l i*l], [sig(i) sig(i)], "r--")
    plot ([(i-1)*l i*l], [sig2(i) sig2(i+1)], "o--")
endfor
```

### Odhad chyby 

$\displaystyle{
\vert e\vert = \int_\Omega \frac{1}{E( x )} \left( \sigma_{x}^*( x ) -
\sigma^{mkp}_x( x ) \right)^2 dx.
}$

$\displaystyle{
\vert e\vert = \sum_e\int_{l_e} \frac{1}{E( x )} \left( \sigma_{x}^*( x ) -
\sigma^{mkp}_x( x ) \right)^2 dx.
}$



```octave
# numericka integrace 
w = [1.0; 1.0];
xi=[-1./sqrt(3); 1/sqrt(3)];

#prispevek do chyby po elementech
e = zeros(n,1);
err = 0.0;
hold on;
for i=1:n
    for ip=1:2
        sig2xi = (1-xi(ip))/2*sig2(i)+(1+xi(ip))/2*sig2(i+1);
        dsig = sig2xi-sig(i);
        e(i) += (dsig^2/E)*w(ip)*l/2;
    endfor
    err = err+e(i);
endfor
e
err
for i=1:n
    plot ([(i-1)*l i*l], [e(i) e(i)], "o--")
endfor 
```

## Příklad 2: Rovinná napjatost - konzola zatižená silou
<img src="figures/2d_konzola.png">
E = 3000
l=1.0, h=0.1

```octave
function [ke,dbe,de,be] = plane_stress (xe,ye,E,nu) 
    Ae=(1/2)*((xe(2)*ye(3)-xe(3)*ye(2))-(xe(1)*ye(3)-xe(3)*ye(1))+(xe(1)*ye(2)-xe(2)*ye(1))) ;
    y23 = ye(2)-ye(3); y31 = ye(3)-ye(1); y12 = ye(1)-ye(2);
    x32 = xe(3)-xe(2); x13 = xe(1)-xe(3); x21 = xe(2)-xe(1);

    be = (1/(2*Ae))*[ y23, 0, y31, 0, y12, 0 ;
                      0, x32, 0, x13, 0, x21 ;
                      x32,y23,x13,y31,x21,y12] ;
                      
    de = (E/(1-nu*nu))*[ 1, nu, 0;
                        nu, 1, 0;
                        0, 0, (1-nu)/2 ];
    dbe = de*be ;
    ke = Ae*be'*dbe;
end

h=0.1;
l=1.0;

%pocet prvku
nrow=4
ncol=10

nelem=2*nrow*ncol;
nnodes=(nrow+1)*(ncol+1);
E=3000; 
nu = 0.1;
id=zeros(nnodes, 2);
x = zeros(nnodes,1);
y = zeros(nnodes,1);
enodes = zeros(nelem, 3);

#generovani uzlu
for c=1:ncol+1
    for r=1:nrow+1
        n = (c-1)*(nrow+1)+r;
        x(n)=(c-1)*l/ncol;
        y(n)=(r-1)*h/nrow;
        id(n,:) = [(n-1)*2+1 (n-1)*2+2];
    endfor
endfor
#x
#y
#id

#generovani prvku
for c=1:ncol
    for r=1:nrow
        en = 2*((c-1)*nrow+r)-1;
        n = (c-1)*(nrow+1)+r;
        enodes(en,:) = [n,n+nrow+1, n+nrow+2];
        enodes(en+1, :) = [n, n+nrow+2, n+1];
    endfor
endfor

#enodes
# zobrazeni site
hold on;
triplot(enodes, x, y);
```

```octave
#reseni ulohy

ke = zeros(6,6);
dbe = zeros(3,6);
be = zeros(3,6);
de = zeros(3,3);

neq = nnodes*2;
k = zeros(neq,neq);
f = zeros(neq,1);

for i = 1:nelem
    xe = [x(enodes(i,:))] ;
    ye = [y(enodes(i,:))] ;
    [ke,dbe,de,be] = plane_stress(xe,ye,E,nu) ;
    lm = [id(enodes(i,1),:), id(enodes(i,2),:), id(enodes(i,3),:)];
    k(lm, lm) = k(lm, lm) + ke;
end

#zatizeni 
for r=1:nrow+1
    eq = id((ncol)*(nrow+1)+r, 2);
    if (r==1 || r==nrow+1)
        f(eq) = 1/nrow/2.;
    else
        f(eq) = 1/nrow;
    endif
endfor
f ;

%reseni posunu
eq = id(nrow+2, 1);
kuu=k(eq:neq, eq:neq);
fuu=f(eq:neq);
u=kuu\fuu;
u;

ug = zeros(neq,1);
ug(eq:neq)=u(:)

%dopocteni napeti
sig = zeros(nelem,3);
for ie=1:nelem
    lm = [id(enodes(ie,1),:), id(enodes(ie,2),:), id(enodes(ie,3),:)];
    re = ug(lm);
    xe = [x(enodes(ie,:))] ;
    ye = [y(enodes(ie,:))] ;
    [ke,dbe,de,be] = plane_stress(xe,ye,E,nu) ;
    sig(ie,:) = dbe*re;
endfor

    

% Plot 
xnew = x;
ynew = y;
for i = 1:nnodes
    xnew(i) = xnew(i) + ug(id(i,1));
    ynew(i) = ynew(i) + ug(id(i,2));
end

# zobrazeni site
hold on;
sig
triplot(enodes,x,y,'k')
triplot(enodes,xnew,ynew,'r')
#for ie=1:nelem
#    trimesh(enodes(ie,:),xnew(enodes(ie,:)),ynew(enodes(ie,:)),[sig(ie,1),sig(ie,1),sig(ie,1)])
#endfor
```

## Odhad chyby
Výpočet vyhlazených napětí

$\displaystyle{\boldsymbol{A}^e=\int_{A_e}
\boldsymbol{N^T N} dA = \int_{A_e}}\left[\begin{array}{r}
N_1\\N_2\\N_3
\end{array}\right] 
\left[\begin{array}{rrr}
N_1 & N_2 & N_3
\end{array}\right] 
dA
$

$\displaystyle{\boldsymbol{b}_i = \int_{A_e} \boldsymbol{N^T \sigma} dA = \int_{A_e}}\left[\begin{array}{r}
N_1\\N_2\\N_3
\end{array}\right] \boldsymbol{\sigma}(\boldsymbol{x}) dx$


$\boldsymbol{Ar}_{\sigma^*} = \boldsymbol{b}$

```octave
function [Ai, bi] = plane_stress_reco (xe,ye,sig) 

    # numericka integrace 
    w = [0.26014666666; 0.26014666666;0.26014666666; -0.28125];
    xi=[0.2, 0.2; 0.6, 0.2; 0.2, 0.6; 0.3333333333, 0.3333333333];

    Ae=(1/2)*((xe(2)*ye(3)-xe(3)*ye(2))-(xe(1)*ye(3)-xe(3)*ye(1))+(xe(1)*ye(2)-xe(2)*ye(1))) ;
    AAi=zeros(3);
    Ai=zeros(3);
    bi=zeros(3,3); #prava strana pro 3 komponenty napeti
    for ip = 1:4
        Ni=[xi(ip,1), xi(ip,2), 1-xi(ip,1)-xi(ip,2)]; # bazove funkce
        AAi = AAi + Ni'*Ni*w(ip)*2*Ae;
        bi = bi+Ni'*sig*w(ip)*2*Ae;
    endfor                
    for i=1:3
        Ai(i,i)=sum(AAi(i,:));
    endfor
end

# evaluate recovered stresses

A = zeros(nnodes);
B = zeros(nnodes,3);
for en=1:nelem
    xe = [x(enodes(en,:))] ;
    ye = [y(enodes(en,:))] ;
    [Ai,bi] = plane_stress_reco (xe,ye,sig(en,:));
    lm = [enodes(en,1), enodes(en,2), enodes(en,3)]; #kodova cisla odpovidaji cislum uzlu
    A(lm, lm) += Ai;
    B(lm,[1,2,3])+= bi;
endfor
sig2 = A\B

if (0)
    hold on;
    caxis ("auto")
    colorbar;
    for en=1:nelem
        xe = [x(enodes(en,:));  x(enodes(en,1))];
        ye = [y(enodes(en,:));  y(enodes(en,1))];
        sig2e = [sig2(enodes(en,:),1); sig2(enodes(en,1),1)];
        patch (xe, ye, sig2e);
    endfor
endif

```

## Odhad chyby
$
\displaystyle \vert\vert e \vert\vert = \int\left(\boldsymbol{\sigma}^*(\boldsymbol{x})-\boldsymbol{\sigma}^{MKP}(\boldsymbol{x})\right)^T \boldsymbol{D}^{-1} \left(\boldsymbol{\sigma}^*(\boldsymbol{x})-\boldsymbol{\sigma}^{MKP}(\boldsymbol{x})\right) d\Omega
$

```octave
# numericka integrace 
w = [0.26014666666; 0.26014666666;0.26014666666; -0.28125];
xi=[0.2, 0.2; 0.6, 0.2; 0.2, 0.6; 0.3333333333, 0.3333333333];

#prispevek do chyby po elementech
e = zeros(nelem,1);
err = 0.0;
sig
for en=1:nelem
    sig2e = [sig2(enodes(en,1),:); sig2(enodes(en,2),:);sig2(enodes(en,3),:)];
    xe = [x(enodes(en,:))] ;
    ye = [y(enodes(en,:))] ;
    Ae=(1/2)*((xe(2)*ye(3)-xe(3)*ye(2))-(xe(1)*ye(3)-xe(3)*ye(1))+(xe(1)*ye(2)-xe(2)*ye(1))) ;
    [ke,dbe,de,be] = plane_stress(xe,ye,E,nu) ; # de

    for ip=1:4
        xiip = [xi(ip,1), xi(ip,2), 1-xi(ip,1)-xi(ip,2)];
        sig2xi = xiip*sig2e;# napeti v integracnim bode
        dsig = sig2xi-sig(en,:); #rozdil sigma_*-sigma_mkp
        e(en) += (dsig)*inv(de)*dsig'*w(ip)*Ae*2;
        
    endfor
    err = err+e(en);
endfor
e
err
```

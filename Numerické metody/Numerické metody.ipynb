### Numerické metody mechaniky I
Předmět je určen studentům doktorského studia, kteří neměli možnost se seznámit s numerickými metodami a zejména s metodou konečných prvků během předchozího studia. V rámci seminářů budou studenti využívat prototypové implementace v prostředí matlab/octave ilustrující problematiku na vybraných příkladech a diskutovat výsledky. 
V rámci předmětu budou studenti řešit samostatné nebo týmové úlohy.

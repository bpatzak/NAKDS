Ukážeme si řešení lineární stability rámových konstrukcí pomocí MKP.

### Diferenciální rovnice Mindlinova prutu
Využijeme znalosti Mindlinova prutu (viz kapitola [Nosníky](../03_Nosniky/Rovinné%20nosníky.ipynb)), podmínky rovnováhy budeme však sestavovat na deformované konstrukci.

<img src = Figures/beam_equlib.png width=750>

Uvažujeme malá pootočení ($\sin\alpha\approx\alpha;\;\cos\alpha=1$).

   * Vodorovná podmínka
   
     $\displaystyle{\del{dN}{dx}+f_x=0}$


   * Svislá podmínka

     $\displaystyle{     \begin{array}{l}     (V(x+dx)-V(x))+(N(x+dx))\left(\del{dw(x+dx)}{dx}\right)-N(x)\del{dw(x)}{dx} + f_z=0  \end{array} }$

     $\displaystyle{\del{dV}{dx}+\del{d}{dx}\left(N\del{dw}{dx}\right)+f_z = 0}$


   * Momentová podmínka

     $\displaystyle{\del{dM}{dx}-V = 0}$


 Uvážením $V=dM/dx;\;M=-EI(w'')$

$\displaystyle{
EI\del{d^4w}{dx}-\del{d}{dx}\left(N(x)\del{dw}{dx}\right) = f_z
}$


#### Ekvivalentní zatížení


Účinek normálové síly na deformované konstrukci můžeme vyjádřit ekvivalentním příčným zatížením.

Podmínka ekvivalence:
$\displaystyle{\downarrow:\;-N(x)w'(x)+N(x+dx)(w'(x+dx))=f_{eq}dx}$


Odtud

$\displaystyle{f_{eq}=\del{d}{dx}\left(N\del{dw}{dx}\right)}$


<img src = Figures/f_eq3.png width=350>



 
## Diferenciální rovnice problému

$\displaystyle{
\begin{array}{ll}
\del{d}{dx}\left(EA\del{d\us}{dx}\right)+f_x&=0\\[3mm]
\del{d}{dx}\left(kGA\left(\del{dw}{dx}+\varphi_y\right)\right)+\del{d}{dx}\left(N\del{dw}{dx}\right)+f_z &=0\\[3mm]
\del{d}{dx}\left(EI_y\del{d\varphi_y}{dx}\right)-kGA\left(\del{dw}{dx}+\varphi_y\right)&=0
\end{array}
}$


Okrajové podmínky

   * Kinematické podmínky: vetknutí ($\us=w=\varphi_y=0$), pevný kloub ($\us=w=0$), atd.
   * Statické okrajové podmínky: $N(x)=\bar{N}(x), V(x)=\bar{V}(x), M(x)=\bar{M}(x)$
 


 
## Slabé řešení

$\displaystyle{\begin{array}{rcl}
   0&=&\displaystyle{\int \vus \left(\del{dN}{dx}+f_x\right)dx}\\
   0&=&\displaystyle{\int \delta w \left(\del{dV}{dx}+\del{d}{dx}\left(N\del{dw}{dx}\right)+f_z\right)dx}\\
   0&=&\displaystyle{\int \delta\varphi_y \left(\del{dM}{dx}-V\right)dx}
 \end{array}}$
 
 
Pro všechny $\vus,\delta w$ a $\delta\varphi_y$ splňující kinematické okrajové podmínky.

 
Integrací per-partes, dosazením konstitutivních rovnic a uvážením okrajových podmínek dostáváme:

$\displaystyle{\begin{array}{rcl}
\displaystyle{\int\del{d(\vus)}{dx}EA\del{d\us}{dx}\,dx}&=&\displaystyle{[\vus\bar{N}]_\Gamma + \int\vus\bar{f}_x\,dx}\\
\displaystyle{\int\del{d(\delta w)}{dx}kGA\left(\del{dw}{dx}+\varphi_y\right)\,dx + \int\del{d(\delta w)}{dx}N\del{dw}{dx}\,dx}&=&\displaystyle{
          [\delta w\bar{V}]_\Gamma + \left[\delta w\bar{N}\del{dw}{dx}\right]_\Gamma + \int\delta w\bar{f}_z\,dx}\\
\displaystyle{\int\del{d(\delta\varphi_y)}{dx}EI\del{d\varphi_y}{dx}\,dx+\int\delta\varphi_ykGA\left(\del{dw}{dx}+\varphi_y\right)\,dx} &=& \displaystyle{[\delta\varphi_y\bar{M}]_\Gamma}
\end{array}}$

 
## Diskretizace MKP

   * Konstrukci rozdělíme na $n$ prvků
   * Na každém prvku budeme aproximovat vodorovný posun $u$, průhyb $w$ a pootočení $\varphi_y$ pomocí interpolačních funkcí a uzlových hodnot $u_i$, $w_i$ a $\varphi_{y_i}$

   $\displaystyle{
\begin{array}{ccc}
u(x)\approx\mbf{N}_u^e r^e& w(x)\approx\mbf{N}_w^e r^e & \varphi_y(x)\approx\mbf{N}_\varphi^e r^e
\end{array}
}$


   * Do slabého řešení potřebujeme i vyjádření derivací

   $\displaystyle{
\begin{array}{ccc}
\pard{u(x)}{x}\approx\mbf{B}_u^e r^e& \pard{w(x)}{x}\approx\mbf{B}_w^e r^e & \pard{\varphi_y(x)}{x}\approx\mbf{B}_\varphi^e r^e
\end{array}
}$


   * Obdobně i pro aproximaci váhových funkcí
   
   $\displaystyle{
\begin{array}{ccc}
\vus(x)\approx\mbf{N}_u^e \delta r^e& \delta w(x)\approx\mbf{N}_w^e \delta r^e & \delta \varphi_y(x)\approx\mbf{N}_\varphi^e \delta r^e\\
\pard{\vus(x)}{x}\approx\mbf{B}_u^e \delta r^e& \pard{\delta w(x)}{x}\approx\mbf{B}_w^e \delta r^e & \pard{\delta \varphi_y(x)}{x}\approx\mbf{B}_\varphi^e \delta r^e
\end{array}
}$
 
 
   * Dosazením do slabého řešení
 
   $\displaystyle{\begin{array}{rcl}
\sum_e\delta\mbf{r}^{eT} \left(\int\mbf{B}_u^{eT}EA\mbf{B}_u^e\ dx\mbf{r}^e- [\mbf{N}_u^{eT}\bar{N}]_\Gamma - \int\mbf{N}_u^{eT}\bar{f}_x\,dx\right)&=&0\\
\sum_e\delta\mbf{r}^{eT}\left(\int\mbf{B}_w^{eT}kGA(\mbf{B}_w^e\mbf{r}^e+\mbf{N}_\varphi^e\mbf{r}^e)\,dx
+\int\mbf{B}_w^{eT} N \mbf{B}_w^e\mbf{r}^e\,dx-\right.
\left. [\mbf{N}_w^{eT}\bar{V}]_\Gamma - \int\mbf{N}_w^{eT}\bar{f}_z\,dx\right)&=&0\\
\sum_e\delta\mbf{r}^{eT}\left(\int\mbf{B}_\varphi^{eT}EI\mbf{B}_\varphi^e\,dx\ \mbf{r}+\int\mbf{N}_\varphi^{eT}kGA(\mbf{B}_w^{e}\mbf{r}^e+\mbf{N}_\varphi^e\mbf{r}^e)\,dx\right. \left.-[\mbf{N}_\varphi^{eT}\bar{M}]_\Gamma\right)&=&0
\end{array}}$


   * V kompaktním zápisu pak

   $\displaystyle{\begin{array}{rcl}
    \sum_e\delta\mbf{r}^{eT}\left(\mbf{K}^e_{uu}\mbf{r}^e - \mbf{R}_u^e\right) &=& 0\\
    \sum_e\delta\mbf{r}^{eT}\left(\mbf{K}^e_{ww}\mbf{r}^e+\mbf{K}_{w\varphi}^e\mbf{r}^e + \mbf{K}_{\sigma}\mbf{r}^e-\mbf{R}_w^e\right) &=& 0\\
    \sum_e\delta\mbf{r}^{eT}\left(\mbf{K}^e_{\varphi w}\mbf{r}^e+\mbf{K}_{\varphi\varphi}^e\mbf{r}^e - \mbf{R}_\varphi^e\right) &=& 0
\end{array}}$


   * Lokalizací příspěvků jednotlivých prvků 
   
   $\displaystyle{ \mbf{\delta\mbf{r}}^T \left([\mbf{K}+\mbf{K}_{\sigma} ]\mbf{r} - \mbf{R}\right) = \mbf{0} }$

   kde 

   * $\mbf{K}$ je matice tuhosti $\mbf{K}=\mbf{K}_{uu} + \mbf{K}_{ww} + \mbf{K}_{w\varphi}+\mbf{K}_{\varphi w} + \mbf{K}_{\varphi\varphi}$
   * $\mbf{K}_{\sigma}$ je matice počátečních napětí
   * Všimněme si, že $\mbf{K}_{w\varphi}^T = \mbf{K}_{\varphi w}$ a tedy matice tuhosti $\mbf{K}$ je opět symetrická
   * Opět potřeba ošetřit smykové zamknutí
 

 
## Lineární stabilita

   * Při řešení lineární stability zatížíme konstrukci tzv. referenčním zatížením $F^{r}$, kterému odpovídá jisté rozložení normálových sil
   
   $\displaystyle{(\mbf{K}+\mbf{K_\sigma^{r}})\mbf{r}=\mbf{F}^{r}}$


   * Pro jinou, proporcionálně zvětšenou úroveň zatížení $\mbf{F}=\lambda\mbf{F}^r$ platí $\mbf{K}_\sigma=\lambda\mbf{K}_\sigma^r$. 
   To je důsledek linearity probému, odkud také plyne, že matice tuhosti bude stejná i pro změněnou úroveň zatížení. A tedy

   $\displaystyle{  (\mbf{K}+\lambda\mbf{K}_\sigma^r)\mbf{r}=\lambda\mbf{F}^r
}$


   * Pokud konstrukce může při dané úrovni zatížení vybočit, pak pro daný rovnovážný stav existuje více než jeden vektor posunutí, splňující podmínky rovnováhy

   $\displaystyle{    (\mbf{K}+\lambda\mbf{K}_\sigma^r)(\mbf{r}+\delta \mbf{r})=\lambda\mbf{F}^r
  }$


   * Odečtením předchozích dvou rovnic obdržíme

   $\displaystyle{(\mbf{K}+\lambda\mbf{K}_\sigma^r) (\delta \mbf{r})=\mbf{0}}$


   * A to je tzv. zobecněný problém vlastních čísel, kde vlastní čísla $\lambda$ určují úroveň referenčního zatížení, při které dojde k vybočení kce (ztrátě stability). Vlastní vektory určují tvar příslušný vybočení, ne však jeho amplitudu, tu lineární stabilitní problém nemůže určit, vyžadovalo by to plně nelineární výpočet.
 

### Propocionální a neproporcionální zatížení

   * V praktických úlohách se často setkáme s tím, že zatížení je nutné rozdělit na část, která se proporcionálně mění (užitné zatížení,..) a část, jenž se nemění (vlastní tíha, ...). 
   * Celkové zatížení $\mbf{F}$ potom rozdělíme na část stálou $\mbf{F}_n$, které odpovídají normálové síly $N_n$ a část proporcionální $\mbf{F}_p$, které odpovídají normálové síly $N_p$.
   * Podmínka rovnováhy pak má tvar

   $\displaystyle{(\mbf{K}+\mbf{K}_{\sigma}(N_n)+\lambda\mbf{K}_{\sigma}(N_p))\delta\mbf{r} = \mbf{0}}$


   * Pokud označíme $\mbf{K}^*=\mbf{K}+\mbf{K}_{\sigma}(N_n)$, zahrnující podmínky rovnováhy na nedeformované konstukci a vliv ekvivalentního příčného zatížení od stálého zatížení. Pak můžeme psát

   $\displaystyle{(\mbf{K}^*+\lambda\mbf{K}_{\sigma}(N_p))\delta\mbf{r} = \mbf{0}}$
 


 
## Problém vlastních čísel - znaménková konvence

   * V předchozí části jsme formulovali úlohu lineární stability ve formě rovnice

   $\displaystyle{(\mbf{K}+\lambda\mbf{K}_\sigma^r) (\delta \mbf{r})=\mbf{0}}$


   * Matematicky jde o zobecněný problém vlastních čísel, který bývá formulován jako

   $\displaystyle{(\mbf{A}-\lambda\mbf{B}) (\delta \mbf{r})=\mbf{0}}$


   * Abychom dostali kompatibilní vyjádření, můžeme
  
     * matici počátečních napětí přenásobit hodnotou -1:
     
       $\displaystyle{(\mbf{K}-\lambda(-1)\mbf{K}_\sigma^r(N)) (\delta \mbf{r})=\mbf{0},}$
     
     * nebo matici  počátečních napětí sestavit pro opačnou hodnotu normálových sil,
    tedy jako kladnou uvažovat tlakovou sílu
     
       $\displaystyle{(\mbf{K}-\lambda\mbf{K}_\sigma^r(-N)) (\delta \mbf{r})=\mbf{0}}$   
 

 
## Řešení problému vlastních čísel

   * Obecně výpočetně velmi náročný problém
   * Z praktického hlediska je rozhodující nejnižší hodnota kritického zatížení $\lambda_1$
   * Vhodnou metodou je tzv. <b> metoda inverzních iterací </b> nebo též <b> Stodolova metoda</b>
   * Je-li $\lambda$ vlastním číslem (kritickým břemenem) a $y$ příslušným vlastním tvarem, platí

   $\displaystyle{\mbf{K}\mbf{y}=\lambda\mbf{K}_\sigma\mbf{y}}$


   * Přenásobením předchozí rovnice členem $\mbf{y}^T$ dostáváme

   $\displaystyle{\mbf{y}^T\mbf{K}\mbf{y}=\lambda\mbf{y}^T\mbf{K}_\sigma\mbf{y}}$
   
   
   * Pro hodnotu kritického břemene tedy platí
  
   $\displaystyle{
    \label{odhad_lambda}
  \lambda=\del{\mbf{y}^T\mbf{K}\mbf{y}}{\mbf{y}^T\mbf{K}_\sigma\mbf{y}}  
  }$
 
### Metoda inverzních iterací

   * V obecném případě neznáme vektor $\mbf{y}$. Určíme ho iteračním způsobem:
  
     * Zvolíme počáteční odhad tvaru vybočení $\mbf{y}^0$ (např. na základě statického výpočtu pro $\lambda$=1)
     * Konstrukci zatížíme ekvivalentními silami  $\mbf{K}_\sigma\mbf{y}^0 \Rightarrow \mbf{y}^1$
  
       $\displaystyle{\mbf{K}\mbf{y}^1=\mbf{K}_\sigma\mbf{y}^0}$
           
     * Konstrukci zatížíme aktualizovanými ekvivalentními silami $\mbf{K}_\sigma\mbf{y}^1\Rightarrow\mbf{y}^2,\cdots$
     * Iteraci ukončíme, pokud jsou dva následující tvary vybočení dostatečně blízko, např. použitím vztahu $\lambda=\del{\mbf{y}^T\mbf{K}\mbf{y}}{\mbf{y}^T\mbf{K}_\sigma\mbf{y}} .$  
 
     * Lze ukázat, že tento postup konverguje k prvnímu tvaru vybočení a poměr $\lambda=\del{\mbf{y}^T\mbf{K}\mbf{y}}{\mbf{y}T\mbf{K}_\sigma\mbf{y}} $ k prvnímu kritickému zatížení.
     * Postupné aproximace vlastních tvarů $\mbf{y}^1, \mbf{y}^2,\cdots$ je vhodné normalizovat.
 
### Řešení v Matlabu (Octave)
Kromě vlastní implementace funkce řešící problém vlastních čísel, založený např. na předchozím algoritmu můžeme s výhodou využít interní funkci v Matlabu či Octave.
 * funkce <code>qz</code> v Octave,
 * funce <code>eig</code> v Matlabu.
   
 

## Příklad: Stabilita rámu

<img src="Figures/simpleframe.png" >

```octave
% Stability of simple beam structure

nnode=4;
nelem=3;
neq=6;

% pole souradnic uzlu prvku (x1,z1,x2,z2)
xz=[ 0, 0,   0, -5; % left columns
     0, -5,   5, -5;
      5, -5,  5,  0];
% pole kodovych cisel
global lm=[0, 0, 0, 1, 2,3; % left columns
	   1, 2, 3, 4, 5, 6;
	   4, 5, 6, 0, 0, 0];

% materialove a geometricke charakteristiky prvku 
EA =21.e7*0.0054;
EI =21.e7*85.e-6;
%nulovani vektoru zatizeni, matice tuhosti
f=zeros(neq,1);
k=zeros(neq);
ks=zeros(neq);
r=zeros(neq,1);

%sestaveni matice tuhosti
for i=1:nelem;
 kel=beam2d_stiffness(xz(i,:), EA, EI);
 k=assembly(k,kel,i);  %lokalizace
end;
%vektor zatizeni
f(1)=0.001;
f(2)=1;
f(5)=1;

%reseni statiky
u=k\f;
%dopocteni normalovych sil na prvcich
s=zeros(nelem,1);
for i=1:nelem;
  r=zeros(6,1);
  for j=1:6;
    if lm(i,j) == 0
      r(j)=0.0;
    else
      r(j)=u(lm(i,j));
end;
end;
endForces = beam2d_postpro(xz(i,:), EA, EI, r);
s(i)=endForces(1);
end;
s


%sestaveni matic tuhosti a poc napeti
for i=1:nelem;
 %kel=beam2d_stiffness(xz(i,:), EA, EI);
 %k=assembly(k,kel,i);  %lokalizace
 ksl=beam2d_initialstress(xz(i,:), s(i));
 ks=assembly(ks,ksl,i);  %lokalizace    
end;
[aa, bb, q, z, v, w, e]=qz(k,-ks); %octave
%E=eig(kuu,-ksuu); %matlab
[e,v]=sorteigval (abs(e),v); %serazeni vl. cisel

r=v(:,1);
lr=(r'*k*r)/(r'*ks*r)



%print eigenvalues
e

%POSTPROCESSING
%
%plot geometry
fmt = ["-r"; "-g"; "-c"];
for ei=1:3 % cyklus pres vlastni tvary a cisla
%clearplot
hold on;
axis ij;
for i=1:nelem;
beam2d_geoplot(xz(i,:)); 
end;

%plot deformation

r=v(:,ei);
for i=1:nelem;
beam2d_defplot (xz(i,:), r, i , fmt(ei,:)); 
end;

hold off;
%pause;
end;
```
